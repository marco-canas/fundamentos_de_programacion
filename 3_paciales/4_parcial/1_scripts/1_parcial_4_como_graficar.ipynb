### Punto 1: Gráficos de Funciones Matemáticas y Personalización Básica (20 puntos)

**Enunciado:**
Grafica la función matemática $f(x) = x^3 - 3x^2 + 2$ en el intervalo de $x$ desde $-2$ hasta $4$.

**Requerimientos del gráfico:**

* Utiliza una línea de color rojo con un estilo discontinuo (dashed) y un grosor de 2.
* Añade un título descriptivo al gráfico.
* Coloca etiquetas claras en los ejes $X$ e $Y$.
* Activa la cuadrícula de fondo para facilitar la lectura de los puntos.
* Añade una leyenda que indique la fórmula de la función graficada.

---


```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Definir el intervalo de x (desde -2 hasta 4)
# Usamos linspace para generar 400 puntos y obtener una curva suave
x = np.linspace(-2, 4, 400)

# 2. Definir la función matemática f(x) = x^3 - 3x^2 + 2
y = x**3 - 3*x**2 + 2

# 3. Crear la figura y el gráfico
plt.figure(figsize=(8, 6))

# 4. Graficar la función aplicando los requerimientos:
# - Color rojo ('red')
# - Estilo discontinuo (linestyle='--')
# - Grosor de 2 (linewidth=2)
# - Leyenda matemática (label)
plt.plot(x, y, color="red", linestyle="--", linewidth=2, label=r"$f(x) = x^3 - 3x^2 + 2$"
)

# 5. Añadir un título descriptivo al gráfico
plt.title("Gráfica de la Función Cúbica f(x)", fontsize=14, fontweight="bold")

# 6. Colocar etiquetas claras en los ejes X e Y
plt.xlabel("Eje X", fontsize=12)
plt.ylabel("Eje Y / f(x)", fontsize=12)

# 7. Activa la cuadrícula de fondo
plt.grid(True, linestyle=":", alpha=0.6)

# 8. Añadir la leyenda en la ubicación óptima
plt.legend(loc="upper center", fontsize=11)

# 9. Mostrar el gráfico en pantalla
plt.savefig("C:\\Users\\marco\\Downloads\\grafica_funcion_cubica.png", dpi=300)  # Guardar la figura con alta resolución
plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt 

x = np.linspace(-2, 4, 400) 
y = x**3-3*x**2 + 2 
plt.figure(figsize = (8, 6))
plt.plot(x, y, color = 'red', linestyle = '--', linewidth = 2, label = r"$f(x) = x^{3} - 3x^{2} + 2$")
plt.title('Gráfica de la función cúbica f(x)', fontsize = 14, fontweight = 'bold') 
plt.xlabel('Eje x', fontsize = 12)
plt.ylabel("Eje y / f(x)", fontsize = 12)
plt.grid(True, linestyle = ':', alpha = 0.6) 
plt.savefig("C:\\Users\\marco\\Downloads\\grafica_funcion_cubica.png", dpi = 300)
plt.legend(loc = 'upper center', fontsize = 11)
plt.show()  
```


### Punto 2: Comparación de Datos Mediante Subgráficos (Subplots) (20 puntos)

**Enunciado:**
Imagina que estás analizando las ventas de dos productos diferentes (Producto A y Producto B) a lo largo de los primeros 6 meses del año. Crea una sola figura que contenga dos subgráficos de líneas dispuestos horizontalmente (1 fila, 2 columnas) para comparar sus rendimientos.

* **Datos de meses:** `['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun']`
* **Ventas Producto A:** `[150, 210, 180, 320, 290, 410]`
* **Ventas Producto B:** `[200, 170, 240, 220, 310, 380]`

**Requerimientos de los subgráficos:**

* El subgráfico de la izquierda debe mostrar las ventas del Producto A en color azul con marcadores circulares `o`.
* El subgráfico de la derecha debe mostrar las ventas del Producto B en color verde con marcadores en forma de estrella `*`.
* Ambos subgráficos deben tener sus propios títulos, etiquetas de ejes y cuadrículas.
* Añade un título principal a toda la figura (Suptitle).

---


```python
import matplotlib.pyplot as plt

# 1. Definición de los datos proporcionados
meses = ["Ene", "Feb", "Mar", "Abr", "May", "Jun"]
ventas_producto_a = [150, 210, 180, 320, 290, 410]
ventas_producto_b = [200, 170, 240, 220, 310, 380]

# 2. Crear la figura y los subgráficos dispuestos horizontalmente (1 fila, 2 columnas)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# ==========================================
# SUBGRÁFICO IZQUIERDO: PRODUCTO A
# ==========================================
# - Color azul ('blue')
# - Marcadores circulares ('o')
ax1.plot(
    meses, ventas_producto_a, color="blue", marker="o", linewidth=2, label="Progreso"
)
ax1.set_title("Rendimiento - Producto A", fontsize=12, fontweight="bold")
ax1.set_xlabel("Meses", fontsize=10)
ax1.set_ylabel("Ventas (Unidades)", fontsize=10)
ax1.grid(True, linestyle="--", alpha=0.5)  # Activa la cuadrícula
ax1.legend()

# ==========================================
# SUBGRÁFICO DERECHO: PRODUCTO B
# ==========================================
# - Color verde ('green')
# - Marcadores en forma de estrella ('*')
# - Se aumenta el tamaño del marcador (markersize) para que la estrella luzca bien
ax2.plot(
    meses,
    ventas_producto_b,
    color="green",
    marker="*",
    markersize=10,
    linewidth=2,
    label="Progreso",
)
ax2.set_title("Rendimiento - Producto B", fontsize=12, fontweight="bold")
ax2.set_xlabel("Meses", fontsize=10)
ax2.set_ylabel("Ventas (Unidades)", fontsize=10)
ax2.grid(True, linestyle="--", alpha=0.5)  # Activa la cuadrícula
ax2.legend()

# ==========================================
# CONFIGURACIÓN GLOBAL DE LA FIGURA
# ==========================================
# Añadir el título principal a toda la figura
fig.suptitle(
    "Comparativa de Ventas Semestrales: Producto A vs. Producto B",
    fontsize=14,
    fontweight="bold",
    y=1.02, # este parámetro y es para ajustar la posición vertical del título principal, 
    # evitando que se superponga con los subgráficos
)

# Ajustar automáticamente el espacio entre los subgráficos para evitar solapamientos
plt.tight_layout()
plt.savefig("C:\\Users\\marco\\Downloads\\comparativa_ventas.png", dpi=300)  # Guardar la figura con alta resolución
# Mostrar el resultado en pantalla
plt.show()
```


### Punto 3: Análisis de Distribuciones y Frecuencias (Histogramas y Dispersión) (20 puntos)

**Enunciado:**
Un equipo de investigadores ha tomado una muestra aleatoria de las estaturas (en cm) y los pesos (en kg) de 50 personas. Crea una figura que evalúe la distribución y la relación de estas variables usando dos tipos de gráficos diferentes.

* *Pistas para los datos simulados en tu código:* Puedes usar arrays de NumPy o listas simples de datos aleatorios.

**Requerimientos:**

* **Gráfico 1 (Histograma):** Representa la distribución de las estaturas utilizando un histograma. Configura el gráfico para que use 10 contenedores (bins) y cambia el color de las barras a un tono naranja con bordes negros visibles.
* **Gráfico 2 (Diagrama de Dispersión / Puntos):** Crea un diagrama de dispersión donde el eje $X$ sea la estatura y el eje $Y$ sea el peso. Utiliza marcadores transparentes (propiedad `alpha`) para identificar zonas de mayor acumulación de puntos.

---


```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Configurar una semilla aleatoria para que los datos sean reproducibles
np.random.seed(42)

# 2. Simulación de datos para 50 personas
# Estaturas: Media de 170 cm con una desviación estándar de 8 cm
estaturas = np.random.normal(loc=170, scale=8, size=50)

# Pesos: Dependiente de la estatura (a mayor estatura, mayor peso promedio) + ruido aleatorio
pesos = (estaturas - 100) * 0.9 + np.random.normal(loc=0, scale=7, size=50)

# 3. Crear la figura contenedora (1 fila, 2 columnas)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# ==========================================
# GRÁFICO 1: HISTOGRAMA (Distribución de Estaturas)
# ==========================================
# - bins=10: Define los 10 contenedores solicitados
# - color='orange': Barras de color naranja
# - edgecolor='black': Bordes negros visibles para delimitar los bins
ax1.hist(estaturas, bins=10, color="orange", edgecolor="black", alpha=0.85)

ax1.set_title("Distribución de Estaturas", fontsize=12, fontweight="bold")
ax1.set_xlabel("Estatura (cm)", fontsize=10)
ax1.set_ylabel("Frecuencia (Número de personas)", fontsize=10)
ax1.grid(True, linestyle=":", alpha=0.6)

# ==========================================
# GRÁFICO 2: DIAGRAMA DE DISPERSIÓN (Estatura vs. Peso)
# ==========================================
# - Eje X: estaturas, Eje Y: pesos
# - alpha=0.6: Añade transparencia para detectar solapamiento y acumulación de puntos
# - color='darkcyan': Un tono azul-verdoso para contrastar con el naranja
ax2.scatter(
    estaturas, pesos, color="darkcyan", s=60, alpha=0.6, edgecolor="darkcyan"
)

ax2.set_title("Relación entre Estatura y Peso", fontsize=12, fontweight="bold")
ax2.set_xlabel("Estatura (cm)", fontsize=10)
ax2.set_ylabel("Peso (kg)", fontsize=10)
ax2.grid(True, linestyle=":", alpha=0.6)

# ==========================================
# CONFIGURACIÓN GLOBAL
# ==========================================
fig.suptitle(
    "Análisis Biométrico: Distribución y Correlación de la Muestra",
    fontsize=14,
    fontweight="bold",
    y=1.02,
)

# Ajustar márgenes para evitar que se encimen elementos
plt.tight_layout()
plt.savefig("C:\\Users\\marco\\Downloads\\analisis_biometrico.png", dpi=300)  # Guardar la figura con alta resolución
# Mostrar gráficos
plt.show()
```


### Punto 4: Representación de Categorías y Proporciones (Barras y Circular) (20 puntos)

**Enunciado:**
Se ha realizado una encuesta en una empresa para conocer el lenguaje de programación favorito de los desarrolladores. Los resultados fueron: Python (45%), JavaScript (30%), Java (15%) y C++ (10%). Representa esta información utilizando la estética de Matplotlib.

**Requerimientos:**

* **Gráfico de Barras:** Muestra las frecuencias en un gráfico de barras verticales. Cada barra debe tener un color diferente utilizando uno de los estilos predefinidos de matplotlib (por ejemplo, `plt.style.use('ggplot')` o similar).
* **Gráfico Circular (Pie Chart):** Representa los mismos datos en un gráfico circular. Asegúrate de mostrar el porcentaje automáticamente dentro de cada porción (propiedad `autopct`), resalta la porción de "Python" separándola ligeramente del resto (`explode`), y añade una sombra para darle profundidad.

---


```python
import matplotlib.pyplot as plt

# 1. Definir el estilo predefinido antes de crear la figura
# 'ggplot' asignará automáticamente colores diferentes y un fondo gris con cuadrícula blanca
plt.style.use("ggplot")

# 2. Datos de la encuesta
lenguajes = ["Python", "JavaScript", "Java", "C++"]
porcentajes = [45, 30, 15, 10]

# 3. Crear la figura y los dos subgráficos (1 fila, 2 columnas)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# ==========================================
# GRÁFICO 1: BARRAS VERTICALES
# ==========================================
# Al usar el estilo ggplot, pasar una lista de colores o dejar que los gestione el ciclo interno
# Asignamos colores explícitos para asegurar que cada barra sea diferente
colores_barras = ["#E74C3C", "#3498DB", "#2ECC71", "#F1C40F"]

barras = ax1.bar(lenguajes, porcentajes, color=colores_barras, edgecolor="none")

ax1.set_title(
    "Preferencia de Lenguajes (Frecuencia)", fontsize=12, fontweight="bold"
)
ax1.set_xlabel("Lenguajes de Programación", fontsize=10)
ax1.set_ylabel("Porcentaje de Desarrolladores (%)", fontsize=10)

# Opcional: Añadir el valor del porcentaje arriba de cada barra para mayor claridad
for barra in barras:
    yval = barra.get_height()
    ax1.text(
        barra.get_x() + barra.get_width() / 2,
        yval + 1,
        f"{yval}%",
        ha="center",
        va="bottom",
        fontweight="bold",
    )


# ==========================================
# GRÁFICO 2: GRÁFICO CIRCULAR (PIE CHART)
# ==========================================
# - explode: Separa la primera porción (Python) un 10% del centro (0.1)
# - autopct: Muestra el porcentaje con un decimal automáticamente
# - shadow=True: Añade la sombra para dar profundidad
# - startangle=140: Gira el gráfico para una mejor disposición visual
explode_config = (0.1, 0, 0, 0)

ax2.pie(
    porcentajes,
    explode=explode_config,
    labels=lenguajes,
    autopct="%1.1f%%",
    shadow=True,
    startangle=140,
    colors=colores_barras,
    textprops={"fontweight": "bold"},
)

ax2.set_title("Proporción de Preferencia", fontsize=12, fontweight="bold")


# ==========================================
# CONFIGURACIÓN GLOBAL
# ==========================================
fig.suptitle(
    "Encuesta Corporativa: Lenguaje de Programación Favorito",
    fontsize=16,
    fontweight="bold",
    color="#333333",
)

# Ajustar márgenes para que no se corten las etiquetas ni los títulos
plt.tight_layout()

# Mostrar los gráficos
plt.show()
```

En la ingeniería automotriz, la desaceleración de un vehículo durante un frenado de emergencia se puede modelar mediante la función física de velocidad: $f(t) = 120 - 15t - 2t^2$, donde $t$ es el tiempo en segundos desde que se presiona el freno (de $0$ a $5$ segundos) y $f(t)$ es la velocidad resultante en km/h. Grafica esta función en dicho intervalo.

**Requerimientos del gráfico:**

* Utiliza una línea de color gris carbón (`#34495E`), con un estilo de línea continuo, un grosor de 3 y marcadores de tipo "octágono" (`p`) colocados en cada segundo.
* Añade un título descriptivo (ej. "Curva de Desaceleración en Frenado de Emergencia").
* Coloca etiquetas claras en los ejes: Eje $X$ ("Tiempo de Frenado (segundos)") y Eje $Y$ ("Velocidad (km/h)").
* Activa la cuadrícula de fondo utilizando líneas discontinuas de color gris claro.
* Añade una leyenda que identifique la curva de telemetría del freno.

### Punto 2: Autonomía de Vehículos Eléctricos en Subgráficos (20 puntos)

**Enunciado:**
Un centro de pruebas evalúa el consumo de batería de dos prototipos de vehículos eléctricos (Modelo Eco y Modelo Sport) a diferentes velocidades constantes. Diseña una sola figura que contenga dos subgráficos de líneas dispuestos horizontalmente (1 fila, 2 columnas) para comparar su rendimiento.

* **Velocidades de prueba:** `['60 km/h', '80 km/h', '100 km/h', '120 km/h']`
* **Autonomía Modelo Eco (km):** `[520, 480, 410, 330]`
* **Autonomía Modelo Sport (km):** `[450, 400, 320, 240]`

**Requerimientos de los subgráficos:**

* El subgráfico de la izquierda debe mostrar la autonomía del Modelo Eco en color verde menta con marcadores circulares (`o`).
* El subgráfico de la derecha debe mostrar la autonomía del Modelo Sport en color naranja encendido con marcadores de diamante (`d`).
* Ambos subgráficos deben incluir obligatoriamente sus propios títulos independientes, etiquetas de ejes y cuadrículas.
* Configura un título global superior (`suptitle`) para la figura completa (ej. "Rendimiento de Autonomía vs. Velocidad de Crucero").

### Punto 3: Análisis de Emisiones e Inspección Vehicular (20 puntos)

**Enunciado:**
Un organismo ambiental analiza un lote de 100 vehículos de carga pesada. Ha registrado los "Años de antigüedad del motor" y las "Emisiones de partículas finas" (en g/km). Representa estos datos mediante dos enfoques gráficos en la misma plantilla.

* *Pistas para los datos simulados:* Genera arrays de NumPy aleatorios (ej. antigüedad entre 1 y 20 años; emisiones entre 0.1 y 2.5 g/km).

**Requerimientos:**

* **Gráfico 1 (Histograma):** Representa la distribución de la antigüedad de los vehículos evaluados. El histograma debe agrupar los datos en 10 contenedores (`bins`), con barras de color azul marino y bordes negros perfectamente perfilados.
* **Gráfico 2 (Diagrama de Puntos / Dispersión):** Diseña un diagrama de dispersión que relacione la Antigüedad (Eje X) con las Emisiones (Eje Y). Cambia la forma del marcador a una cruz de suma (`+`), ajusta un tamaño resaltado y una transparencia (`alpha=0.7`) para evaluar visualmente si los vehículos más viejos concentran mayores niveles de contaminación.
