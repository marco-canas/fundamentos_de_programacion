
# Punto 1: Modelado de Depreciación Comercial y Estética Básica (20 puntos)

**Enunciado:**
En contabilidad y finanzas minoristas, la pérdida de valor de exhibición de ciertos activos tecnológicos en inventario sigue una tasa exponencial decreciente.   

Grafica la función $f(t) = e^{-0.08t}$ que representa la tasa de valor residual del artículo en función del tiempo $t$, en un intervalo de $t$ desde $0$ hasta $30$ meses.

**Requerimientos del gráfico:**

* Utiliza una línea de color azul oscuro (`darkblue`) con un estilo de línea de puntos (estilo `':'`), y un grosor de 4.0.
* Añade un título descriptivo (ej. "Curva de Depreciación Exponencial de Activos en Inventario - Variante Alexander").
* Coloca etiquetas claras en los ejes: Eje $X$ ("Tiempo (meses)") y Eje $Y$ ("Porcentaje de Valor Residual").
* Activa la cuadrícula de fondo utilizando líneas continuas finas con una opacidad (`alpha`) de 0.3.
* Añade una leyenda que describa la curva matemática utilizando el parámetro `label` de la función `plot()`.
* Guarde el gráfico en un archivo de imagen .png en la carpeta de descargas. 



```python
import numpy as np 
import matplotlib.pyplot as plt 
incremento = 1
t = np.arange(0, 30 + incremento, incremento)
f = np.exp(-0.08*t)
plt.plot(t, f, color = 'darkblue', linestyle = ':', linewidth = 4.0, label = r'$f(x) = e^{-0.08t}$')
plt.title('Tasa de valor residual en función del tiempo')
plt.xlabel('Tiempo (meses)')
plt.ylabel('Porcentaje de valor residual')
plt.grid(alpha = 0.3)
plt.legend() 
plt.savefig(r"C:\Users\marco\Downloads\tasa_valor_residual.png")
plt.show() 
```
