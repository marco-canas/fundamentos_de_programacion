````python
import pandas as pd

# 1. Cargar el DataFrame original
ruta_csv = r"C:\Users\marco\Documentos\docencia\fundamentos_de_programacion\2_clases\unidad4\2_datos\rendimiento_estudiantes.csv"
df_estudiantes = pd.read_csv(ruta_csv)

print(f"Dimensiones iniciales del Dataset: {df_estudiantes.shape[0]} filas.")
print("-" * 60)
````

````python

# =========================================================================
# PASO A: Corrección de la Nota Final Fuera de Rango (> 5.0)
# =========================================================================
# Forzamos tipo float para evitar conflictos de tipo de datos
df_estudiantes["Nota_Final"] = df_estudiantes["Nota_Final"].astype(float)

# Filtramos las notas válidas (menores o iguales a 5.0) para calcular la mediana de control
mediana_notas_validas = df_estudiantes.loc[df_estudiantes["Nota_Final"] <= 5.0, "Nota_Final"].median()

# Sustituimos el outlier usando .loc
df_estudiantes.loc[df_estudiantes["Nota_Final"] > 5.0, "Nota_Final"] = mediana_notas_validas
print(f"Nota Final saneada: Valores anómalos reemplazados por la mediana válida ({mediana_notas_validas}).")


````

````python

#
# =========================================================================
# PASO B: Rectificación de Horas de Estudio Negativas (< 0)
# =========================================================================
# El límite lógico inferior del tiempo dedicado al estudio es cero horas
df_estudiantes.loc[df_estudiantes["Horas_Estudio_Semana"] < 0, "Horas_Estudio_Semana"] = 0
print("✔️ Horas de Estudio corregidas: Valores negativos nivelados al piso lógico de 0 horas.")


# =========================================================================
# PASO C: Tratamiento e Imputación de Celdas Vacías en Asistencia
# =========================================================================
df_estudiantes["Asistencia_Porcentaje"] = df_estudiantes["Asistencia_Porcentaje"].astype(float)

# Calculamos la mediana de la asistencia para usarla como métrica de sustitución
mediana_asistencia = df_estudiantes["Asistencia_Porcentaje"].median()

# Rellenamos los valores nulos (NaN) de forma directa
df_estudiantes["Asistencia_Porcentaje"] = df_estudiantes["Asistencia_Porcentaje"].fillna(mediana_asistencia)
print(f"✔️ Asistencia normalizada: Filas nulas imputadas con la mediana del atributo ({mediana_asistencia}%).")


# =========================================================================
# PASO D: Verificación de Coherencia Matemática y Consistencia
# =========================================================================
print("\n" + "=" * 60)
print("=== COMPROBACIÓN Y ESTADÍSTICA DESCRIPTIVA FINAL ===")
print("=" * 60)
print(df_estudiantes.describe().round(2))
````


# Sustentación de la Estrategia de Curación de Datos

En el diseño de pipelines de análisis de datos, las acciones correctivas aplicadas a las variables deben guiarse por el principio de conservación de la muestra y la mitigación de sesgos estadísticos:

**1. Criterio de Imputación de la Variable Asistencia (`Asistencia_Porcentaje`):**

* **Decisión:** Se determinó **imputar los valores faltantes utilizando la mediana** del grupo en lugar de eliminar los registros del DataFrame.
* **Justificación de Ingeniería de Datos:** Eliminar filas mediante `.dropna()` reduce artificialmente el tamaño de la muestra activa de la institución, lo cual afecta el poder estadístico de los análisis subsiguientes. Se seleccionó la **mediana** por encima de la media aritmética debido a su naturaleza robusta: la asistencia suele ser una variable con un sesgo hacia la izquierda (la mayoría de los estudiantes regulares registran asistencias altas, entre el **85%** y el **100%**), por lo que el promedio se vería arrastrado y subestimado por unos pocos alumnos desertores. La mediana preserva la tendencia central real de la masa estudiantil sin alterar de forma masiva la varianza del atributo.

**2. Corrección de la Nota Final Atípica con la Mediana Válida:**

* **Justificación:** Una nota que supera el límite de 5.0 representa una falla de software en el frontend del sistema escolar o un error tipográfico humano.
````python
import os
import matplotlib.pyplot as plt
import numpy as np  # <- Importamos numpy para calcular los extremos exactos
import pandas as pd

# 1. Configuración Estética Global
plt.style.use("ggplot")

# 2. Creación del Lienzo y Subplots
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(14, 6))

# -------------------------------------------------------------------------
# GRÁFICO IZQUIERDO: Histograma con Etiquetas en los Extremos de los Intervalos
# -------------------------------------------------------------------------
# Definimos el número de intervalos deseados
numero_bins = 10

# Calculamos los extremos de los intervalos de forma equidistante entre el mínimo y el máximo real
# 'np.linspace' genera exactamente (numero_bins + 1) puntos de corte
extremos_intervalos = np.linspace(
    df_estudiantes["Nota_Final"].min(), 
    df_estudiantes["Nota_Final"].max(), 
    numero_bins + 1
)

# Dibujamos el histograma pasándole el arreglo de extremos exactos en 'bins'
ax1.hist(
    df_estudiantes["Nota_Final"],
    bins=extremos_intervalos,  # <- CLAVE: Extremos explícitos
    color="#3498db",
    edgecolor="white",
    linewidth=1.2,
    alpha=0.85,
)

# CLAVE DE LA CORRECCIÓN: Forzamos a que los marcadores (ticks) coincidan con los extremos
ax1.set_xticks(extremos_intervalos)

# Rotamos las etiquetas si quedan muy juntas y les damos un formato decimal limpio
ax1.set_xticklabels([f"{x:.2f}" for x in extremos_intervalos], rotation=45, fontsize=9)

ax1.set_title("Distribución Global de Notas Finales", fontsize=13, fontweight="bold", pad=15)
ax1.set_xlabel("Límites de los Intervalos de Clase (Notas)", fontsize=11)
ax1.set_ylabel("Frecuencia (Número de Estudiantes)", fontsize=11)
ax1.grid(True, linestyle="--", alpha=0.5)

# -------------------------------------------------------------------------
# GRÁFICO DERECHO: Gráfico Circular de Proporción por Género (Permanece igual)
# -------------------------------------------------------------------------
conteo_genero = df_estudiantes["Género"].value_counts()
categorias = conteo_genero.index
explode_config = [0.1 if count == conteo_genero.min() else 0.0 for count in conteo_genero]
colores_pie = ["#2ecc71", "#e74c3c", "#9b59b6"][: len(categorias)]

ax2.pie(
    conteo_genero,
    labels=categorias,
    autopct="%1.1f%%",
    startangle=140,
    explode=explode_config,
    shadow=True,
    colors=colores_pie,
    textprops={"fontsize": 11, "weight": "semibold"},
)

ax2.set_title("Proporción del Estudiantado por Género", fontsize=13, fontweight="bold", pad=15)
ax2.legend(title="Géneros", loc="upper right", bbox_to_anchor=(1.15, 1))

# =========================================================================
# PASO 3: Ajustes Finales y Exportación
# =========================================================================
plt.tight_layout()

directorio_reportes = r"C:\Users\marco\Documentos\docencia\fundamentos_de_programacion\2_clases\unidad4\3_reportes"
os.makedirs(directorio_reportes, exist_ok=True)
ruta_pdf = os.path.join(directorio_reportes, "informe_composicion_estudiantil.pdf")

plt.savefig(ruta_pdf, format="pdf", dpi=300, bbox_inches="tight")
print(f"✔️ ¡Gráfico corregido y generado! Exportado en: {ruta_pdf}")

plt.show()
````

````python

# =========================================================================
# PASO A: Corrección de la Anomalía Térmica (-40.0 °C)
# =========================================================================
# Como analistas, decidimos que un valor de -40°C es un error de medición.
# Lo reemplazaremos por la mediana de la temperatura de la ciudad a la que pertenece
# ese registro. Primero identificamos qué ciudad tiene ese problema.
fila_error_temp = df_ambiental[df_ambiental["Temperatura_C"] == -40.0]
fila_error_temp.empty   
````

````python

if not fila_error_temp.empty:
    ciudad_afectada = fila_error_temp["Ciudad"].values[0]
    # Calcular la mediana de temperatura de esa ciudad (excluyendo el error de -40)
    mediana_temp_ciudad = df_ambiental[
        (df_ambiental["Ciudad"] == ciudad_afectada)
        & (df_ambiental["Temperatura_C"] > -40)
    ]["Temperatura_C"].median()

    # Aplicar la corrección usando .loc
    df_ambiental.loc[
        df_ambiental["Temperatura_C"] == -40.0, "Temperatura_C"
    ] = mediana_temp_ciudad
    print(
        f"- Corrección térmica: El registro erróneo de -40.0°C en {ciudad_afectada} "
        f"fue sustituido por su mediana local ({mediana_temp_ciudad:.1f}°C)."
    )


# =========================================================================
# PASO B: Tratamiento de Ubicaciones Huérfanas (Nulos en la columna 'Ciudad')
# =========================================================================
# Dado que no se puede deducir o inventar la ciudad de origen en un análisis regional,
# procedemos a eliminar de forma controlada los registros donde 'Ciudad' sea NaN.
df_ambiental = df_ambiental.dropna(subset=["Ciudad"])
print(
    f"- Saneamiento geográfico: Se eliminaron las filas con valores nulos en 'Ciudad'."
)


# =========================================================================
# PASO C: Validación Final del Estado de los Datos
# =========================================================================
print("\n" + "=" * 50)
print("=== COMPROBACIÓN POST-LIMPIEZA ===")
print("=" * 50)

print(f"Nuevo tamaño del dataset: {df_ambiental.shape[0]} filas.")
print("\n1. Conteo actual de valores nulos:")
print(df_ambiental.isnull().sum())

print("\n2. Nuevos estadísticos de control para la Temperatura:")
print(df_ambiental["Temperatura_C"].describe()[["min", "mean", "max"]])
````

# Celda 1: Código de Dispersión Multidimensional y Mapeo de Calor (`cmap`)


````python
import matplotlib.pyplot as plt
import pandas as pd

# 1. Configuración de la ruta y carga de datos
ruta_csv = r"C:\Users\marco\Documentos\docencia\fundamentos_de_programacion\3_paciales\4_parcial\2_datos\contaminacion_ambiental_ciudades.csv"
df_ambiental = pd.read_csv(ruta_csv)

# 2. Pipeline de limpieza previo (Idéntico a las actividades anteriores)
df_ambiental = df_ambiental.dropna(subset=["Ciudad"])
mediana_medellin = df_ambiental[
    (df_ambiental["Ciudad"] == "Medellín") & (df_ambiental["Temperatura_C"] > -40)
]["Temperatura_C"].median()
df_ambiental.loc[
    df_ambiental["Temperatura_C"] == -40.0, "Temperatura_C"
] = mediana_medellin


# =========================================================================
# CONSTRUCCIÓN DEL GRÁFICO DE DISPERSIÓN AVANZADO
# =========================================================================
plt.figure(figsize=(10, 7))

# Seteamos el estilo global
plt.style.use("seaborn-v0_8-whitegrid")

# Requerimiento: Eje X = Temperatura, Eje Y = Nivel_PM25
# Requerimiento: Tercera dimensión -> c = Humedad_Porcentaje usando el mapa de color 'viridis'
# Requerimiento: Opacidad -> alpha = 0.75 para controlar el solapamiento de puntos
dispersion = plt.scatter(
    x=df_ambiental["Temperatura_C"],
    y=df_ambiental["Nivel_PM25"],
    c=df_ambiental["Humedad_Porcentaje"],
    cmap="viridis",
    s=75,  # Tamaño de los marcadores aumentado para mejor visualización
    alpha=0.75,
    edgecolors="#333333",
    linewidths=0.8,
)

# =========================================================================
# PERSONALIZACIÓN Y ELEMENTOS DE REFERENCIA
# =========================================================================
# Añadir títulos y etiquetas con unidades del sistema internacional
plt.title(
    "Análisis de Correlación: Contaminación por PM2.5 frente a Temperatura y Humedad",
    fontsize=13,
    fontweight="bold",
    pad=15,
)
plt.xlabel("Temperatura Ambiente (°C)", fontsize=11)
plt.ylabel("Concentración de Material Particulado PM2.5 (µg/m³)", fontsize=11)

# Requerimiento: Configurar la barra de colores lateral para que la escala sea legible
cbar = plt.colorbar(dispersion)
cbar.set_label(
    "Porcentaje de Humedad Relativa (%)", fontsize=11, fontweight="bold"
)
cbar.ax.tick_params(labelsize=10)

# Ajustar los límites de los ejes para enmarcar la nube de puntos estéticamente
plt.xlim(df_ambiental["Temperatura_C"].min() - 2, df_ambiental["Temperatura_C"].max() + 2)
plt.ylim(0, df_ambiental["Nivel_PM25"].max() + 10)

# Evitar cortes en los bordes antes de mostrar
plt.tight_layout()

# Mostrar gráfico en el cuaderno
plt.savefig(r"C:\Users\marco\Documentos\docencia\fundamentos_de_programacion\3_paciales\4_parcial\3_imagenes\grafico_dispersion_contaminacion.png", \
    dpi=300)
plt.show()
````
