### **Sección I: Componente Teórico (Conceptos y Lógica)**
*Esta sección evalúa la comprensión de la naturaleza de los datos y la lógica de los operadores.*

**1. Si definimos `a = [1, 2, 3]` y `b = a`, ¿qué diferencia hay entre evaluar `a == b` y `a is b`?**
* **Respuesta:** `a == b` evalúa si los valores de ambas listas son iguales (igualdad de contenido). `a is b` evalúa si ambos nombres apuntan al mismo objeto en la memoria (identidad). En este caso, ambos son `True`.

**2. En el contexto de conjuntos ($Set$), ¿por qué no es posible acceder a un elemento mediante un índice como `mi_conjunto[0]`?**
* **Respuesta:** Porque los conjuntos en Python son estructuras **no ordenadas**. Al no haber un orden garantizado, no existe el concepto de posición o índice.

**3. ¿Cuál es el resultado de la operación lógica `(5 > 3) or (10 / 0 == 1)` y por qué no genera un error de división por cero?**
* **Respuesta:** El resultado es `True`. Python usa **evaluación de cortocircuito**: como la primera parte de la disyunción (`5 > 3`) es `True`, el intérprete ya conoce que toda la expresión será verdadera y no evalúa la segunda parte, evitando así el error.

**4. Represente la siguiente proposición lógica en Python: $P \lor (Q \land \neg R)$. Suponga que P, Q y R son booleanos.**
* **Respuesta:** `P or (Q and not R)`

**5. ¿Qué significa que una estructura de datos sea "inmutable"? Mencione un ejemplo visto en clase.**
* **Respuesta:** Significa que una vez creada, sus elementos no pueden ser modificados, añadidos o eliminados. Ejemplo: La **tupla** (`tuple`).

**6. Si tenemos la cadena `s = "UdeA"`, ¿cuál es el resultado de `s * 2` y qué operador se está sobrecargando?**
* **Respuesta:** El resultado es `"UdeAUdeA"`. Se está utilizando el operador de **repetición** (el asterisco, que usualmente es multiplicación).

**7. Explique la diferencia entre la división `/` y la división entera `//`.**
* **Respuesta:** `/` realiza una división de punto flotante (siempre devuelve un `float`), mientras que `//` devuelve el cociente entero (piso de la división), útil en algoritmos de teoría de números.

**8. Dado un diccionario `D`, ¿cuál es la restricción principal sobre las "llaves" ($keys$)?**
* **Respuesta:** Las llaves deben ser objetos **inmutables** (como strings, números o tuplas) y deben ser únicas dentro del diccionario.

**9. ¿Cómo se representa el conjunto vacío en Python y por qué no se puede usar `{}`?**
* **Respuesta:** Se representa como `set()`. No se puede usar `{}` porque ese literal está reservado para crear un diccionario vacío.

**10. Si $x = 7$ y $y = 2$, ¿cuál es el resultado de $x \% y$ y qué concepto matemático representa?**
* **Respuesta:** El resultado es `1`. Representa el **residuo** o módulo de la división entera, fundamental en aritmética modular.

**11. Defina dos vectores de $\mathbb{R}^3$ como listas, $u = [1, -2, 3]$ y $v = [4, 0, -1]$. Calcule su producto escalar usando únicamente operadores de indexación y suma.**
```python
u = [1, -2, 3]
v = [4, 0, -1]
producto_escalar = u[0]*v[0] + u[1]*v[1] + u[2]*v[2]
print(producto_escalar) # Resultado: 1
```

**13. Un estudiante pertenece al conjunto de "Álgebra" y al conjunto de "Geometría". Use operadores de conjuntos para hallar los estudiantes que están en ambas materias (Intersección).**
```python
algebra = {"Ana", "Juan", "Pedro", "Luis"}
geometria = {"Luis", "Marta", "Ana", "Diego"}
ambas = algebra & geometria # o algebra.intersection(geometria)
print(ambas)
```

**16. Dada la lista `L = [10, 20, 30, 40, 50]`, intercambie el primer elemento con el último sin crear una lista nueva.**

**17. Verifique si el número $2^{10} - 1$ es divisible por 3 usando el operador módulo.**
```python
numero = 2**10 - 1
es_divisible = (numero % 3 == 0)
print(es_divisible) # True
```

**18. Construya una tupla que represente una línea recta en el plano con el formato $(m, b)$, donde $y = mx + b$. Luego, calcule el valor de $y$ para un $x=5$.**
```python
recta = (2, -1) # y = 2x - 1
x = 5
y = recta[0] * x + recta[1]
print(y)
```

**19. Dada una lista de conjuntos `S = [{1, 2}, {2, 3}, {3, 4}]`, obtenga la unión de todos los elementos manualmente usando operadores.**
```python
S = [{1, 2}, {2, 3}, {3, 4}]
union_total = S[0] | S[1] | S[2]
print(union_total)
```
