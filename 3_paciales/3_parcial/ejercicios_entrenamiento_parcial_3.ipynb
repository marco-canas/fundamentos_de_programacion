# Bloque 1: Gestión de Datos y Persistencia (Módulo `csv` y Python nativo)

1.  **Matriz de Adyacencia y Lógica:** Crea un programa que lea un archivo CSV que representa una matriz de adyacencia de un grafo dirigido. El programa debe verificar, mediante operadores lógicos, si el grafo es reflexivo y transitivo.


Para resolver este problema con el rigor matemático que requiere tu formación, vamos a abordar las definiciones de **reflexividad** y **transitividad** desde una perspectiva computacional utilizando Python estándar (para la lectura y lógica básica) y NumPy (para la eficiencia en operaciones de matrices).

### Fundamentos Matemáticos del Algoritmo
Dado un grafo $G = (V, E)$ y su matriz de adyacencia $A$:
1.  **Reflexividad:** Un grafo es reflexivo si para todo $v \in V$, $(v, v) \in E$. En la matriz $A$, esto significa que todos los elementos de la diagonal principal deben ser $1$ (o `True`).
2.  **Transitividad:** Un grafo es transitivo si siempre que $(u, v) \in E$ y $(v, w) \in E$, entonces $(u, w) \in E$. En términos matriciales, si $A^2_{ij} > 0 \implies A_{ij} > 0$.

---

### Implementación en Python

Primero, supongamos que tienes un archivo llamado `grafo.csv` con este contenido (un grafo reflexivo y transitivo de 3 nodos):
```csv
1,1,1
0,1,0
0,0,1
```

Aquí tienes el código completo:

```python
import numpy as np
import csv

def cargar_matriz_csv(nombre_archivo):
    """Lee un archivo CSV y devuelve una matriz de NumPy de tipo entero."""
    matriz = []
    try:
        with open(nombre_archivo, 'r') as f:
            lector = csv.reader(f)
            for fila in lector:
                matriz.append([int(dato) for dato in fila])
        return np.array(matriz)
    except FileNotFoundError:
        print("Error: El archivo no existe.")
        return None

def es_reflexivo(matriz):
    """
    Verifica si la diagonal principal contiene solo 1s.
    Utiliza una máscara booleana sobre la diagonal.
    """
    diagonal = np.diag(matriz)
    # Verificamos si todos los elementos son iguales a 1
    return np.all(diagonal == 1)

def es_transitivo(matriz):
    """
    Verifica la transitividad: Si A[i,j] y A[j,k], entonces A[i,k].
    Lógicamente: A^2 debe implicar A.
    """
    n = matriz.shape[0]
    # Calculamos A^2 usando el producto de matrices
    # En un grafo, A^2[i,k] > 0 indica que hay un camino de longitud 2
    matriz_cuadrado = np.dot(matriz, matriz)
    
    # Verificamos la condición: (A^2[i,j] > 0) => (A[i,j] == 1)
    # Esto es equivalente a decir que no puede ocurrir que (A^2 > 0) Y (A == 0)
    condicion = (matriz_cuadrado > 0) & (matriz == 0)
    
    # Si hay algún elemento que cumpla la falla de transitividad, devolvemos False
    return not np.any(condicion)

# --- Ejecución del Programa ---
nombre_archivo = 'grafo.csv'
# Para propósitos de este ejemplo, crearemos el archivo si no existe
with open(nombre_archivo, 'w', newline='') as f:
    f.write("1,1,1\n0,1,0\n0,0,1")

A = cargar_matriz_csv(nombre_archivo)

if A is not None:
    print(f"Matriz de Adyacencia Cargada:\n{A}\n")
    
    reflexivo = es_reflexivo(A)
    transitivo = es_transitivo(A)
    
    print(f"¿Es el grafo reflexivo? {'SÍ' if reflexivo else 'NO'}")
    print(f"¿Es el grafo transitivo? {'SÍ' if transitivo else 'NO'}")
    
    # Conclusión Lógica
    if reflexivo and transitivo:
        print("\nEl grafo cumple con las propiedades de un preorden.")


# Explicación Técnica para Matemáticos:

1.  **Lectura con `csv` y `np.array`**: Convertimos los datos planos en un objeto tensorial de rango 2 para aprovechar el álgebra lineal computacional.
2.  **Operación `np.diag`**: Extrae el vector de la diagonal principal. Al aplicar `== 1`, generamos un vector de booleanos. `np.all` actúa como un cuantificador universal ($\forall$).
3.  **Transitividad y Producto Punto**: 
    * Recordemos que el elemento $(i, j)$ de $A^2$ cuenta el número de caminos de longitud 2 entre el nodo $i$ y el nodo $j$.
    * La lógica utilizada en `(matriz_cuadrado > 0) & (matriz == 0)` busca **contraejemplos**. Si existe al menos un par $(i, j)$ donde haya un camino indirecto de longitud 2 (`matriz_cuadrado > 0`) pero no una conexión directa (`matriz == 0`), la propiedad de transitividad falla.
4.  **Vectorización**: En lugar de usar tres ciclos `for` anidados ($O(n^3)$), `np.dot` utiliza bibliotecas optimizadas (BLAS/LAPACK) que son mucho más eficientes para matrices de gran tamaño en análisis de datos.

2.  **Serialización de Objetos Geométricos:** Define una estructura que almacene puntos en $\mathbb{R}^3$. Implementa una función que guarde estos puntos en un archivo `.txt` y otra que los recupere, calculando la norma euclidiana de cada uno al ser leídos.
3.  **Filtro de Datos de Experimentos:** Dado un archivo con resultados de un experimento de física (formato `tiempo, posición`), crea un script que use un *Context Manager* para escribir en un nuevo archivo solo aquellas muestras donde la velocidad instantánea estimada sea superior a un umbral $\epsilon$.
4.  **Limpieza de Ruido en Series:** Implementa un programa que lea un archivo con una serie de datos numéricos y genere un archivo de salida donde cada valor sea reemplazado por el promedio de sus $k$ vecinos (suavizado), manejando correctamente los punteros del archivo con `.seek()`.
5.  **Validación de Formatos:** Escribe un script que verifique si un archivo CSV de datos experimentales cumple con el "esquema": debe tener $N$ columnas y la primera columna debe ser un identificador único (clave primaria).

### Bloque 2: Computación Numérica y Álgebra Lineal (NumPy)

6.  **Transformaciones Lineales:** Crea una función que reciba una matriz $A \in \mathbb{R}^{2 \times 2}$ y un conjunto de vectores (un arreglo de NumPy). Aplica la transformación $T(v) = Av$ y utiliza Matplotlib para visualizar cómo se deforma el círculo unitario.
7.  **Sistemas de Ecuaciones:** Implementa el método de eliminación de Gauss-Jordan utilizando únicamente operaciones de arreglos de NumPy (sin usar `linalg.solve`) para resolver un sistema $Ax = b$.
8.  **Proyección Ortogonal:** Dados dos vectores $u, v \in \mathbb{R}^n$ en arreglos de NumPy, calcula la proyección de $u$ sobre $v$ utilizando el producto punto (`np.dot`). Extiéndelo para proyectar un vector sobre un subespacio definido por una base ortonormal.
9.  **Descomposición Espectral:** Escribe un script que calcule los autovalores y autovectores de una matriz simétrica y verifique numéricamente la propiedad $A = V \Lambda V^T$.
10. **Aproximación de Funciones:** Utiliza `np.linspace` para generar una partición del intervalo $[0, 2\pi]$ y aproxima la función $\sin(x)$ mediante su polinomio de Taylor de grado $n$. Compara el error cuadrático medio.
11. **Simulación de Caminatas Aleatorias:** Crea una matriz de $1000 \times 1000$ donde cada fila sea una caminata aleatoria en 1D. Usa funciones vectorizadas de NumPy para encontrar en qué paso cada caminata se alejó más del origen.
12. **Tensores y Operaciones:** Implementa una función que reciba un arreglo tridimensional (tensor de orden 3) y devuelva la "traza" de cada una de sus capas frontales utilizando `np.trace`.

### Bloque 3: Análisis de Datos y Estadística (Pandas)

13. **Análisis de Correlación:** Carga un DataFrame con datos socioeconómicos y calcula la matriz de correlación de Pearson. Identifica automáticamente qué pares de variables tienen una correlación superior a 0.85.
14. **Indexación Booleana Compleja:** Dado un DataFrame de estudiantes con columnas `promedio`, `estrato` y `edad`, filtra a aquellos que están por encima del percentil 90 en promedio pero pertenecen a los estratos 1 o 2.
15. **Agregación y Agrupamiento:** Utiliza `.groupby()` para analizar un conjunto de datos de ventas por ciudad. Calcula el total de ventas, el promedio y la desviación estándar, y ordena las ciudades de mayor a menor rendimiento.
16. **Tratamiento de Datos Faltantes (NaN):** En un DataFrame de mediciones climáticas, reemplaza los valores `NaN` de la temperatura utilizando una interpolación lineal basada en el tiempo.
17. **Normalización de Datos:** Implementa una función que reciba un DataFrame y normalice todas sus columnas numéricas al intervalo $[0, 1]$ usando la transformación Min-Max: $x_{norm} = \frac{x - x_{min}}{x_{max} - x_{min}}$.
18. **Manejo de Series Temporales:** Carga un archivo CSV que contenga fechas y precios de una acción. Convierte la columna de fecha al tipo `datetime` y extrae el promedio de precio por mes.
19. **Tablas de Contingencia:** Crea una tabla de frecuencias cruzadas entre dos variables categóricas (ej. `Nivel educativo` vs `Ocupación`) y realiza un análisis visual rápido de las tendencias.
20. **Exportación de Reportes Técnicos:** Genera un DataFrame con el análisis estadístico descriptivo completo de un experimento (`.describe()`) y expórtalo a un archivo CSV llamado `resultados_finales.csv` con separador `;`.

**Sugerencia:** Para estos ejercicios, recuerda siempre el rigor matemático: verifica dimensiones de matrices antes de operar y asegúrate de que los archivos existan antes de intentar abrirlos. ¡Mucho éxito en tu parcial!

Para complementar el entrenamiento de tus estudiantes de Matemáticas Puras, he diseñado cinco ejercicios adicionales que integran conceptos de **Álgebra Lineal** y **Cálculo** con la potencia de **Pandas** y la gestión de archivos. 

Aquí tienes los ejercicios con sus respectivas soluciones explicadas paso a paso:

---

### Ejercicio 21: El Espacio de los Datos (Manejo de Archivos y Matrices)
**Contexto:** En matemáticas, una matriz esparsa es aquella con muchos ceros. Crea un archivo que represente una "matriz de datos" de un experimento donde muchos sensores fallaron (marcando 0).
**Reto:** Generar un archivo CSV con una matriz de $10 \times 10$ donde solo la diagonal principal y la subdiagonal tengan valores (una matriz de banda). Luego, léelo y calcula cuántos elementos son distintos de cero.

```python
import csv

# 1. Creación didáctica de la matriz de banda
nombre_archivo = 'matriz_banda.csv'
n = 10

with open(nombre_archivo, 'w', newline='') as f:
    writer = csv.writer(f)
    for i in range(n):
        # Lógica matemática: a_ij != 0 si j == i o j == i-1
        fila = [round((i+j)*0.5, 2) if (j == i or j == i-1) else 0 for j in range(n)]
        writer.writerow(fila)

# 2. Procesamiento para contar elementos no nulos
no_nulos = 0
with open(nombre_archivo, 'r') as f:
    reader = csv.reader(f)
    for fila in reader:
        # Convertimos a float y usamos lógica booleana
        no_nulos += sum(1 for x in fila if float(x) != 0)

print(f"Análisis de densidad: Se encontraron {no_nulos} elementos significativos de {n*n} posibles.")
```

### Ejercicio 22: Aproximación de la Derivada (Pandas y Cálculo)
**Contexto:** La derivada es el límite de la razón de cambio. En análisis de datos, usamos diferencias finitas.
**Reto:** Dado un conjunto de datos de posición $x(t) = t^2$, utiliza Pandas para calcular la velocidad instantánea aproximada en cada punto mediante la diferencia entre registros consecutivos.

```python
import pandas as pd
import numpy as np

# Creamos un dominio t en [0, 5] con paso h = 0.1
t = np.arange(0, 5.1, 0.1)
posicion = t**2

df = pd.DataFrame({'tiempo': t, 'posicion': posicion})

# Cálculo de la velocidad: delta_x / delta_t
# .diff() calcula la resta entre la fila i e i-1
df['velocidad'] = df['posicion'].diff() / df['tiempo'].diff()

print("Primeras filas del análisis de movimiento:")
print(df.head()) # La velocidad en t=1 debería ser cercana a 2 (derivada de t^2)
```

### Ejercicio 23: Normalización de Vectores de Características (Álgebra Lineal)
**Contexto:** En análisis de datos, "normalizar" a menudo significa llevar los datos al intervalo $[0, 1]$ para que todas las variables tengan el mismo peso en un modelo.
**Reto:** Tienes un DataFrame con mediciones de tres sensores con escalas muy distintas. Aplica la transformación lineal: $x' = \frac{x - \min(x)}{\max(x) - \min(x)}$.

```python
data = {
    'Sensor_Presion': [101.3, 102.5, 100.8, 101.9],
    'Sensor_Temp': [25.5, 45.0, 30.2, 35.8],
    'Sensor_Humedad': [0.1, 0.8, 0.4, 0.9]
}
df = pd.DataFrame(data)

# Aplicación vectorizada de la fórmula (sin usar ciclos for)
df_norm = (df - df.min()) / (df.max() - df.min())

print("Matriz de datos normalizada (cada columna en escala [0, 1]):")
print(df_norm)
```

### Ejercicio 24: Detección de "Ruido" en la Señal (Análisis de Diagnóstico)
**Contexto:** En matemáticas aplicadas, el ruido son valores que se alejan demasiado del comportamiento esperado.
**Reto:** Encuentra los registros que están a más de 2 desviaciones estándar ($\sigma$) de la media ($\mu$) en un dataset de voltajes, lo cual es una técnica común de limpieza de datos.

```python
import pandas as pd

# Generamos datos con un "outlier" (valor atípico)
voltajes = [5.1, 5.2, 4.9, 5.0, 5.1, 12.5, 4.8, 5.0] # 12.5 es el ruido
df = pd.DataFrame(voltajes, columns=['voltaje'])

media = df['voltaje'].mean()
desviacion = df['voltaje'].std()

# Máscara lógica para identificar ruido
limite_superior = media + 2 * desviacion
limite_inferior = media - 2 * desviacion

outliers = df[(df['voltaje'] > limite_superior) | (df['voltaje'] < limite_inferior)]

print(f"Media: {media:.2f}, Desviación: {desviacion:.2f}")
print("Registros identificados como ruido:")
print(outliers)
```

### Ejercicio 25: Integración de Datos (Join Lineal)
**Contexto:** A menudo los datos están en archivos separados. Debes unirlos usando una "llave" común, similar a cómo mapeas elementos de dos conjuntos distintos mediante una relación.
**Reto:** Tienes un archivo con `ID_Estudiante` y `Nota`, y otro con `ID_Estudiante` y `Nombre`. Únelos en un solo DataFrame y calcula el promedio general.

```python
# Creación de DataFrames de ejemplo
df_nombres = pd.DataFrame({
    'ID': [101, 102, 103],
    'Nombre': ['Ana', 'Bernardo', 'Carla']
})

df_notas = pd.DataFrame({
    'ID': [101, 102, 103],
    'Examen_Final': [4.5, 3.8, 4.2]
})

# Operación de unión (Merge) basada en la columna ID
df_final = pd.merge(df_nombres, df_notas, on='ID')

promedio = df_final['Examen_Final'].mean()

print("Dataset consolidado:")
print(df_final)
print(f"\nEl promedio del curso de Matemáticas es: {promedio:.1f}")
```
