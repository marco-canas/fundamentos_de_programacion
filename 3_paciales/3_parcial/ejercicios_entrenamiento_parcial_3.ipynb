1.  **Matriz identidad:** Crea un programa que lea un archivo CSV que representa una matriz de identidad de orden 10 por 10. El programa debe verificar si efectivamente, la matriz dada es una matriz identidad.


Para resolver este problema con el rigor matemático que requiere tu formación, vamos a abordar las definiciones de **reflexividad** y **transitividad** desde una perspectiva computacional utilizando Python estándar (para la lectura y lógica básica) y NumPy (para la eficiencia en operaciones de matrices).



# Fundamentos Matemáticos del Algoritmo
Dado un grafo $G = (V, E)$ y su matriz de adyacencia $A$:
1.  **Reflexividad:** Un grafo es reflexivo si para todo $v \in V$, $(v, v) \in E$. En la matriz $A$, esto significa que todos los elementos de la diagonal principal deben ser $1$ (o `True`).
2.  **Transitividad:** Un grafo es transitivo si siempre que $(u, v) \in E$ y $(v, w) \in E$, entonces $(u, w) \in E$. En términos matriciales, si $A^2_{ij} > 0 \implies A_{ij} > 0$.




# Implementación en Python

Primero, supongamos que tienes un archivo llamado `grafo.csv` con este contenido (un grafo reflexivo y transitivo de 3 nodos):
```csv
1,1,1
0,1,0
0,0,1
```



Aquí tienes el código completo:


````python
import numpy as np
import csv

def cargar_matriz_csv(nombre_archivo):
    """Lee un archivo CSV y devuelve una matriz de NumPy de tipo entero."""
    matriz = [] # porque construiremos la matriz como una lista de listas. 
    with open(nombre_archivo, 'r') as f:
        lector = csv.reader(f)
        for fila in lector:
            matriz.append([int(dato) for dato in fila])
    return np.array(matriz)
````

````python
import numpy as np
np.array([[1,2],[3,4]]) + np.array([[5,6],[7,8]])
````

````python

def es_reflexivo(matriz):
    """
    Verifica si la diagonal principal contiene solo 1s.
    Utiliza una máscara booleana sobre la diagonal.
    """
    diagonal = np.diag(matriz)
    # Verificamos si todos los elementos son iguales a 1
    return np.all(diagonal == 1)

def es_transitivo(matriz):
    """
    Verifica la transitividad: Si A[i,j] y A[j,k], entonces A[i,k].
    Lógicamente: A^2 debe implicar A.
    """
    n = matriz.shape[0]
    # Calculamos A^2 usando el producto de matrices
    # En un grafo, A^2[i,k] > 0 indica que hay un camino de longitud 2
    matriz_cuadrado = np.dot(matriz, matriz)
    
    # Verificamos la condición: (A^2[i,j] > 0) => (A[i,j] == 1)
    # Esto es equivalente a decir que no puede ocurrir que (A^2 > 0) Y (A == 0)
    condicion = (matriz_cuadrado > 0) & (matriz == 0)
    
    # Si hay algún elemento que cumpla la falla de transitividad, devolvemos False
    return not np.any(condicion)

# --- Ejecución del Programa ---
nombre_archivo = 'grafo.csv'
# Para propósitos de este ejemplo, crearemos el archivo si no existe
with open(nombre_archivo, 'w', newline='') as f:
    f.write("1,1,1\n0,1,0\n0,0,1")

A = cargar_matriz_csv(nombre_archivo)

if A is not None:
    print(f"Matriz de Adyacencia Cargada:\n{A}\n")
    
    reflexivo = es_reflexivo(A)
    transitivo = es_transitivo(A)
    
    print(f"¿Es el grafo reflexivo? {'SÍ' if reflexivo else 'NO'}")
    print(f"¿Es el grafo transitivo? {'SÍ' if transitivo else 'NO'}")
    
    # Conclusión Lógica
    if reflexivo and transitivo:
        print("\nEl grafo cumple con las propiedades de un preorden.")
````

# Explicación Técnica para Matemáticos:

1.  **Lectura con `csv` y `np.array`**: Convertimos los datos planos en un objeto tensorial de rango 2 para aprovechar el álgebra lineal computacional.


2.  **Operación `np.diag`**: Extrae el vector de la diagonal principal. Al aplicar `== 1`, generamos un vector de booleanos. `np.all` actúa como un cuantificador universal ($\forall$).


3.  **Transitividad y Producto Punto**: 
    * Recordemos que el elemento $(i, j)$ de $A^2$ cuenta el número de caminos de longitud 2 entre el nodo $i$ y el nodo $j$.
    * La lógica utilizada en `(matriz_cuadrado > 0) & (matriz == 0)` busca **contraejemplos**. Si existe al menos un par $(i, j)$ donde haya un camino indirecto de longitud 2 (`matriz_cuadrado > 0`) pero no una conexión directa (`matriz == 0`), la propiedad de transitividad falla.


4.  **Vectorización**: En lugar de usar tres ciclos `for` anidados ($O(n^3)$), `np.dot` utiliza bibliotecas optimizadas (BLAS/LAPACK) que son mucho más eficientes para matrices de gran tamaño en análisis de datos.

2.  **Serialización de Objetos Geométricos:** Define una estructura que almacene puntos en $\mathbb{R}^3$. Implementa una función que guarde estos puntos en un archivo `.txt` y otra que los recupere, calculando la norma euclidiana de cada uno al ser leídos.


Para resolver este problema, utilizaremos un enfoque orientado a objetos para definir la estructura del punto en $\mathbb{R}^3$ y el módulo `math` para los cálculos geométricos. La **serialización** en este contexto consiste en convertir la estructura de datos en memoria a un formato de texto plano que pueda ser almacenado y reconstruido fielmente.

### Fundamento Matemático
Para un punto $P = (x, y, z)$, la norma euclidiana (o magnitud del vector posición) se define como:
$$\|P\| = \sqrt{x^2 + y^2 + z^2}$$

### Implementación en Python


````python
import math

class Punto3D:
    """Representa un punto en el espacio R^3."""
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z

    def __repr__(self):
        return f"Punto3D(x={self.x}, y={self.y}, z={self.z})"

def guardar_puntos(puntos, nombre_archivo):
    """
    Serializa una lista de objetos Punto3D a un archivo .txt.
    Cada línea tendrá el formato: x,y,z
    """
    try:
        with open(nombre_archivo, 'w') as archivo:
            for p in puntos:
                archivo.write(f"{p.x},{p.y},{p.z}\n")
        print(f"Éxito: Se han guardado {len(puntos)} puntos en '{nombre_archivo}'.")
    except Exception as e:
        print(f"Error al guardar: {e}")

def recuperar_y_analizar(nombre_archivo):
    """
    Lee el archivo, reconstruye los objetos y calcula su norma.
    """
    puntos_recuperados = []
    print(f"{'Punto':<30} | {'Norma Euclidiana':<15}")
    print("-" * 50)
    
    try:
        with open(nombre_archivo, 'r') as archivo:
            for linea in archivo:
                # Deserialización: de string a flotantes
                partes = linea.strip().split(',')
                if len(partes) == 3:
                    x, y, z = map(float, partes)
                    punto = Punto3D(x, y, z)
                    
                    # Cálculo de la norma
                    norma = math.sqrt(punto.x**2 + punto.y**2 + punto.z**2)
                    
                    puntos_recuperados.append(punto)
                    print(f"{str(punto):<30} | {norma:.4f}")
                    
        return puntos_recuperados
    except FileNotFoundError:
        print("Error: El archivo no existe.")
        return []

# --- Ejemplo de uso ---

# 1. Definir una lista de puntos (ejemplos matemáticos notables)
mis_puntos = [
    Punto3D(0, 0, 0),      # Origen
    Punto3D(1, 1, 1),      # Diagonal del cubo unidad
    Punto3D(3, 4, 0),      # Terna pitagórica en el plano XY
    Punto3D(-1, 2, -2)     # Punto con coordenadas negativas
]

archivo_datos = "puntos_espaciales.txt"

# 2. Guardar
guardar_puntos(mis_puntos, archivo_datos)

# 3. Leer y calcular normas
print("\nRecuperando datos y calculando normas:")
lista_final = recuperar_y_analizar(archivo_datos)
````

2.  **Deserialización**: Al leer el archivo, los datos llegan como `string`. El uso de `map(float, partes)` es una forma funcional y elegante de convertir todos los elementos de la lista a números de punto flotante simultáneamente.

3.  **Filtro de Datos de Experimentos:** Dado un archivo con resultados de un experimento de física (formato `tiempo, posición`), crea un script que use un *Context Manager* para escribir en un nuevo archivo solo aquellas muestras donde la velocidad instantánea estimada sea superior a un umbral $\epsilon$.


Para resolver este ejercicio, aplicaremos el concepto de **derivación numérica** mediante diferencias finitas. Dado que los datos son discretos (puntos en el tiempo), la velocidad instantánea en un punto $t_i$ se puede estimar como la tasa de cambio entre dos mediciones consecutivas:

$$v_i \approx \frac{x_{i} - x_{i-1}}{t_i - t_{i-1}}$$

### Implementación en Python

Este script utiliza un **Context Manager** anidado (usando la sentencia `with`) para leer el archivo de origen y escribir en el de destino de manera simultánea y segura.


````python
import csv

def filtrar_experimento(archivo_entrada, archivo_salida, umbral_epsilon):
    """
    Filtra muestras basadas en la estimación de la velocidad instantánea.
    """
    try:
        # Uso de Context Manager para manejar ambos archivos
        with open(archivo_entrada, mode='r') as f_in, \
             open(archivo_salida, mode='w', newline='') as f_out:
            
            lector = csv.reader(f_in)
            escritor = csv.writer(f_out)
            
            # Intentar leer el encabezado (si existe) y escribirlo en la salida
            try:
                encabezado = next(lector)
                escritor.writerow(encabezado)
            except StopIteration:
                return # Archivo vacío

            # Inicializar variables para el cálculo de la primera velocidad
            # Necesitamos al menos dos puntos para calcular una diferencia
            linea_previa = next(lector)
            t_prev, x_prev = map(float, linea_previa)
            
            puntos_procesados = 1
            puntos_filtrados = 0

            for fila in lector:
                t_actual, x_actual = map(float, fila)
                
                # Cálculo de la velocidad instantánea (Diferencia finita hacia atrás)
                delta_t = t_actual - t_prev
                
                if delta_t <= 0:
                    print(f"Advertencia: Intervalo de tiempo nulo o negativo en t={t_actual}")
                    continue
                
                velocidad = abs((x_actual - x_prev) / delta_t)
                
                # Criterio lógico: |v| > epsilon
                if velocidad > umbral_epsilon:
                    escritor.writerow([t_actual, x_actual])
                    puntos_filtrados += 1
                
                # Actualizar estados para la siguiente iteración
                t_prev, x_prev = t_actual, x_actual
                puntos_procesados += 1
                
        print(f"Proceso finalizado. Se conservaron {puntos_filtrados} de {puntos_procesados} muestras.")

    except FileNotFoundError:
        print("Error: El archivo de entrada no fue localizado.")
    except ValueError:
        print("Error: El archivo contiene datos no numéricos.")

# --- Preparación de datos de prueba ---
datos_ejemplo = "experimento.csv"
with open(datos_ejemplo, "w", newline="") as f:
    f.write("tiempo,posicion\n")
    f.write("0.0,0.0\n")   # v=0
    f.write("0.1,0.05\n")  # v=0.5
    f.write("0.2,0.5\n")   # v=4.5 (Alta)
    f.write("0.3,0.9\n")   # v=4.0 (Alta)
    f.write("0.4,0.95\n")  # v=0.5

# --- Ejecución del Script ---
epsilon = 2.0  # Queremos solo velocidades mayores a 2.0 unidades/s
filtrar_experimento(datos_ejemplo, "muestras_criticas.csv", epsilon)
````

### Ejercicio 21: El Espacio de los Datos (Manejo de Archivos y Matrices)
**Contexto:** En matemáticas, una matriz esparsa es aquella con muchos ceros. Crea un archivo que represente una "matriz de datos" de un experimento donde muchos sensores fallaron (marcando 0).


**Reto:** Generar un archivo CSV con una matriz de $10 \times 10$ donde solo la diagonal principal y la subdiagonal tengan valores (una matriz de banda). Luego, léelo y calcula cuántos elementos son distintos de cero.


````python
import csv

# 1. Creación didáctica de la matriz de banda
nombre_archivo = 'matriz_banda.csv'
n = 10

with open(nombre_archivo, 'w', newline='') as f:
    writer = csv.writer(f)
    for i in range(n):
        # Lógica matemática: a_ij != 0 si j == i o j == i-1
        fila = [round((i+j)*0.5, 2) if (j == i or j == i-1) else 0 for j in range(n)]
        writer.writerow(fila)

# 2. Procesamiento para contar elementos no nulos
no_nulos = 0
with open(nombre_archivo, 'r') as f:
    reader = csv.reader(f)
    for fila in reader:
        # Convertimos a float y usamos lógica booleana
        no_nulos += sum(1 for x in fila if float(x) != 0)

print(f"Análisis de densidad: Se encontraron {no_nulos} elementos significativos de {n*n} posibles.")
````

# Ejercicio 24: Detección de "Ruido" en la Señal (Análisis de Diagnóstico)
**Contexto:** En matemáticas aplicadas, el ruido son valores que se alejan demasiado del comportamiento esperado.
**Reto:** Encuentra los registros que están a más de 2 desviaciones estándar ($\sigma$) de la media ($\mu$) en un dataset de voltajes, lo cual es una técnica común de limpieza de datos.


````python
import pandas as pd

# Generamos datos con un "outlier" (valor atípico)
voltajes = [5.1, 5.2, 4.9, 5.0, 5.1, 12.5, 4.8, 5.0] # 12.5 es el ruido
df = pd.DataFrame(voltajes, columns=['voltaje'])

media = df['voltaje'].mean()
desviacion = df['voltaje'].std()

# Máscara lógica para identificar ruido
limite_superior = media + 2 * desviacion
limite_inferior = media - 2 * desviacion

outliers = df[(df['voltaje'] > limite_superior) | (df['voltaje'] < limite_inferior)]

print(f"Media: {media:.2f}, Desviación: {desviacion:.2f}")
print("Registros identificados como ruido:")
print(outliers)
````

1.  **Generador de Matrices Esparsas:** Escribe un script que genere un archivo `matriz_esparsa.csv` de $100 \times 100$. La matriz debe ser una matriz identidad más una matriz con valores aleatorios solo en la banda superior inmediata ($a_{i, i+1}$). Usa el módulo `csv` y *Context Managers*.


Para un matemático, una matriz esparsa (o rala) es fundamental en la computación científica porque permite ahorrar memoria. En este ejercicio, construiremos una matriz con una estructura de banda específica: la diagonal principal (matriz identidad) y la superdiagonal (banda superior inmediata).

Matemáticamente, estamos generando una matriz $A$ tal que:

$$a_{i,j} = 
\begin{cases} 
1 & \text{si } i = j \\
\text{uniforme}(0, 1) & \text{si } j = i + 1 \\
0 & \text{en cualquier otro caso}
\end{cases}$$

### Script de Python



````python
import csv
import random

def generar_matriz_esparsa(nombre_archivo, n):
    """
    Genera una matriz de n x n con estructura:
    - Diagonal principal = 1
    - Superdiagonal = valores aleatorios entre 0 y 1
    - Resto = 0
    """
    try:
        # Uso de Context Manager para asegurar el cierre del archivo
        with open(nombre_archivo, mode='w', newline='') as archivo_csv:
            escritor = csv.writer(archivo_csv)
            
            for i in range(n):
                fila = []
                for j in range(n):
                    if i == j:
                        # Diagonal principal (Identidad)
                        valor = 1.0
                    elif j == i + 1:
                        # Superdiagonal (Banda superior inmediata)
                        valor = round(random.uniform(0, 1), 4)
                    else:
                        # Elementos fuera de la estructura
                        valor = 0.0
                    
                    fila.append(valor)
                
                # Escribimos la fila completa en el CSV
                escritor.writerow(fila)
                
        print(f"Éxito: Se ha generado la matriz de {n}x{n} en '{nombre_archivo}'.")

    except IOError as e:
        print(f"Error al escribir el archivo: {e}")

# --- Parámetros y Ejecución ---
DIMENSION = 100
NOMBRE_FILE = "matriz_esparsa.csv"

generar_matriz_esparsa(NOMBRE_FILE, DIMENSION)

# --- Verificación opcional: Mostrar las primeras 5x5 submatriz ---
print("\nVista previa de la submatriz superior izquierda (5x5):")
with open(NOMBRE_FILE, mode='r') as f:
    lector = csv.reader(f)
    for i, fila in enumerate(lector):
        if i < 5:
            # Solo mostramos los primeros 5 elementos de las primeras 5 filas
            print([float(x) for x in fila[:5]])
        else:
            break
````
