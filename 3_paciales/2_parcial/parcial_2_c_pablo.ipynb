# Geometría Analítica y Lógica (Condicionales)   

En Geometría Vectorial, la posición relativa de un punto $P = (x, y)$ respecto a una circunferencia centrada en el origen con radio $R$ es fundamental.   

Escriba un programa que reciba las coordenadas $x$, y $y$ el valor del radio $R$. El programa debe determinar e imprimir si el punto está:
1.	Dentro de la circunferencia $x^2 + y^2 < R^2$.
2.	Sobre la frontera de la circunferencia $x^2 + y^2 = R^2$.
3.	Fuera de la circunferencia $x^2 + y^2 > R^2$.



````python
def posicion_punto(x, y, R): 
    valor = x**2 + y**2
    R_2 = R**2
    if valor < R_2:
        return "Dentro"
    elif valor == R_2:
        return "Sobre la circunferencia"
    else:  
        return "Fuera"
````

````python
posicion_punto(1, 1, 2)
````

````python
# graficar con matplotlib la circunferencia de radio R y el punto (x, y) para verificar la respuesta de la función.
import matplotlib.pyplot as plt
import numpy as np
def graficar_circunferencia(x, y, R):
    # Crear un rango de ángulos para la circunferencia
    theta = np.linspace(0, 2 * np.pi, 100)
    
    # Calcular las coordenadas de la circunferencia
    circunferencia_x = R * np.cos(theta)
    circunferencia_y = R * np.sin(theta)
    
    # Crear la figura y el eje
    plt.figure(figsize=(6, 6))
    plt.plot(circunferencia_x, circunferencia_y, label='Circunferencia de radio R')
    
    # Graficar el punto (x, y)
    plt.scatter(x, y, color='red', label='Punto (x, y)')
    
    # Configurar el gráfico
    plt.axhline(y = 0, color='black', lw=0.5)
    plt.axvline(x = 0, color='black', lw=0.5)
    plt.xlim(-R-1, R+1)
    plt.ylim(-R-1, R+1)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.title('Posición del punto respecto a la circunferencia')
    plt.legend()
    plt.grid()
    plt.savefig(r"C:\Users\marco\Downloads\circunferencia.png")  # Guardar la imagen
    plt.show()
````

````python
graficar_circunferencia(1, 1, 2)
````
