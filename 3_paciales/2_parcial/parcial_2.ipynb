### **Sección I: Componente Teórico (Conceptos y Lógica)**
*Esta sección evalúa la comprensión de las estructuras de control y la modularización.*



**1. ¿Cuál es la diferencia fundamental entre un bucle `for` y un bucle `while` en términos de terminación del ciclo?**
* **Respuesta:** El `for` se utiliza cuando el número de iteraciones es conocido o está definido por una colección (iterable). El `while` depende de una condición de verdad y se usa cuando no se sabe de antemano cuántas veces se ejecutará el código.



**2. En lógica de predicados, ¿qué estructura de Python usarías para verificar si "Existe un $x$ en el conjunto $S$ tal que $P(x)$ es verdadero"?**
* **Respuesta:** Un bucle `for` que recorra el conjunto $S$ con un condicional `if P(x):` y una sentencia `break` o un `return True` apenas se cumpla la condición.



**3. ¿Qué sucede si una función en Python no tiene una sentencia `return` explícita?**
* **Respuesta:** La función ejecuta su bloque de código y retorna automáticamente el valor `None`.



**4. Explique el concepto de "Short-circuit evaluation" en un condicional `if A and B:`.**
* **Respuesta:** Si `A` es falso, Python no evalúa `B`, ya que el resultado de la conjunción será falso independientemente del valor de `B`. Esto es crucial para evitar errores (ej. evitar división por cero).



**5. ¿Qué es el "scope" o alcance de una variable definida dentro de una función?**
* **Respuesta:** Es local. La variable solo existe y es accesible mientras la función se está ejecutando. Al terminar, la variable se destruye.



**6. ¿Cómo se representa matemáticamente la composición de funciones $f(g(x))$ en Python?**
* **Respuesta:** `resultado = f(g(x))`, asumiendo que ambas han sido definidas previamente con `def`.



**7. ¿Para qué sirve el operador `is` en comparación con `==`?**
* **Respuesta:** `==` compara igualdad de valor (contenido), mientras que `is` compara identidad de objeto (si apuntan al mismo lugar en memoria).



**8. Defina qué es una función pura en el contexto de programación funcional.**
* **Respuesta:** Es una función que, dados los mismos argumentos, siempre devuelve el mismo resultado y no tiene efectos secundarios (no modifica variables globales ni imprime en consola).



**9. ¿Cuál es la complejidad lógica de un condicional anidado frente a un `elif`?**
* **Respuesta:** El `elif` es más eficiente y legible para evaluar condiciones mutuamente excluyentes, evitando niveles innecesarios de indentación.



**10. ¿Qué hace la función `enumerate()` al iterar sobre una lista?**
* **Respuesta:** Devuelve pares ordenados (tuplas) que contienen el índice y el valor del elemento actual: $(i, x_i)$.

**11. Escriba una función `es_ortogonal(u, v)` que reciba dos listas (vectores de $\mathbb{R}^n$) y devuelva `True` si su producto escalar es 0.**


````python
def es_ortogonal(u, v):
    dot_product = sum(u[i] * v[i] for i in range(len(u)))
    return dot_product == 0

````

**13. Cree una función que reciba los coeficientes $(a, b, c)$ de una ecuación cuadrática y retorne las raíces reales usando condicionales para validar el discriminante $\Delta$.**


````python
def resolver_cuadratica(a, b, c):
    disc = b**2 - 4*a*c
    if disc < 0:
        return "Raíces complejas"
    x1 = (-b + disc**0.5) / (2*a)
    x2 = (-b - disc**0.5) / (2*a)
    return x1, x2
````

**15. Escriba una función que reciba una lista de números y devuelva una nueva lista solo con los números primos (aplique lógica de predicados).**


````python
def es_primo(num):
    if num < 2: return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0: return False
    return True

def filtrar_primos(lista):
    return [x for x in lista if es_primo(x)]
````

**16. Diseñe una función que calcule la norma euclidiana $\|v\|$ de un vector de dimensión arbitraria.**

````python
def norma_euclidiana(v):
    return sum(x**2 for x in v)**0.5
````


**17. Implemente una función que reciba una función $f$ y un punto $x$, y devuelva una aproximación de la derivada $f'(x)$ usando la definición de límite con $h = 10^{-7}$.**


````python
def derivada(f, x, h=1e-7):
    return (f(x + h) - f(x)) / h
````

**19. Escriba una función que verifique si una matriz (lista de listas) es cuadrada.**

````python
def es_cuadrada(matriz):
    filas = len(matriz)
    for fila in matriz:
        if len(fila) != filas:
            return False
    return True
````

**20. Desarrolle un programa que reciba un conjunto de puntos $(x, y)$ y determine en qué cuadrante se encuentra cada uno usando `if-elif-else`.**


````python
def determinar_cuadrante(puntos):
    for x, y in puntos:
        if x > 0 and y > 0: print(f"({x},{y}) Cuadrante I")
        elif x < 0 and y > 0: print(f"({x},{y}) Cuadrante II")
        elif x < 0 and y < 0: print(f"({x},{y}) Cuadrante III")
        elif x > 0 and y < 0: print(f"({x},{y}) Cuadrante IV")
        else: print(f"({x},{y}) Sobre un eje u origen")
````
