## Sesión 1: Gestión de Archivos y Persistencia de Datos

### 1. La Abstracción del "File Object" y Modos de Apertura
En Python, un archivo es un objeto que actúa como un flujo (*stream*) de datos. La función `open()` establece un canal de comunicación.

#### Modos de apertura (Modelado de acceso)
Para un matemático, esto es definir las **permisos y restricciones** del operador sobre el conjunto de datos:

| Modo | Nombre | Descripción | Comportamiento del Puntero |
| :--- | :--- | :--- | :--- |
| `'r'` | **Read** | Solo lectura. | Inicia al principio (0). |
| `'w'` | **Write** | Escritura. Trunca (borra) el contenido previo. | Inicia al principio (0). |
| `'a'` | **Append** | Anexar. Mantiene datos previos. | Inicia al final ($n$). |
| `'b'` | **Binary** | Modo binario (ej. para imágenes o archivos `.npy`). | Sin decodificación UTF-8. |

---

### 2. Métodos de Lectura y Escritura
La lectura se puede entender como una función que mapea el contenido del archivo a estructuras de Python (strings o listas).

* `.read(n)`: Lee $n$ caracteres. Si no se especifica, lee el archivo completo (Cuidado: carga todo en RAM).
* `.readline()`: Lee hasta encontrar un carácter de nueva línea `\n`. Ideal para archivos grandes.
* `.readlines()`: Convierte el archivo en una **lista de strings**, donde cada elemento es una línea. Útil para iterar con un `for`.

---

### 3. Anatomía del Puntero: `.seek()` y `.tell()`
Un archivo no es una estructura estática, sino una secuencia direccionable. El puntero es el índice actual de lectura/escritura.

* **`.tell()`**: Devuelve la posición actual del puntero (un entero que representa bytes desde el inicio).
* **`.seek(offset)`**: Mueve el puntero a una posición específica. Esto permite **acceso no secuencial**, similar a cómo indexamos un vector en $\mathbb{R}^n$.



---

### 4. Context Managers: El rigor en el manejo de recursos
En matemáticas, si abrimos un paréntesis, debemos cerrarlo. En programación, si abrimos un canal de E/S, debemos cerrarlo para evitar "fugas de memoria" (*memory leaks*).

La sentencia `with` implementa el protocolo de gestión de contextos. Garantiza que el método `__exit__` se ejecute (cerrando el archivo) incluso si ocurre un error en el bloque de código.

```python
# Malas prácticas (estilo imperativo)
f = open('matriz.txt', 'r')
datos = f.read()
f.close() # Si ocurre un error arriba, esta línea nunca se ejecuta.

# Práctica profesional (estilo declarativo/contextual)
with open('matriz.txt', 'r') as f:
    datos = f.read()
# El archivo se cierra automáticamente al salir de la indentación.
```

### 5. CSV y Organización de Experimentos Reproducibles
Para un matemático, la **reproducibilidad** es el equivalente a la **demostración**. Si un colega no puede ejecutar su código y obtener los mismos resultados, su análisis carece de validez científica.

#### El formato CSV (Comma Separated Values)
Es esencialmente una matriz $M_{m \times n}$ representada como texto plano. Python ofrece el módulo `csv` para manejar las sutilezas de los delimitadores y las comillas.

#### Estructura de Proyecto Sugerida
Para sus investigaciones, utilicen siempre esta jerarquía de directorios:
```text
proyecto_analisis/
├── data/               # Archivos fuente (Inmutables)
│   ├── raw/            # Datos originales
│   └── processed/      # Datos limpios tras la limpieza
├── scripts/            # Código fuente (.py)
├── results/            # Gráficos y archivos de salida (CSV, Excel)
└── README.md           # Explicación del experimento
```

### Ejercicio Práctico: El Analista de Matrices
**Contexto:** Tienes un archivo llamado `experimento.csv` que contiene una matriz de mediciones físicas.

1.  **Crea el archivo:** Escribe un script que genere una matriz identidad de $5 \times 5$ y la guarde en `data/raw/identidad.csv` usando el módulo `csv`.
2.  **Lectura y Cálculo:** Lee el archivo, convierte los datos a flotantes y calcula la traza de la matriz ($\sum_{i=1}^{n} a_{ii}$).
3.  **Persistencia:** Guarda el resultado de la traza en un nuevo archivo `results/analisis.txt` indicando la fecha y el valor obtenido.



````python
import csv

# 1. Escritura
with open('identidad.csv', 'w', newline='') as f: # lo que hace esta línea es abrir un archivo
    #llamado 'identidad.csv' en modo escritura ('w') y asegura que no se añadan líneas en blanco
    # adicionales al escribir en el archivo (newline='').
    writer = csv.writer(f) # esta línea crea un objeto escritor de CSV que se utilizará para 
    #escribir filas en el archivo.
    for i in range(5):
        fila = [1 if j == i else 0 for j in range(5)]
        writer.writerow(fila)
````


# 2. Lectura y procesamiento
traza = 0
with open('identidad.csv', 'r') as f:
    reader = csv.reader(f)
    for i, fila in enumerate(reader):
        traza += float(fila[i])

print(f"La traza calculada es: {traza}")
