Esta es una propuesta de diseño curricular estructurada para una clase intensiva o un módulo de **Python para Ciencia de Datos**. El diseño está organizado de lo fundamental (manejo de archivos) a lo aplicado (análisis de datos con Pandas).

---



## Módulo: Fundamentos de Python para Ciencia de Datos

### Sesión 1: Gestión de Archivos y Persistencia de Datos
El objetivo es dominar cómo Python interactúa con el sistema de archivos antes de automatizar procesos.

* **Apertura y Modos:** Uso de `open()` con modos `'r'`, `'w'`, `'a'` y `'b'`.
* **Lectura y Escritura:** Diferencias entre `.read()`, `.readline()` y `.readlines()`.
* **Funciones de E/S:** El puntero del archivo (`.seek()`, `.tell()`).
* **Context Managers:** Implementación de `with open(...) as file:` para evitar fugas de memoria y archivos corruptos.
* **CSV y Reproducción:** Uso del módulo `csv` nativo y la importancia de organizar carpetas (`/data`, `/scripts`, `/results`) para experimentos reproducibles.

---

### Sesión 2: Introducción al Análisis de Datos
Teoría fundamental para entender el "porqué" antes del "cómo".

* **Conceptos Clave:**
    * **Data Analysis vs. Data Analytics:** El primero busca entender el pasado/presente; el segundo usa esos insights para predecir o tomar decisiones estratégicas.
    * **¿Por qué analizar datos?:** Optimización de procesos, identificación de tendencias y mitigación de riesgos.
* **Tipos de Análisis:** Descriptivo (qué pasó), Diagnóstico (por qué pasó), Predictivo (qué pasará) y Prescriptivo (cómo hacer que pase).
* **Flujo de Proceso:** 1. Definición del problema.
    2. Recolección.
    3. Limpieza.
    4. Análisis Exploratorio (EDA).
    5. Modelado/Interpretación.
* **Tipos de Datos:** Cualitativos (nominales, ordinales) y Cuantitativos (discretos, continuos).



---

### Sesión 3: El Ecosistema Pandas
Inmersión en la biblioteca estándar de facto para manipulación de datos.

* **¿Qué es y por qué Pandas?:** Basado en NumPy, ofrece estructuras de alto nivel y herramientas de alineación de datos.
* **Estructuras de Datos:** * **Series:** Arreglos unidimensionales con etiquetas.
    * **DataFrames:** Estructuras bidimensionales (tablas) con ejes etiquetados.
* **Ingesta de Datos:**
    * Creación desde colecciones (diccionarios y listas).
    * Lectura desde fuentes externas: `pd.read_csv()`, `pd.read_excel()`, `pd.read_json()`.



---

### Sesión 4: Manipulación y Limpieza de Datos (Data Wrangling)
La fase que consume el 80% del tiempo de un científico de datos.

* **Exploración Inicial:** Uso de `.head()`, `.info()`, `.describe()` y `.shape`.
* **Selección y Filtrado:** Diferencias entre `.loc[]` (etiquetas) e `.iloc[]` (índices enteros). Filtrado booleano.
* **Limpieza Profunda:**
    * **Duplicados:** `.duplicated()` y `.drop_duplicates()`.
    * **Valores Faltantes (NaN):** Detección con `.isnull()`, eliminación con `.dropna()` e imputación con `.fillna()`.
* **Transformación de la Estructura:**
    * Agregar/Eliminar columnas y filas (`.drop()`, asignación directa).
    * Ordenamiento (`.sort_values()`) y Ranking.

---

### Sesión 5: Operaciones Avanzadas y Exportación
Consolidación de múltiples fuentes y salida de resultados.

* **Agregación y Agrupamiento:** El patrón **Split-Apply-Combine** con `.groupby()` y funciones de agregación (`.sum()`, `.mean()`, `.count()`).
* **Combinación de Datasets:**
    * **Concatenación:** `.concat()` para unir tablas por ejes (pegar una debajo de otra).
    * **Merging/Joining:** `.merge()` para realizar operaciones tipo SQL (Inner, Left, Right, Outer Join) mediante llaves comunes.
* **Persistencia de Resultados:**
    * Exportación a formatos estructurados: `.to_csv()`, `.to_excel()`, `.to_json()`.

---

### Proyecto Práctico Sugerido
**"Análisis de Experimentos Clínicos/Ventas":**
1. Cargar múltiples archivos CSV usando un loop y `os.listdir`.
2. Limpiar valores nulos y registros duplicados.
3. Unir la tabla de "Resultados" con la de "Metadatos de Usuarios" mediante un `merge`.
4. Agrupar por categoría para obtener métricas clave.
5. Exportar un reporte final en Excel.
