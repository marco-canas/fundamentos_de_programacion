# Sesión 1: Gestión de Archivos y Persistencia de Datos



### 1. La Abstracción del "File Object" y Modos de Apertura
En Python, un archivo es un objeto que actúa como un flujo (*stream*) de datos. La función `open()` establece un canal de comunicación.



## Modos de apertura (Modelado de acceso)
Para un matemático, esto es definir las **permisos y restricciones** del operador sobre el conjunto de datos:

| Modo | Nombre | Descripción | Comportamiento del Puntero |
| :--- | :--- | :--- | :--- |
| `'r'` | **Read** | Solo lectura. | Inicia al principio (0). |
| `'w'` | **Write** | Escritura. Trunca (borra) el contenido previo. | Inicia al principio (0). |
| `'a'` | **Append** | Anexar. Mantiene datos previos. | Inicia al final ($n$). |
| `'b'` | **Binary** | Modo binario (ej. para imágenes o archivos `.npy`). | Sin decodificación UTF-8. |

---



### 2. Métodos de Lectura y Escritura
La lectura se puede entender como una función que mapea el contenido del archivo a estructuras de Python (strings o listas).

* `.read(n)`: Lee $n$ caracteres. Si no se especifica, lee el archivo completo (Cuidado: carga todo en RAM).
* `.readline()`: Lee hasta encontrar un carácter de nueva línea `\n`. Ideal para archivos grandes.
* `.readlines()`: Convierte el archivo en una **lista de strings**, donde cada elemento es una línea. Útil para iterar con un `for`.

### Ejercicio Práctico: El Analista de Matrices
**Contexto:** Tienes un archivo llamado `experimento.csv` que contiene una matriz de mediciones físicas.

1.  **Crea el archivo:** Escribe un script que genere una matriz identidad de $5 \times 5$ y la guarde en `data/raw/identidad.csv` usando el módulo `csv`.
2.  **Lectura y Cálculo:** Lee el archivo, convierte los datos a flotantes y calcula la traza de la matriz ($\sum_{i=1}^{n} a_{ii}$).
3.  **Persistencia:** Guarda el resultado de la traza en un nuevo archivo `results/analisis.txt` indicando la fecha y el valor obtenido.



````python
import csv

# 1. Escritura
with open('identidad.csv', 'w', newline='') as f: # lo que hace esta línea es abrir un archivo
    #llamado 'identidad.csv' en modo escritura ('w') y asegura que no se añadan líneas en blanco
    # adicionales al escribir en el archivo (newline='').
    writer = csv.writer(f) # esta línea crea un objeto escritor de CSV que se utilizará para 
    #escribir filas en el archivo.
    for i in range(5):
        fila = [1 if j == i else 0 for j in range(5)]
        writer.writerow(fila)
````

````python

# 2. Lectura y procesamiento
traza = 0
with open('identidad.csv', 'r') as f:
    reader = csv.reader(f)
    for i, fila in enumerate(reader):
        traza += float(fila[i])

print(f"La traza calculada es: {traza}")
````

## Sesión 2: Introducción al Análisis de Datos

### 1. Conceptos Clave: La diferencia semántica y funcional
Aunque a menudo se usan como sinónimos, en el ámbito profesional tienen enfoques distintos que podemos ver desde una línea temporal:



* **Data Analysis (El qué):** Es un proceso retrospectivo. Tomamos un conjunto de datos histórico $D$ y aplicamos funciones estadísticas para describir sus propiedades. Es el equivalente a estudiar las propiedades de una función ya graficada.


* **Data Analytics (El para qué):** Es el campo más amplio que incluye al análisis pero busca la **acción**. Utiliza el razonamiento lógico-inductivo para inferir comportamientos futuros o tomar decisiones basadas en la probabilidad de éxito.



**¿Por qué analizar datos?**
Desde el álgebra lineal y el cálculo, sabemos que los sistemas complejos pueden modelarse. Analizar datos nos permite:
1.  **Optimizar:** Encontrar los máximos o mínimos de una función de costo real.
2.  **Mitigar Riesgos:** Calcular la varianza y la probabilidad de eventos atípicos (outliers).



### 2. Tipos de Análisis: Los niveles de madurez
Podemos clasificar el análisis según el valor que aporta y la complejidad técnica requerida.





| Tipo | Pregunta | Herramienta Matemática Sugerida |
| :--- | :--- | :--- |
| **Descriptivo** | ¿Qué pasó? | Estadística descriptiva (media, mediana, desviación estándar). |
| **Diagnóstico** | ¿Por qué pasó? | Correlación, análisis de causalidad, pruebas de hipótesis. |
| **Predictivo** | ¿Qué pasará? | Regresión lineal/no lineal, series de tiempo, cálculo de probabilidades. |
| **Prescriptivo** | ¿Cómo lograrlo? | Optimización restringida, teoría de juegos, investigación de operaciones. |



# 3. El Flujo de Proceso: Del problema al insight
Para un matemático, este es el **algoritmo de trabajo**. No se puede pasar a la etapa $i+1$ sin haber validado la etapa $i$.



1.  **Definición del problema:** Plantear la hipótesis. Ejemplo: "¿Existe una correlación lineal entre la inversión en I+D y el crecimiento del PIB?".


2.  **Recolección:** Obtención de datos mediante APIs, Web Scraping o bases de datos SQL (visto en la Sesión 1).


3.  **Limpieza (Data Cleaning):** El paso crítico. Tratar con datos faltantes o ruidosos. Si los datos están "sucios", cualquier modelo matemático posterior será inválido.


4.  **Análisis Exploratorio (EDA):** Aquí es donde "escuchamos" a los datos. Usamos histogramas, diagramas de dispersión y matrices de correlación.


5.  **Modelado/Interpretación:** Aplicamos el rigor matemático para concluir si la hipótesis inicial se sostiene o se rechaza.



### 4. Tipos de Datos: Clasificación Estructural
Desde la lógica y el álgebra, clasificamos los datos según su naturaleza escalar y su escala de medición.



## A. Cualitativos (Categóricos)
Representan atributos o cualidades no numéricas.
* **Nominales:** No tienen un orden intrínseco. (Ej: Color de ojos, Ciudad de origen). En computación, se suelen representar con *One-Hot Encoding*.


* **Ordinales:** Tienen una jerarquía lógica. (Ej: Nivel de satisfacción: Bajo < Medio < Alto). Existe una relación de orden $\leq$.



#### B. Cuantitativos (Numéricos)
Representan magnitudes sobre las que podemos realizar operaciones aritméticas.
* **Discretos:** Valores aislados en un conjunto contable (generalmente $\mathbb{Z}^+$). Ej: Número de hijos, cantidad de colisiones en un experimento.


* **Continuos:** Valores que pueden tomar cualquier número en un intervalo real $\mathbb{R}$. Ej: Tiempo de reacción, temperatura, presión. Aquí es donde el **Cálculo Diferencial** entra en juego para analizar tasas de cambio.

## Desarrollo Práctico: Análisis de Datos "Desde Cero"

Vamos a analizar un pequeño dataset de un experimento físico. Los datos representan:
* **Variable Nominal:** `id_sensor`
* **Variable Ordinal:** `calidad_señal` (1: Mala, 2: Regular, 3: Buena)
* **Variable Discreta:** `pulsos_detectados`
* **Variable Continua:** `voltaje_mv`

### 1. El Dataset como Estructura de Datos
Representaremos nuestros datos como una **lista de diccionarios**, una estructura común en el intercambio de datos (JSON).


````python
# Dataset: Mediciones de sensores en un laboratorio
data = [
    {"id": "S1", "calidad": 3, "pulsos": 10, "voltaje": 5.2},
    {"id": "S2", "calidad": 1, "pulsos": 15, "voltaje": 4.8},
    {"id": "S1", "calidad": 2, "pulsos": 12, "voltaje": 5.0},
    {"id": "S3", "calidad": 3, "pulsos": 8, "voltaje": 5.5},
    {"id": "S2", "calidad": 2, "pulsos": 14, "voltaje": 4.9}
]
data 
````

# 2. Análisis Descriptivo (Estadística de Tendencia Central)
Para un matemático, la media ($\bar{x}$) es el primer momento estadístico. Vamos a calcularla para el voltaje.


````python
voltajes = [d["voltaje"] for d in data]
media_voltaje = sum(voltajes) / len(voltajes)

print(f"Análisis Descriptivo: El voltaje promedio es {media_voltaje:.2f} mV")
````

### 3. Análisis de Diagnóstico (Buscando Correlaciones)
Intentemos ver si existe una relación entre los `pulsos` (discreta) y el `voltaje` (continua). Si al aumentar los pulsos, el voltaje disminuye, hay una correlación negativa.


````python
# Verificamos la relación visualmente o mediante lógica
print("Diagnóstico: Comparativa Pulsos vs Voltaje")
for d in data:
    print(f"Pulsos: {d['pulsos']} -> Voltaje: {d['voltaje']}")

# Nota: Observarán que a mayor número de pulsos (15), el voltaje es menor (4.8).
````


### 4. Análisis Predictivo (Inferencia Simple)
Usando una aproximación lineal básica (asumiendo una tasa de cambio constante $m = \frac{\Delta V}{\Delta P}$), podríamos predecir un valor.


````python
# Supongamos una relación lineal simplificada: V = -0.1 * P + 6.2
def predecir_voltaje(pulsos):
    return -0.1 * pulsos + 6.2

prediccion = predecir_voltaje(20)
print(f"Análisis Predictivo: Para 20 pulsos, se espera un voltaje de {prediccion:.1f} mV")
````

# 5. Análisis Prescriptivo (Toma de Decisiones)
Basado en la `calidad_señal`, decidimos qué sensores requieren mantenimiento inmediato.


````python
sensores_criticos = [d["id"] for d in data if d["calidad"] == 1]

print(f"Análisis Prescriptivo: Acción requerida en los sensores: {sensores_criticos}")
````

# ¿Por qué pasar a Pandas después de esto?


Como matemáticos, notarán que trabajar con listas de diccionarios para 1,000,000 de registros es ineficiente computacionalmente ($O(n)$ para muchas búsquedas). 

Para ilustrar por qué pasar de una **lista de diccionarios** a un **DataFrame** o una **Base de Datos** no es solo una cuestión de "comodidad", sino de eficiencia matemática y computacional, analicemos un caso de búsqueda en un millón de registros.




# 1. El Problema: Búsqueda en una Lista de Diccionarios ($O(n)$)

Imagina que tenemos una lista que representa a un millón de ciudadanos. Cada elemento es un diccionario con su `id` y `nombre`. 

```python
# Estructura: Lista de diccionarios
data = [
    {"id": 1, "nombre": "Ana"},
    {"id": 2, "nombre": "Betto"},
    # ... 999,997 registros más ...
    {"id": 1000000, "nombre": "Zulema"}
]
```


Si queremos encontrar a la persona con el `id: 999,999`, el ordenador debe realizar una **búsqueda lineal o secuencial**. 

En el peor de los casos, debe inspeccionar cada uno de los $n$ elementos.

* **Complejidad Temporal:** $O(n)$.
* **Realidad física:** Si cada comparación tarda un microsegundo, buscar al último elemento nos tomaría aproximadamente 1 segundo. Si hacemos 1,000 búsquedas, ¡estaríamos esperando casi 17 minutos!




## 2. La Solución: Estructuras Optimizadas ($O(1)$ o $O(\log n)$)

Cuando pasamos estos datos a herramientas diseñadas para grandes volúmenes, la arquitectura cambia radicalmente:



## A. Pandas (DataFrames)
Pandas utiliza **bloques de memoria contiguos** y permite crear **índices**. Al indexar la columna `id`, la búsqueda deja de ser lineal.
* Internamente, Pandas puede usar *Hash Tables* para búsquedas de igualdad.
* **Complejidad:** $O(1)$ en promedio. Encontrar el registro 1 o el 1,000,000 toma prácticamente el mismo tiempo.



## B. Bases de Datos SQL (B-Trees)
Una base de datos no recorre la tabla de arriba a abajo. Utiliza una estructura de árbol (normalmente **B-Tree**) para sus índices.





* **Complejidad:** $O(\log n)$. 
* Para $1,000,000$ de registros, $\log_2(1,000,000) \approx 20$. En lugar de un millón de comparaciones, el motor de la base de datos solo realiza unas 20 operaciones para localizar el dato exacto.




## Tabla Comparativa de Eficiencia

| Estructura | Tipo de Acceso | Operaciones para 1M de filas | Eficiencia Relativa |
| :--- | :--- | :--- | :--- |
| **Lista de Diccionarios** | Búsqueda Lineal | $1,000,000$ | Muy Baja |
| **SQL (con Índice)** | B-Tree | $\approx 20$ | Muy Alta |
| **Pandas (con Índice)** | Hash Table | $\approx 1$ | Instantánea |



# Conclusión para el perfil matemático


Trabajar con una lista de diccionarios es tratar a los datos como una **sucesión simple**, donde el coste computacional escala proporcionalmente al tamaño del conjunto. 

Al usar DataFrames o SQL, tratamos los datos como **conjuntos organizados o espacios vectoriales indexados**, donde la estructura misma de la memoria reduce drásticamente la entropía de la búsqueda.


**Pandas** optimiza estas operaciones utilizando estructuras vectorizadas similares a las matrices de Álgebra Lineal, permitiendo operaciones en bloque mucho más rápidas.


Vamos a bajar la teoría a la práctica con un script de Python. 

Vamos a comparar el tiempo que tarda una **Lista de Diccionarios** (fuerza bruta) frente a un **DataFrame de Pandas** (indexado) para encontrar un registro específico en un universo de 1,000,000 de entradas.


````python
import pandas as pd
import time

# 1. Preparación del Dataset
print("Generando datos...")
n_registros = 1_000_000
data_list = [{"id": i, "valor": f"usuario_{i}"} for i in range(n_registros)]
data_list 
````

````python
df = pd.DataFrame(data_list)
df   
````

````python

id_a_buscar = 999_999

# --- PRUEBA 1: Lista (O(n)) ---
inicio_lista = time.time()
for registro in data_list:
    if registro["id"] == id_a_buscar:
        break
tiempo_lista = time.time() - inicio_lista
tiempo_lista 
````

````python

# --- PRUEBA 2: Pandas con Índice (O(1)) ---
df_indexed = df.set_index("id")
df_indexed  
````

````python
inicio_pandas = time.time()
resultado_pandas = df_indexed.loc[id_a_buscar]
tiempo_pandas = time.time() - inicio_pandas
tiempo_pandas 
````

# ¿Qué está pasando aquí matemáticamente?

1.  **En la Lista ($O(n)$):** El intérprete de Python tiene que entrar en el primer diccionario, verificar la llave, comparar el valor, y si no coincide, saltar al siguiente espacio de memoria. Para llegar al final, hizo **1,000,000 de saltos y comparaciones**.


2.  **En Pandas/SQL ($O(1)$ aprox.):** Al ejecutar `.set_index("id")`, Pandas crea una **Tabla Hash**. Matemáticamente, se aplica una función $f(id)$ que te dice exactamente en qué dirección de memoria está el dato. No hay "recorrido", hay "salto directo".





## ¿Por qué esto escala de forma crítica?

Si duplicas los datos a 2,000,000:
* La **lista** tardará el **doble** de tiempo (relación lineal).
* **Pandas/SQL** tardará prácticamente lo **mismo** (relación constante o logarítmica).



Como notarás, en sistemas de producción (donde no tienes uno, sino miles de usuarios haciendo consultas simultáneas), usar una lista de diccionarios simplemente colapsaría el procesador por una ineficiencia algorítmica de base.


### La Intuición del Matemático
¿Por qué preferimos `fillna(media)` sobre `dropna()`? 
* Si borramos la fila (`dropna`), perdemos la información del parámetro `Tiempo`. 
* Si imputamos con la media, mantenemos el tamaño del vector y no alteramos la esperanza matemática $E[X]$ del conjunto, aunque reducimos artificialmente la varianza $\sigma^2$. Es un **trade-off** (compromiso) entre integridad y precisión.

## Script de Solución: Taller de Integración Final


````python
import pandas as pd
import numpy as np
import csv
import os

# ==========================================
# PARTE 1: Preparación y Gestión de Archivos
# ==========================================
# Creación de estructura de directorios para reproducibilidad
os.makedirs('data', exist_ok=True)
os.makedirs('results', exist_ok=True)

# Generación del dataset crudo (Raw Data)
raw_data = [
    ['timestamp', 'exp_id', 'densidad'],
    [1, 'EXP_01', 1.2],
    [2, 'EXP_01', 2.4],
    [2, 'EXP_01', 2.4], # Duplicado (Error de redundancia)
    [3, 'EXP_01', None], # Dato faltante (Indeterminación)
    [1, 'EXP_02', 1.1],
    [2, 'EXP_02', 1.8],
    [3, 'EXP_02', 2.5]
]

with open('data/mediciones.csv', 'w', newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerows(raw_data)

# ==========================================
# PARTE 2: Ingesta y Limpieza (Wrangling)
# ==========================================
# Lectura del archivo CSV
df = pd.read_csv('data/mediciones.csv')

# 1. Eliminación de duplicados (Asegurando la unicidad de las observaciones)
df = df.drop_duplicates()

# 2. Imputación de valores faltantes (NaN)
# Calculamos la esperanza matemática (media) local para el experimento EXP_01
media_exp1 = df[df['exp_id'] == 'EXP_01']['densidad'].mean()
df['densidad'] = df['densidad'].fillna(media_exp1)

# ==========================================
# PARTE 3: Merging y Transformación
# ==========================================
# Definición del diccionario de condiciones (Metadatos)
condiciones_dict = {
    "exp_id": ["EXP_01", "EXP_02"],
    "nutriente": ["Nutriente_A", "Nutriente_B"]
}
df_condiciones = pd.DataFrame(condiciones_dict)

# Unión de conjuntos (Inner Join) mediante la llave 'exp_id'
df_final = pd.merge(df, df_condiciones, on='exp_id')

# Creación de variable derivada (Tasa de crecimiento instantánea aproximada)
# Tc = densidad / tiempo
df_final['tasa_crecimiento'] = df_final['densidad'] / df_final['timestamp']

# ==========================================
# PARTE 4: Agregación y Conclusión
# ==========================================
# Reducción de dimensionalidad: Agrupamos por nutriente y calculamos la media
analisis_nutrientes = df_final.groupby('nutriente')['tasa_crecimiento'].mean().reset_index()

# Guardar resultados
analisis_nutrientes.to_csv('results/analisis_final.csv', index=False)

# Mostrar resultados en consola
print("--- DATAFRAME PROCESADO ---")
print(df_final)
print("\n--- RESULTADO DE LA HIPÓTESIS ---")
print(analisis_nutrientes)

# Verificación lógica de la hipótesis
tasa_a = analisis_nutrientes.loc[analisis_nutrientes['nutriente'] == 'Nutriente_A', 'tasa_crecimiento'].values[0]
tasa_b = analisis_nutrientes.loc[analisis_nutrientes['nutriente'] == 'Nutriente_B', 'tasa_crecimiento'].values[0]

if tasa_a > tasa_b:
    print("\nConclusión: La hipótesis se acepta. El Nutriente A es más eficiente.")
else:
    print("\nConclusión: La hipótesis se rechaza. El Nutriente B tiene mayor o igual eficiencia.")
````

### Ejercicio 22: Aproximación de la Derivada (Pandas y Cálculo)
**Contexto:** La derivada es el límite de la razón de cambio. En análisis de datos, usamos diferencias finitas.
**Reto:** Dado un conjunto de datos de posición $x(t) = t^2$, utiliza Pandas para calcular la velocidad instantánea aproximada en cada punto mediante la diferencia entre registros consecutivos.

```python
import pandas as pd
import numpy as np

# Creamos un dominio t en [0, 5] con paso h = 0.1
t = np.arange(0, 5.1, 0.1)
posicion = t**2

df = pd.DataFrame({'tiempo': t, 'posicion': posicion})

# Cálculo de la velocidad: delta_x / delta_t
# .diff() calcula la resta entre la fila i e i-1
df['velocidad'] = df['posicion'].diff() / df['tiempo'].diff()

print("Primeras filas del análisis de movimiento:")
print(df.head()) # La velocidad en t=1 debería ser cercana a 2 (derivada de t^2)
```

### Ejercicio 24: Detección de "Ruido" en la Señal (Análisis de Diagnóstico)
**Contexto:** En matemáticas aplicadas, el ruido son valores que se alejan demasiado del comportamiento esperado.
**Reto:** Encuentra los registros que están a más de 2 desviaciones estándar ($\sigma$) de la media ($\mu$) en un dataset de voltajes, lo cual es una técnica común de limpieza de datos.

```python
import pandas as pd

# Generamos datos con un "outlier" (valor atípico)
voltajes = [5.1, 5.2, 4.9, 5.0, 5.1, 12.5, 4.8, 5.0] # 12.5 es el ruido
df = pd.DataFrame(voltajes, columns=['voltaje'])

media = df['voltaje'].mean()
desviacion = df['voltaje'].std()

# Máscara lógica para identificar ruido
limite_superior = media + 2 * desviacion
limite_inferior = media - 2 * desviacion

outliers = df[(df['voltaje'] > limite_superior) | (df['voltaje'] < limite_inferior)]

print(f"Media: {media:.2f}, Desviación: {desviacion:.2f}")
print("Registros identificados como ruido:")
print(outliers)
```
