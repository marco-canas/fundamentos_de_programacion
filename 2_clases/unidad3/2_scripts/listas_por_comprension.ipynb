### Ejercicio 2:

Filtrar números primos:



````python
def es_primo(x):
    """
    Determina si un número entero x es primo.
    Retorna True si es primo, False en caso contrario.
    """
    if x < 2:
        return False
    for i in range(2, int(x**0.5) + 1):
        if x % i == 0:
            return False
    return True
````

````python
primos = [x for x in range(2,50) if es_primo(x)]
primos 
````

**Enunciado:**

````python
f = [x**2 for x in range(10)]
derivada = [f[i+1] - f[i] for i in range(len(f)-1)]
````


# **16. Números primos (básico)**

**Enunciado:**
Genera números primos menores que 20 (usa función auxiliar).

**Solución:**


````python
def es_primo(n):
    if n < 2:
        return False
    return all(n % i != 0 for i in range(2, int(n**0.5)+1))

primos = [x for x in range(20) if es_primo(x)]
primos 
````
