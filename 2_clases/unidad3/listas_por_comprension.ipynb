## 🔗 **3. Conexión matemática (clave en tu contexto)**

Se establece la analogía central:

* En matemáticas:
  $$
  A = { x^2 \mid x \in \mathbb{N}, x < 5 }
  $$

* En Python:

  ```python
  A = [x**2 for x in range(5)]
  ```

👉 Esto conecta directamente con:

* Definición de conjuntos por propiedad
* Funciones como reglas de asignación
* Pensamiento computacional como formalización algorítmica

Estructura general:

```python
[expresión for elemento in iterable if condición]
```

* `expresión`: función $f(x)$
* `elemento`: variable $x$
* `iterable`: dominio
* `condición`: restricción lógica

### 📌 Ejemplo 2: Filtrado (Lógica matemática)

```python
pares = [x for x in range(20) if x % 2 == 0]
```

$$

Interpretación:
$$
\vec{v} + \vec{w}
$$

# 📝 **Ejercicios tipo examen: Listas por comprensión en Python**

---

## 🔢 **1. Cuadrados perfectos**

**Enunciado:**
Construye la lista de los cuadrados de los enteros del 1 al 10.

**Solución:**

```python
cuadrados = [x**2 for x in range(1, 11)]
```

---

## 🔢 **2. Cubos negativos**

**Enunciado:**
Lista de los cubos de los enteros desde -5 hasta -1.

**Solución:**

```python
cubos = [x**3 for x in range(-5, 0)]
```

---

## 🔍 **3. Números pares**

**Enunciado:**
Lista de los números pares entre 1 y 20.

**Solución:**

```python
pares = [x for x in range(1, 21) if x % 2 == 0]
```

---

## 🔍 **4. Números impares elevados al cuadrado**

**Enunciado:**
Construye la lista de los cuadrados de los números impares entre 1 y 15.

**Solución:**

```python
impares_cuadrados = [x**2 for x in range(1, 16) if x % 2 != 0]
```

---

## 📈 **5. Evaluación de función cuadrática**

**Enunciado:**
Evalúa la función $f(x)=x^2-4x+3$ para $x \in {-3,-2,...,3}$.

**Solución:**

```python
f = [x**2 - 4*x + 3 for x in range(-3, 4)]
```

---

## 📈 **6. Valores positivos de una función**

**Enunciado:**
Filtra los valores positivos de la lista anterior.

**Solución:**

```python
positivos = [y for y in f if y > 0]
```

---

## 🔁 **7. Recíprocos**

**Enunciado:**
Lista de los recíprocos de los números del 1 al 10.

**Solución:**

```python
reciprocos = [1/x for x in range(1, 11)]
```

---

## 🔁 **8. Aproximación de derivada discreta**

**Enunciado:**
Dada $f(x)=x^2$, calcula la derivada discreta en puntos $x=0,...,9$.

**Solución:**

```python
f = [x**2 for x in range(10)]
derivada = [f[i+1] - f[i] for i in range(len(f)-1)]
```

---

## 🧮 **9. Suma de vectores**

**Enunciado:**
Suma los vectores $v=[1,2,3]$ y $w=[4,5,6]$.

**Solución:**

```python
v = [1,2,3]
w = [4,5,6]
suma = [v[i] + w[i] for i in range(len(v))]
```

---

## 🧮 **10. Producto escalar**

**Enunciado:**
Calcula el producto escalar de los vectores anteriores.

**Solución:**

```python
producto = sum([v[i]*w[i] for i in range(len(v))])
```

---

## 🔬 **11. Función seno (aproximación discreta)**

**Enunciado:**
Evalúa $\sin(x)$ en 10 puntos entre 0 y 2.

**Solución:**

```python
import numpy as np
x = np.linspace(0, 2, 10)
seno = [np.sin(val) for val in x]
```

---

## 🔢 **12. Números divisibles por 3 y 5**

**Enunciado:**
Lista los números entre 1 y 100 divisibles por 3 y 5.

**Solución:**

```python
divisibles = [x for x in range(1, 101) if x % 3 == 0 and x % 5 == 0]
```

---

## 🔢 **13. Transformación de lista**

**Enunciado:**
Dada la lista `[1,2,3,4]`, multiplica cada elemento por 10.

**Solución:**

```python
lista = [1,2,3,4]
resultado = [x*10 for x in lista]
```

---

## 🔁 **14. Filtrar mayores que un valor**

**Enunciado:**
De `[3,7,2,9,5]`, obtener los mayores que 5.

**Solución:**

```python
lista = [3,7,2,9,5]
mayores = [x for x in lista if x > 5]
```

---

## 🧠 **15. Lógica proposicional**

**Enunciado:**
Construye una lista de valores booleanos de la proposición $x > 3$ para $x=0,...,6$.

**Solución:**

```python
valores = [x > 3 for x in range(7)]
```

---

## 🔢 **16. Números primos (básico)**

**Enunciado:**
Genera números primos menores que 20 (usa función auxiliar).

**Solución:**

```python
def es_primo(n):
    if n < 2:
        return False
    return all(n % i != 0 for i in range(2, int(n**0.5)+1))

primos = [x for x in range(20) if es_primo(x)]
```

---

## 📊 **17. Valor absoluto**

**Enunciado:**
Calcula el valor absoluto de los números entre -5 y 5.

**Solución:**

```python
abs_vals = [abs(x) for x in range(-5, 6)]
```

---

## 🔄 **18. Producto de dos listas (tipo matriz)**

**Enunciado:**
Multiplica elemento a elemento dos listas.

**Solución:**

```python
a = [1,2,3]
b = [4,5,6]
producto = [a[i]*b[i] for i in range(len(a))]
```

---

## 🔢 **19. Tabla de multiplicar**

**Enunciado:**
Genera la tabla del 5.

**Solución:**

```python
tabla5 = [5*x for x in range(1, 11)]
```

## 🔁 **20. Lista de pares ordenados**

**Enunciado:**
Construye los pares $(x,y)$ con $x,y \in {1,2,3}$.

**Solución:**

```python
pares = [(x,y) for x in range(1,4) for y in range(1,4)]
```
