````python
import qrcode
from IPython.display import display
from PIL import Image

def generar_qr(url, version=1, box_size=5, border=4, fill_color="black", back_color="white"):
    """
    Genera y muestra un código QR para una URL dada en un Jupyter Notebook.
    
    Parámetros:
    - url: La URL para la cual generar el código QR
    - version: Tamaño del QR (1-40, donde 1 es el más pequeño)
    - box_size: Número de píxeles por cada "caja" del QR
    - border: Grosor del borde blanco alrededor del QR (en cajas)
    - fill_color: Color del código QR
    - back_color: Color de fondo del código QR
    
    Retorna:
    - Muestra el código QR directamente en el notebook
    - Retorna el objeto QR generado
    """
    # Configurar el generador de QR
    qr = qrcode.QRCode(
        version=version,
        error_correction=qrcode.constants.ERROR_CORRECT_L,
        box_size=box_size,
        border=border,
    )
    
    # Añadir la URL al QR
    qr.add_data(url)
    qr.make(fit=True)
    
    # Crear la imagen del QR
    img = qr.make_image(fill_color=fill_color, back_color=back_color)
    return img
````

# Ejemplo de un umbral lineal de decisión en plano   

Un **umbral lineal de decisión en el plano** es una recta que separa dos clases de datos. Matemáticamente suele escribirse como:

$$
w_1 x + w_2 y + b = 0
$$

* Si $w_1 x + w_2 y + b > 0$ → clase 1
* Si $w_1 x + w_2 y + b < 0$ → clase 2



A continuación tienes un ejemplo en matplotlib que:

1. Genera dos grupos de puntos.
2. Define un clasificador lineal.
3. Dibuja la recta de decisión y las regiones del plano.

---



## Ejemplo con matplotlib




````python
import numpy as np
import matplotlib.pyplot as plt

# =========================
# Datos simulados bien separables
# =========================
np.random.seed(1)

# Clase positiva (arriba-derecha)
X_pos = np.random.randn(60, 2) + np.array([2.5, 2])

# Clase negativa (abajo-izquierda)
X_neg = np.random.randn(60, 2) + np.array([-2.5, -2])

# =========================
# Umbral lineal de decisión
# w1*x + w2*y + b = 0
# =========================
w = np.array([1, 1])
b = 0

# Recta de decisión
x_vals = np.linspace(-6, 6, 200)
y_vals = -(w[0]*x_vals + b) / w[1]

# =========================
# Región de decisión
# =========================
xx, yy = np.meshgrid(np.linspace(-6,6,200),
                     np.linspace(-6,6,200))
Z = w[0]*xx + w[1]*yy + b

# =========================
# Gráfica
# =========================
plt.figure(figsize=(7,7))

# Regiones coloreadas según signo
plt.contourf(xx, yy, Z > 0, alpha=0.2)

# Puntos
plt.scatter(X_pos[:,0], X_pos[:,1], label='Clase positiva (+1)')
plt.scatter(X_neg[:,0], X_neg[:,1], label='Clase negativa (-1)')

# Recta de decisión
plt.plot(x_vals, y_vals, linewidth=2, label='Umbral lineal', color = 'red')

# Ejes
plt.axhline(0)
plt.axvline(0)
plt.grid(True)
plt.legend()
plt.title("Umbral lineal separando clase positiva y negativa")
plt.xlabel("x")
plt.ylabel("y")
plt.savefig(r"C:\Users\marco\Downloads\umbral_lineal_de_decision.png")
plt.show()
````

## Interpretación geométrica

* La recta es la frontera de decisión.
* Divide el plano en dos semiplanos.
* Cada semiplano corresponde a una clase distinta.
* Este modelo es la base de:

  * perceptrón,
  * regresión logística,
  * máquinas de soporte vectorial lineales (SVM lineales).

Geométricamente, el vector $w$ es perpendicular a la recta y determina la dirección de separación de las clases.

## 6.5 Sentencia `elif`

````python
x = 0

if x > 0:
    print("positivo")
elif x < 0:
    print("negativo")
else:
    print("cero")
````
