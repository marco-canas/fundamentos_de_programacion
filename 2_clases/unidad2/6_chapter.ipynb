````python
import qrcode
from IPython.display import display
from PIL import Image

def generar_qr(url, version=1, box_size=4, border=4, fill_color="black", back_color="white"):
    """
    Genera y muestra un código QR para una URL dada en un Jupyter Notebook.
    
    Parámetros:
    - url: La URL para la cual generar el código QR
    - version: Tamaño del QR (1-40, donde 1 es el más pequeño)
    - box_size: Número de píxeles por cada "caja" del QR
    - border: Grosor del borde blanco alrededor del QR (en cajas)
    - fill_color: Color del código QR
    - back_color: Color de fondo del código QR
    
    Retorna:
    - Muestra el código QR directamente en el notebook
    - Retorna el objeto QR generado
    """
    # Configurar el generador de QR
    qr = qrcode.QRCode(
        version=version,
        error_correction=qrcode.constants.ERROR_CORRECT_L,
        box_size=box_size,
        border=border,
    )
    
    # Añadir la URL al QR
    qr.add_data(url)
    qr.make(fit=True)
    
    # Crear la imagen del QR
    img = qr.make_image(fill_color=fill_color, back_color=back_color)
     
    return img

# Ejemplo de uso:
# generar_qr("https://www.ejemplo.com")

````

Podemos definir funciones más complejas:

### Valor absoluto

$$
|x| =
\begin{cases}
x & \text{si } x \ge 0 \\  
-x & \text{si } x < 0
\end{cases}
$$



````python
def valor_absoluto(x):
    if x >= 0:
        return x
    else:
        return -x
````

$$
g(x)=
\begin{cases}
x^2 & x<0 \\  
\sqrt{x} & x\ge 0
\end{cases}
$$



````python
import math

def g(x):
    if x < 0:
        return x**2
    else:
        return math.sqrt(x)
````

$$

````python
def sucesion(n):
    if n == 0:
        return 1
    else:
        return 2*sucesion(n-1) + 1
````

### Ejemplo: verificar si una función es par

$$

````python
def es_par(func, valores):
    for x in valores:
        if func(x) != func(-x):
            return False
    return True

def h(x):
    return x**2

es_par(h, range(-10,11))
````

# Cómo empezar a crear funciones con la sentencia `lambda`  



### Introducción: funciones anónimas (lambda) en Python

*(para estudiantes de Matemáticas Puras)*

En matemáticas, muchas veces trabajamos con funciones sin necesidad de nombrarlas explícitamente, por ejemplo:

$$
x \mapsto x^2,\quad x \mapsto x+1,\quad (x,y) \mapsto x+y
$$

En Python, esta idea se implementa mediante **funciones anónimas o funciones lambda**, que permiten definir reglas funcionales de forma compacta, directa y operativa, especialmente útiles en:

* construcción de ejemplos rápidos,
* evaluación puntual de funciones,
* programación funcional,
* manipulación de listas, secuencias y estructuras matemáticas,
* experimentación computacional.

Son herramientas clave para pensar las funciones como **objetos matemáticos manipulables**.

---



## 1) ¿Qué es una función lambda?

Es una función definida en una sola línea, sin nombre, usando la palabra clave `lambda`.

Estructura general:

```python
lambda argumentos: expresion
```



Ejemplo matemático:

$$
f(x)=x^2
$$



````python
f = lambda x: x**2
````

Evaluación:


````python
f(2)
f(5)
f(-3)
````


## 2) Interpretación matemática

La expresión:


````python
lambda x: x**2
````


se interpreta como:

$$
x \mapsto x^2
$$

Es decir: una regla funcional inmediata.

# 9) Mini pipeline de ciencia de datos con funciones


````python
datos = [2, 5, None, 8, 12, None, 15, 3]

def limpiar(lista):
    return [x for x in lista if x is not None]

def escalar(lista):
    max_val = max(lista)
    return [x/max_val for x in lista]

def transformar(lista):
    return list(map(lambda x: x**2, lista))

datos_procesados = transformar(escalar(limpiar(datos)))
datos_procesados
````

Reflexión:

* pipeline de preprocesamiento
* encadenamiento funcional

## Crear módulo propio

Archivo: `analisis_datos.py`


````python
def limpiar(lista):
    return [x for x in lista if x is not None]

def escalar(lista):
    max_val = max(lista)
    return [x/max_val for x in lista]

def media(lista):
    return sum(lista)/len(lista)
````

## Usarlo en Jupyter



````python
import analisis_datos as ad

datos = [2, 5, None, 8, 12]

datos_limpios = ad.limpiar(datos)
datos_escalados = ad.escalar(datos_limpios)
media = ad.media(datos_limpios)

datos_escalados, media
````

## Problema 1: Composición de funciones

Sea:



````python
def f(x): return x**2 - 1
def g(x): return 2*x + 3
````

**Problema:** Construya una función que calcule $(f \circ g)(x)$.

### Solución


````python
def composicion(x):
    return f(g(x))

print(composicion(2))  # 48
````


##  Problema 2: Función inversa (verificación numérica)

**Problema:** Verifique si dos funciones son inversas en un intervalo.

````python
def f(x): return 3*x - 5
def g(x): return (x + 5)/3
````

### Solución


````python
def verificar_inversa(valores):
    for x in valores:
        if abs(f(g(x)) - x) > 1e-6:
            return False
    return True

print(verificar_inversa(range(-10, 10)))
````


##  Problema 3: Aproximación de derivada

**Problema:** Aproxime la derivada de una función usando diferencia finita.

###  Solución


````python
def derivada(f, x, h=1e-5):
    return (f(x + h) - f(x)) / h

print(derivada(lambda x: x**2, 3))  # ≈ 6
````


## Problema 4: Función como argumento

**Problema:** Cree una función que reciba otra función y devuelva su valor máximo en un intervalo.

### Solución


````python
import numpy as np

def maximo(f, a, b):
    x = np.linspace(a, b, 1000)
    return max(f(xi) for xi in x)

print(maximo(lambda x: -x**2 + 4, -3, 3))
````


## Problema 5: Raíz por bisección

### Solución



````python
def biseccion(f, a, b, tol=1e-5):
    while (b - a)/2 > tol:
        c = (a + b)/2
        if f(c) == 0:
            return c
        elif f(a)*f(c) < 0:
            b = c
        else:
            a = c
    return (a + b)/2

print(biseccion(lambda x: x**2 - 4, 0, 3))
````


##  Problema 6: Función recursiva (Fibonacci)

### Solución


````python
def fib(n):
    if n <= 1:
        return n
    return fib(n-1) + fib(n-2)

print(fib(6))
````


##  Problema 7: Memoización

### Solución


````python
memo = {}

def fib_memo(n):
    if n in memo:
        return memo[n]
    if n <= 1:
        return n
    memo[n] = fib_memo(n-1) + fib_memo(n-2)
    return memo[n]
````


## Problema 8: Función generadora

### Solución


````python
def generador_fib(n):
    a, b = 0, 1
    for _ in range(n):
        yield a
        a, b = b, a + b

print(list(generador_fib(10)))
````


## Problema 9: Map con funciones

### Solución


````python
lista = [1, 2, 3, 4]
resultado = list(map(lambda x: x**2, lista))
print(resultado)
````


## Problema 10: Filtrado funcional

### Solución

````python
lista = range(20)
pares = list(filter(lambda x: x % 2 == 0, lista))
print(pares)
````


## Problema 11: Integral numérica (trapecio)

### Solución


````python
def trapecio(f, a, b, n=1000):
    h = (b - a) / n
    suma = 0.5*(f(a) + f(b))
    for i in range(1, n):
        suma += f(a + i*h)
    return h * suma
````


## Problema 12: Función que retorna función

### Solución


````python
def crear_lineal(m, b):
    return lambda x: m*x + b

f = crear_lineal(2, 3)
print(f(5))
````


## Problema 13: Optimización simple

### Solución


````python
def minimo(f, a, b):
    import numpy as np
    x = np.linspace(a, b, 1000)
    valores = [f(xi) for xi in x]
    return x[valores.index(min(valores))]
````


## Problema 14: Serie de Taylor

### Solución


````python

def exp_taylor(x, n=10):
    from math import factorial
    return sum(x**k / factorial(k) for k in range(n))
````


## Problema 15: Composición múltiple

### Solución


````python
def componer(funcs, x):
    for f in reversed(funcs):
        x = f(x)
    return x
````


## Problema 16: Función escalón

### Solución

````python
def escalon(x):
    return 1 if x >= 0 else 0
````


## Problema 17: Norma de una función

### Solución



````python
def norma_L2(f, a, b, n=1000):
    import numpy as np
    x = np.linspace(a, b, n)
    return np.sqrt(sum(f(xi)**2 for xi in x)*(b-a)/n)
````


## Problema 18: Ceros múltiples

### Solución


````python

def encontrar_ceros(f, a, b, pasos=100):
    import numpy as np
    x = np.linspace(a, b, pasos)
    ceros = []
    for i in range(len(x)-1):
        if f(x[i]) * f(x[i+1]) < 0:
            ceros.append((x[i]+x[i+1])/2)
    return ceros
````


## Problema 19: Función logística

### Solución


````python
def logistica(x, L=1, k=1, x0=0):
    import numpy as np
    return L / (1 + np.exp(-k*(x - x0)))

````


## Problema 20: Iteración funcional (dinámica). O iteración de punto fijo. Encuentre el punto fijo de una función dada en un dominio dado. 

### Solución


````python
def iterar(f, x0, n):
    valores = [x0]
    for _ in range(n):
        x0 = f(x0)
        valores.append(x0)
    return valores

print(iterar(lambda x: 0.5*x + 1, 0, 10))
````
