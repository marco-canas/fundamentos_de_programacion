````python
import qrcode
from IPython.display import display
from PIL import Image

def generar_qr(url, version=1, box_size=4, border=4, fill_color="black", back_color="white"):
    """
    Genera y muestra un código QR para una URL dada en un Jupyter Notebook.
    
    Parámetros:
    - url: La URL para la cual generar el código QR
    - version: Tamaño del QR (1-40, donde 1 es el más pequeño)
    - box_size: Número de píxeles por cada "caja" del QR
    - border: Grosor del borde blanco alrededor del QR (en cajas)
    - fill_color: Color del código QR
    - back_color: Color de fondo del código QR
    
    Retorna:
    - Muestra el código QR directamente en el notebook
    - Retorna el objeto QR generado
    """
    # Configurar el generador de QR
    qr = qrcode.QRCode(
        version=version,
        error_correction=qrcode.constants.ERROR_CORRECT_L,
        box_size=box_size,
        border=border,
    )
    
    # Añadir la URL al QR
    qr.add_data(url)
    qr.make(fit=True)
    
    # Crear la imagen del QR
    img = qr.make_image(fill_color=fill_color, back_color=back_color)
     
    return img

# Ejemplo de uso:
# generar_qr("https://www.ejemplo.com")

````

### `pass`

Marcador sintáctico.



La sentencia `pass` se usa como “marcador de posición”: no hace nada, pero permite que la estructura del programa sea válida mientras decides qué lógica agregar después.


````python
for i in range(10):
    if i % 2 == 0:
        # Aquí luego agregaremos lógica para los números pares
        pass 
    else:
        print(f"{i} es impar y se procesa normalmente")
````


**Qué muestra este ejemplo:**

* `pass` mantiene la estructura del `if` sin ejecutar ninguna acción.
* El programa sigue funcionando sin errores de sintaxis.
* Es útil cuando estás diseñando el algoritmo y aún no implementas una parte del código.

# Aproximemos a pi 

$$ \sum_{i = 1}^{\infty} 4 (-1)^{n} \frac{1}{2n - 1} = \pi $$

$$ \sum_{i = 1}^{100} 4 (-1)^{n+1} \frac{1}{2n - 1} = \pi $$

````python
suma = 0
N = 100_000_000 
for n in range(0, N+1): 
    suma += 4*(-1)**(n)/(2*n+1)
suma     
````

````python
import numpy as np 

np.pi
````

````python
# algoritmo para aproximar el número pi con ciclo mientras y break
pi_aproximado = 0
n = 1
while True:
    pi_aproximado += 4 * ((-1)**(n+1)) / (2*n - 1)
    n += 1
    if abs(pi_aproximado - 3.14159) < 0.000001:
        break
pi_aproximado    
````

Tu algoritmo está bien encaminado (usa la serie de Serie de Leibniz), pero tiene un problema conceptual fuerte:



- Estás comparando contra un valor fijo de $\pi$ (`3.14159`), lo cual **no es adecuado** si queremos un criterio interno de convergencia.

Lo correcto, como planteas, es usar:

$$
| \pi_n - \pi_{n-1} | < \text{tolerancia}
$$

---



# Versión mejorada (criterio entre aproximaciones)

```python


````python
# Aproximación de pi con criterio de tolerancia entre iteraciones

tolerancia = float(input("Ingrese la tolerancia: "))

pi_aproximado = 0
pi_anterior = 0
n = 1

while True:
    pi_anterior = pi_aproximado
    
    pi_aproximado += 4 * ((-1)**(n+1)) / (2*n - 1)
    
    # criterio de parada
    if abs(pi_aproximado - pi_anterior) < tolerancia:
        break
    
    n += 1

print("Aproximación de pi:", pi_aproximado)
print("Número de iteraciones:", n)
````

# Mejora conceptual

Este algoritmo ahora:

* No usa el valor real de $\pi$
* Usa un **criterio de convergencia numérica**
* Evalúa estabilidad de la sucesión
* Es consistente con métodos de análisis numérico

---



# Observación matemática clave

La serie de Leibniz:

$$
\pi = 4 \sum_{n=1}^{\infty} \frac{(-1)^{n+1}}{2n-1}
$$

- **Converge MUY lento**

Por ejemplo:

* Para $10^{-3}$ → miles de iteraciones
* Para $10^{-6}$ → millones

---



# Versión optimizada (evitando potencias)

Puedes mejorar eficiencia evitando `(-1)**(n+1)`:



````python
tolerancia = float(input("Ingrese la tolerancia: "))

pi_aproximado = 0
pi_anterior = 0
n = 1
signo = 1  # alterna entre +1 y -1

while True:
    pi_anterior = pi_aproximado
    
    pi_aproximado += 4 * signo / (2*n - 1)
    
    if abs(pi_aproximado - pi_anterior) < tolerancia:
        break
    
    signo *= -1
    n += 1

print("Aproximación de pi:", pi_aproximado)
print("Iteraciones:", n)
````


# Nivel alto (para Matemáticas Puras)

veamos que:

* Están trabajando con una **sucesión de sumas parciales**
* El criterio implementa:

$$
\lim_{n \to \infty} |\pi_n - \pi_{n-1}| = 0
$$

* Esto conecta con:

  * convergencia de series
  * criterio de Cauchy
  * análisis numérico

---



# Versión con gráfica de convergencia

Este script:

* Calcula las aproximaciones sucesivas
* Guarda los valores
* Grafica cómo converge hacia $\pi$

## Código con `matplotlib`

```python


````python
import matplotlib.pyplot as plt
import math

# Parámetros
tolerancia = float(input("Ingrese la tolerancia: "))
max_iter = 100000  # límite de seguridad

# Inicialización
pi_aproximado = 0
pi_anterior = 0
n = 1
signo = 1

# Listas para graficar
aproximaciones = []
iteraciones = []

while n <= max_iter:
    pi_anterior = pi_aproximado
    
    pi_aproximado += 4 * signo / (2*n - 1)
    
    aproximaciones.append(pi_aproximado)
    iteraciones.append(n)
    
    if abs(pi_aproximado - pi_anterior) < tolerancia:
        break
    
    signo *= -1
    n += 1

# Gráfica
plt.figure()
plt.plot(iteraciones, aproximaciones, label="Aproximación de π")
plt.axhline(y=math.pi, linestyle='--', label="π real", color='red')

plt.xlabel("Número de iteraciones")
plt.ylabel("Valor aproximado")
plt.title("Convergencia de la serie de Leibniz para π")
plt.legend()
plt.grid()
plt.savefig(r"C:\Users\marco\Downloads\convergencia_pi.png")  # Guardar la gráfica como imagen
plt.show()

# Resultados
print("Aproximación:", pi_aproximado)
print("Iteraciones:", n)
````


## Versión con `plotly` (interactiva)

````python
import plotly.graph_objects as go
import math

tolerancia = float(input("Ingrese la tolerancia: "))
max_iter = 100000

pi_aproximado = 0
pi_anterior = 0
n = 1
signo = 1

aproximaciones = []
iteraciones = []

while n <= max_iter:
    pi_anterior = pi_aproximado
    
    pi_aproximado += 4 * signo / (2*n - 1)
    
    aproximaciones.append(pi_aproximado)
    iteraciones.append(n)
    
    if abs(pi_aproximado - pi_anterior) < tolerancia:
        break
    
    signo *= -1
    n += 1

# Gráfica interactiva
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=iteraciones,
    y=aproximaciones,
    mode='lines',
    name='Aproximación de π'
))

fig.add_trace(go.Scatter(
    x=iteraciones,
    y=[math.pi]*len(iteraciones),
    mode='lines',
    name='π real',
    line=dict(dash='dash')
))

fig.update_layout(
    title="Convergencia de la serie de Leibniz",
    xaxis_title="Iteraciones",
    yaxis_title="Valor",
)

fig.show()
fig.write_image(r"C:\Users\marco\Downloads\convergencia_pi.png")

print("Aproximación:", pi_aproximado)
print("Iteraciones:", n)
````

````python
import numpy as np
import time
````

````python

# ============================================================
# DEMOSTRACIÓN: Loop FOR en Python vs Método SUM() de NumPy
# ============================================================

print("=" * 70)
print("COMPARACIÓN DE RENDIMIENTO: loop for vs numpy.sum()")
print("=" * 70)

# Crear un arreglo grande de NumPy (10 millones de elementos)
tamaño = 10_000_000
print(f"\n📊 Creando un arreglo numpy con {tamaño:,} elementos aleatorios...")
datos = np.random.random(tamaño)
print(f"   Arreglo creado. Tipo de dato: {datos.dtype}")
print(f"   Memoria aproximada: {datos.nbytes / 1e6:.2f} MB")

# ============================================================
# MÉTODO 1: Loop FOR tradicional en Python
# ============================================================
print("\n" + "-" * 70)
print("🔵 MÉTODO 1: Suma con loop FOR en Python puro")
print("-" * 70)

inicio_for = time.time()

suma_for = 0
for valor in datos:
    suma_for += valor

fin_for = time.time()
tiempo_for = fin_for - inicio_for

print(f" Resultado de la suma: {suma_for:.6f}")
print(f" Tiempo de ejecución: {tiempo_for:.4f} segundos")

# ============================================================
# MÉTODO 2: Método SUM() de NumPy (vectorizado)
# ============================================================
print("\n" + "-" * 70)
print("MÉTODO 2: Suma con numpy.sum() (operación vectorizada en C)")
print("-" * 70)

inicio_numpy = time.time()

suma_numpy = np.sum(datos)

fin_numpy = time.time()
tiempo_numpy = fin_numpy - inicio_numpy

print(f" Resultado de la suma: {suma_numpy:.6f}")
print(f"  Tiempo de ejecución: {tiempo_numpy:.4f} segundos")

# ============================================================
# COMPARACIÓN Y ANÁLISIS
# ============================================================
print("\n" + "=" * 70)
print("ANÁLISIS DE RENDIMIENTO")
print("=" * 70)

# Calcular la diferencia y el factor de mejora
diferencia = tiempo_for - tiempo_numpy
factor_mejora = tiempo_for / tiempo_numpy

print(f"\n Comparación directa:")
print(f"   • Tiempo del loop for:      {tiempo_for:.4f} s")
print(f"   • Tiempo de numpy.sum():    {tiempo_numpy:.4f} s")
print(f"   • Diferencia absoluta:      {diferencia:.4f} s")
print(f"   • Factor de mejora:         {factor_mejora:.1f}x más rápido con NumPy")
````

````python

# Verificar que ambos métodos dan el mismo resultado (dentro de tolerancia)
if np.abs(suma_for - suma_numpy) < 1e-10:
    print(f"\n✅ VERIFICACIÓN: Ambos métodos producen el mismo resultado.")
else:
    print(f"\n  VERIFICACIÓN: Los resultados difieren (revisar precisión numérica).")

print("\n" + "=" * 70)
print(" EXPLICACIÓN DEL FENÓMENO")
print("=" * 70)

explicacion = """
 ¿Por qué numpy.sum() es tan más rápido?

1. 🐍 Loop FOR en Python:
   - Cada iteración requiere interpretación del código Python
   - Verificación de tipos en cada operación
   - Múltiples llamadas a funciones internas de Python
   - Todo esto genera una sobrecarga significativa

2. ⚡ numpy.sum() (vectorizado):
   - Está implementado en C, un lenguaje compilado y de muy bajo nivel
   - Opera sobre bloques continuos de memoria (buffers)
   - Las operaciones se ejecutan en loops optimizados en C
   - Aprovecha instrucciones SIMD (Single Instruction, Multiple Data) del procesador

3. Consecuencia práctica:
   - Para 10 millones de elementos, NumPy es hasta 50-100x más rápido
   - La diferencia crece con el tamaño de los datos
   - En ciencia de datos, la vectorización es CRUCIAL para el rendimiento
"""
print(explicacion)

# ============================================================
# DEMOSTRACIÓN ADICIONAL: Comparación con diferentes tamaños
# ============================================================
print("\n" + "=" * 70)
print(" COMPARACIÓN CON DIFERENTES TAMAÑOS DE DATOS")
print("=" * 70)

tamaños = [1_000, 100_000, 1_000_000, 10_000_000]
resultados = []

print("\nEjecutando pruebas con diferentes tamaños...\n")
print(f"{'Tamaño':>12} | {'Tiempo For (s)':>15} | {'Tiempo NumPy (s)':>17} | {'Factor mejora':>14}")
print("-" * 70)

for n in tamaños:
    # Crear datos de prueba
    datos_prueba = np.random.random(n)
    
    # Medir loop for
    inicio = time.time()
    suma_f = 0
    for v in datos_prueba:
        suma_f += v
    tiempo_f = time.time() - inicio
    
    # Medir numpy sum
    inicio = time.time()
    suma_np = np.sum(datos_prueba)
    tiempo_np = time.time() - inicio
    
    factor = tiempo_f / tiempo_np
    resultados.append((n, tiempo_f, tiempo_np, factor))
    
    print(f"{n:12,} | {tiempo_f:15.4f} | {tiempo_np:17.4f} | {factor:14.1f}x")

print("\n" + "=" * 70)
print("🎓 CONCLUSIÓN FINAL")
print("=" * 70)

conclusion = """
 Siempre que trabajes con arreglos numéricos grandes en Python:
   - USA NumPy y sus operaciones vectorizadas
   - EVITA loops explícitos de Python cuando sea posible
   - La diferencia de rendimiento es abismal y crece con los datos

 Excepciones (cuando un loop puede ser aceptable):
   - Datos muy pequeños (menos de 1000 elementos)
   - Lógica condicional muy compleja que no se puede vectorizar fácilmente
   - Cuando la claridad del código es más importante que la velocidad

 En ciencia de datos y matemáticas computacionales, la vectorización 
   NO es una opción, es una NECESIDAD para trabajar con datos a gran escala.
"""
print(conclusion)
````

# Método de bisección con `while` (criterio de tolerancia)

Recordemos la idea central del método de Método de bisección:

Se itera hasta que:

* el intervalo sea pequeño: $|b - a| < \text{tolerancia}$
* o la función esté cerca de cero: $|f(c)| < \text{tolerancia}$

---



## Script en Python



````python
# Método de bisección usando ciclo while

import math

# Definir la función
def f(x):
    return x**3 - x - 2   # puedes cambiarla

# Datos iniciales
a = float(input("Ingrese el extremo izquierdo a: "))
b = float(input("Ingrese el extremo derecho b: "))
tolerancia = float(input("Ingrese la tolerancia: "))
max_iter = 1000  # seguridad

# Verificación inicial
if f(a) * f(b) > 0:
    print("Error: no hay cambio de signo en el intervalo")
else:
    iteracion = 0
    
    while abs(b - a) > tolerancia and iteracion < max_iter:
        c = (a + b) / 2
        
        print(f"Iteración {iteracion}: c = {c}, f(c) = {f(c)}")
        
        # Criterio de parada por valor de la función
        if abs(f(c)) < tolerancia:
            break
        
        # Cambio de intervalo
        if f(a) * f(c) < 0:
            b = c
        else:
            a = c
        
        iteracion += 1

    print("\nAproximación de la raíz:", c)
    print("Número de iteraciones:", iteracion)
````


# ¿Por qué `while` es la estructura correcta?

Porque el algoritmo depende de un criterio dinámico:

* No sabemos cuántas iteraciones tomará
* Depende de:

  * la función
  * el intervalo inicial
  * la tolerancia

- Esto refleja perfectamente la lógica:

$$
\text{repetir mientras no se cumpla el criterio de convergencia}
$$

    *   **Explicación:** Esta opción es correcta y promueve una buena práctica. La oscilación es una característica de la serie, no un error. La condición de parada debe ser entre aproximaciones sucesivas (criterio de Cauchy), no contra el valor real. La lentitud de convergencia refuerza la utilidad de usar un `while` con una tolerancia en lugar de un `for` con un número fijo de iteraciones, pero no hace que este último sea una "mala estrategia" per se, solo menos flexible.

# Visualización de la convergencia de π (oscilación)

## Versión con `matplotlib`


````python
import matplotlib.pyplot as plt
import math

# Parámetros
tolerancia = float(input("Ingrese la tolerancia (ej: 1e-4): "))
max_iter = 50_000  # límite de seguridad

# Inicialización
pi_aprox = 0 # la variable que almacenará la aproximación actual de pi
pi_prev = 0  # variable para almacenar la aproximación anterior
n = 1        # Contador de iteraciones
signo = 1    # alterna entre +1 y -1 para la serie de Leibniz

# Almacenamiento de datos
aproximaciones = []
iteraciones = []
errores = []

while n <= max_iter:
    pi_prev = pi_aprox
    
    # Serie de Leibniz
    pi_aprox += 4 * signo / (2*n - 1)
    
    # Guardar datos
    aproximaciones.append(pi_aprox)
    iteraciones.append(n)
    errores.append(abs(pi_aprox - math.pi))
    
    # Criterio de parada (Cauchy)
    if abs(pi_aprox - pi_prev) < tolerancia:
        break
    
    signo *= -1
    n += 1

# =========================
# GRÁFICA
# =========================
plt.figure()

plt.plot(iteraciones, aproximaciones, label="Aproximaciones de π")
plt.axhline(y=math.pi, linestyle='--', label="π real", color = 'red')

plt.xlabel("Iteraciones")
plt.ylabel("Valor")
plt.title("Oscilación y convergencia de la serie de Leibniz")
plt.legend()
plt.grid()
plt.savefig(r"C:\Users\marco\Downloads\convergencia_pi.png")  # Guardar la gráfica
plt.show()

# =========================
# GRÁFICA DEL ERROR
# =========================
plt.figure()

plt.plot(iteraciones, errores)
plt.xlabel("Iteraciones")
plt.ylabel("Error absoluto")
plt.title("Error |π_n - π|")
plt.grid()
plt.savefig(r"C:\Users\marco\Downloads\error_convergencia_pi.png")  # Guardar la gráfica del error
plt.show()

# Resultados
print("Aproximación final:", pi_aprox)
print("Iteraciones:", n)
````

````python
def factorial(n):
    """
    Calcula el factorial de n usando un ciclo for.
    """
    if not isinstance(n, int) or n < 0:
        return "Error: n debe ser un entero no negativo."
        
    resultado = 1
    for i in range(1, n + 1):
        resultado *= i
    return resultado

    # Ejemplo de uso:
print(factorial(5))  # Salida: 120
print(factorial(0))  # Salida: 1
````

````python

import math

def es_primo(n):
    """
    Determina si un número n es primo.
    """
    if not isinstance(n, int) or n <= 1:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False

    limite = int(math.sqrt(n)) + 1
    for i in range(3, limite, 2):
        if n % i == 0:
            return False
    return True

    # Ejemplo de uso:
print(es_primo(17))  # Salida: True
print(es_primo(15))  # Salida: False
    
````

**Solución:**


````python
def fibonacci(n):
    """
    Genera los primeros n números de la sucesión de Fibonacci.
    """
    if n <= 0:
        return []
    elif n == 1:
        return [0]
    elif n == 2:
        return [0, 1]

    fib_list = [0, 1]
    for i in range(2, n):
        siguiente = fib_list[i-1] + fib_list[i-2]
        fib_list.append(siguiente)
    return fib_list

    # Ejemplo de uso:
N = 10
sucesion = fibonacci(N)
print(f"Los primeros {N} números de Fibonacci son: {sucesion}")
````
