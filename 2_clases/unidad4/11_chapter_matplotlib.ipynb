```python
import qrcode
from IPython.display import display
from PIL import Image

def generar_qr(url, version=1, box_size=4, border=4, fill_color="black", back_color="white"):
    qr = qrcode.QRCode(
        version=version,
        error_correction=qrcode.constants.ERROR_CORRECT_L,
        box_size=box_size,
        border=border,
    )
    
    # Añadir la URL al QR
    qr.add_data(url)
    qr.make(fit=True)
    
    # Crear la imagen del QR
    img = qr.make_image(fill_color=fill_color, back_color=back_color)
     
    return img

```

# 3. ¿Qué es la visualización de datos?

La **visualización de datos** es la representación gráfica de información mediante diagramas, gráficos o figuras.



Su objetivo es **comunicar información de manera clara y eficiente**.



Matemáticamente, muchas estructuras de datos pueden representarse visualmente:

* una función $f(x)$ puede representarse en el plano cartesiano
* un vector $v \in \mathbb{R}^2$ puede representarse como un segmento orientado o como un punto en el plano. 
* una distribución de datos puede representarse mediante un histograma.



En ciencia de datos, la visualización permite descubrir patrones que no son evidentes únicamente observando tablas de números.



Por ejemplo, una serie temporal $x(t)$ puede mostrar tendencias, periodicidad o crecimiento exponencial.

# 6. Gráfico de línea simple

El gráfico más básico es el **gráfico de líneas**.

Supongamos que queremos graficar la función

$f(x)=x^2$


```python
import pandas as pd 

x = np.arange(-5,6,1) # Genera 11 puntos entre -5 y 5. 
# Una progresión aritmética con inicio -5, fin 5 y 11 puntos.

y = x**2 # Eleva cada elemento de x al cuadrado para obtener y. 

diccionario = {'x': x, 'y': y}
df = pd.DataFrame(diccionario) 
df 
```

```python
import numpy as np
import matplotlib.pyplot as plt


plt.figure(figsize=(4,3)) # Crea una figura de 8 pulgadas de ancho y 6 pulgadas de alto.
plt.scatter(x,y, color = 'blue') # Crea un gráfico de dispersión con x en el eje horizontal y y en el eje vertical.

plt.hlines(0, -5, 5, colors='red', linestyles='dashed') # Agrega una línea horizontal en y=0 desde x=-5 hasta x=5.
plt.vlines(0, 0, 25, colors='red', linestyles='dashed') # Agrega una línea vertical en x=0 desde y=0 hasta y=25.
plt.show()
```

```python
plt.scatter(x,y, color = 'blue') # Crea un gráfico de dispersión con x en el eje horizontal y y en el eje vertical.
plt.plot(x,y, color = 'skyblue') # Crea un gráfico de líneas con x en el eje horizontal y y en el eje vertical.
```

Este código:

1. genera 100 puntos en el intervalo $[-5,5]$
2. evalúa la función
3. dibuja la gráfica

---



# 7. API orientada a objetos en Matplotlib

Matplotlib también permite trabajar mediante una **interfaz orientada a objetos**.



```python
fig, ax = plt.subplots()

ax.plot(x,y)

plt.show()
```


Aquí:

* `fig` representa la figura
* `ax` representa el sistema de ejes

Esta forma es más adecuada para gráficos complejos.

---



# 8. Uso de subplots

Los **subplots** permiten mostrar varios gráficos en una misma figura.

Ejemplo:

graficar

$f(x)=x^2$  
$f(x)=x^3$



```python
fig, ax = plt.subplots(1,2)

x = np.linspace(-3,3,100)

ax[0].plot(x,x**2)
ax[0].hlines(0, -3, 3, colors='red', linestyles='dashed') # Agrega una línea horizontal en y=0 desde x=-3 hasta x=3.
ax[0].vlines(0, 0, 9, colors='red', linestyles='dashed') # Agrega una línea vertical en x=0 desde y=0 hasta y=9.

ax[1].plot(x,x**3)
ax[1].hlines(0, -3, 3, colors='red', linestyles='dashed') # Agrega una línea horizontal en y=0 desde x=-3 hasta x=3.
ax[1].vlines(0, 0, 27, colors='red', linestyles='dashed') # Agrega una línea vertical en x=0 desde y=0 hasta y=27.
plt.show()
```

Esto es muy útil para comparar funciones.

# 9. Personalización de gráficos

Matplotlib permite modificar múltiples aspectos del gráfico.





# 9.1 Agregar título



```python
plt.title("Gráfica de f(x)=x^2")

```


# 9.2 Etiquetas de los ejes



```python
plt.xlabel("x")
plt.ylabel("f(x)")

```


# 9.3 Agregar texto



```python
plt.text(0.5, 0.5, "Punto importante")

```


# 9.4 Marcadores



```python
x = np.array([-5, -3, 0, 2, 4, 5])
y = x**2
plt.plot(x, y, marker=">")

```


# 9.5 Estilo de línea



```python
plt.plot(x,y, linestyle = "--", marker="o", color = "red", markersize = 10, markerfacecolor = "blue",\
    markeredgecolor = "black", markeredgewidth = 2,linewidth = 4)

```


# 9.6 Color



```python
plt.plot(x,y, color="green")

```


# 9.7 Grosor de línea



```python
plt.plot(x,y, linewidth=3)

```


# 9.8 Cuadrícula



```python
plt.grid(True)

```


# 9.9 Límites de los ejes



```python
plt.xlim(-5,5)
plt.ylim(0,25)
```


# 9.10 Ticks



```python
plt.xticks(np.arange(-5,6,1))
plt.yticks(np.arange(0,26,5))

```


# 10. Tipos básicos de gráficos




# 10.1 Gráfico de barras

Representa datos categóricos.



```python
categorias = ["A","B","C","D"]

frecuencias  = [10,15,7,12]

plt.bar(categorias,frecuencias)

plt.show()

```


# 10.2 Histogramas

Muestran la distribución de datos.


```python
data = np.random.randn(1000)

plt.hist(data, bins=10)

plt.show()
```


# 10.3 Scatter plot

Representa pares ordenados.



```python
x = np.random.randn(100)

y = np.random.randn(100)

plt.scatter(x,y)

plt.show()
```


# 10.4 Pie chart

Representa proporciones.



```python
sizes = [90, 180, 30, 60]

labels = ["A","B","C","D"]

plt.pie(sizes, labels=labels)

plt.show()

```
