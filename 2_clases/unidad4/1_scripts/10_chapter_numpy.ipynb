````python
import qrcode
from IPython.display import display
from PIL import Image

def generar_qr(url, version=1, box_size=4, border=4, fill_color="black", back_color="white"):
    qr = qrcode.QRCode(
        version=version,
        error_correction=qrcode.constants.ERROR_CORRECT_L,
        box_size=box_size,
        border=border,
    )
    
    # Añadir la URL al QR
    qr.add_data(url)
    qr.make(fit=True)
    
    # Crear la imagen del QR
    img = qr.make_image(fill_color=fill_color, back_color=back_color)
     
    return img
````

# 1. ¿Qué es NumPy?

NumPy es una biblioteca de Python diseñada para realizar **cálculo numérico eficiente**.



Su elemento central es el **ndarray (N-dimensional array)**.



Un ndarray es similar a:

* un **vector** en matemáticas
* una **matriz**
* un **tensor**

pero optimizado para computación.



Ventajas de NumPy frente a listas de Python:

| Listas Python               | NumPy                    |
| --------------------------- | ------------------------ |
| Más flexibles               | Mucho más rápidas        |
| Menos eficientes en memoria | Alta eficiencia          |
| No vectorizadas             | Operaciones vectorizadas |

Producto punto



````python
v = np.array([1,2])
w = np.array([3,4])

np.dot(v,w)
````

# Mini-proyecto



````python
import matplotlib.pyplot as plt

t = np.linspace(0,100,200)

P0 = 10

r1 = 0.02
r2 = 0.05
r3 = 0.08

P1 = P0*np.exp(r1*t)
P2 = P0*np.exp(r2*t)
P3 = P0*np.exp(r3*t)

plt.plot(t,P1,label="r=0.02")
plt.plot(t,P2,label="r=0.05")
plt.plot(t,P3,label="r=0.08")

plt.legend()
plt.grid()
plt.savefig(r'C:\Users\marco\Downloads\crecimiento.png')  # Guardar la figura como un archivo PNG
````


Conclusión:

A mayor $r$, mayor velocidad de crecimiento.
