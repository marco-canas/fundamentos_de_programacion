## 1. Introducción Teórica: La Anatomía de una Consulta `SELECT`



Antes de ir al código, tengamos en cuenta que el orden en que escribimos SQL **no** es el orden en que la computadora lo ejecuta.



# El Orden Sintáctico (Cómo se escribe):

1. **`SELECT`** ¿Qué columnas quiero ver?
2. **`FROM`** ¿De qué tabla se van a sacar?
3. **`WHERE`** ¿Qué filtros deben cumplir las filas?
4. **`ORDER BY`** ¿Cómo quiero ordenar el resultado?
5. **`LIMIT`** ¿Cuántos registros quiero mostrar?


# 3. Ejercicios Prácticos Guiados

A partir de aquí, encuentras una guía para resolver retos de la administración de la cafetería.



# Reto 1: Consultar todo y columnas específicas (El inventario)

> **Problema de negocio:** El nuevo encargado de la cafetería quiere conocer qué productos vendemos y sus precios exactos.

* **Caso A: Traer toda la tabla (`*`)**


```python
# El asterisco trae absolutamente todo
query_todos = "SELECT * FROM productos;"
df_productos = pd.read_sql_query(query_todos, conexion)
df_productos
```

# Práctica de codificación  


* **Caso B: Traer columnas específicas (Buenas prácticas)**


```python
# Es mejor especificar solo lo que el negocio necesita para ahorrar memoria
query_columnas = "SELECT nombre, precio FROM productos;"
df_columnas = pd.read_sql_query(query_columnas, conexion)
print(df_columnas)
```


# Reto 2: Filtrar filas con `WHERE` (Análisis de Precios)

> **Problema de negocio:** Finanzas quiere evaluar si es necesario crear un "impuesto de lujo" o promoción especial para los productos que son caros (con un precio mayor o igual a $4.00).


```python
# Usamos el operador >= para filtrar
query_filtro = """
SELECT nombre, categoria, precio 
FROM productos 
WHERE precio >= 4.00;
"""
df_filtro = pd.read_sql_query(query_filtro, conexion)
print(df_filtro)
```


*Tip docente:* Explica que aquí se pueden usar operadores como `=`, `<`, `>`, `!=` y también texto (ej. `WHERE categoria = 'Bebidas Calientes'`).




# Reto 3: Ordenar datos con `ORDER BY` (Prioridades de Venta)

> **Problema de negocio:** El encargado de marketing quiere acomodar la pizarra del menú digital de la sucursal para que los productos aparezcan ordenados del más costoso al más barato.


```python
# 'DESC' ordena de mayor a menor. 'ASC' (por defecto) de menor a mayor.
query_orden = """
SELECT nombre, precio 
FROM productos 
ORDER BY precio DESC;
"""
df_orden = pd.read_sql_query(query_orden, conexion)
print(df_orden)

```



# Reto 4: Limitar resultados con `LIMIT` (El "Top")

> **Problema de negocio:** El Director General quiere ver una muestra rápida de las primeras 5 transacciones registradas en el sistema para verificar que los identificadores se estén generando correctamente.


```python
# LIMIT restringe el número de filas devueltas
query_limite = """
SELECT id_transaccion, fecha_hora, cantidad 
FROM ventas 
LIMIT 5;
"""
df_limite = pd.read_sql_query(query_limite, conexion)
print(df_limite)
```


# 4. El Gran Reto Integrador (Evaluación o Práctica Autónoma)

Para consolidar el aprendizaje, se le pide a los alumnos que resuelvan el siguiente problema combinando todo lo aprendido:

> **Reto para el alumno:** El departamento de operaciones detectó una anomalía en las ventas. Necesitan un reporte que muestre únicamente el **nombre** y la **categoría** de los productos que pertenezcan a la categoría de **'Bebidas Calientes'** y cuyo precio sea **mayor a $3.00**, ordenados de forma **ascendente por su nombre**.



**Solución esperada que el alumno debe construir:**


```python
query_reto = """
SELECT nombre, categoria, precio
FROM productos
WHERE categoria = 'Bebidas Calientes' AND precio > 3.00
ORDER BY nombre ASC;
"""
df_resultado = pd.read_sql_query(query_reto, conexion)
print(df_resultado)
```
