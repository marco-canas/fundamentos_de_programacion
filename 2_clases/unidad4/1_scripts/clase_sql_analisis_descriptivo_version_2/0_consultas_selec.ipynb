## 1. La Anatomía de una Consulta `SELECT`



# El Orden Sintáctico (Cómo se escribe):

1. **`SELECT`** ¿Qué columnas quiero ver?
2. **`FROM`** ¿De qué tabla o archivo se van a sacar?
3. **`WHERE`** ¿Qué filtros deben cumplir las filas o instancias u observaciones?
4. **`ORDER BY`** ¿Cómo quiero ordenar el resultado?
5. **`LIMIT`** ¿Cuántos registros quiero mostrar?

# Código para consultar de cada archivo .db los nombres de todas las tablas que contiene

Para conocer qué tablas existen dentro de un archivo `.db` (SQLite), debemos consultar una tabla oculta del sistema que se genera automáticamente en cada base de datos, llamada **`sqlite_master`**.



Aquí tienes el código limpio y estructurado utilizando **Pandas** y **SQLite**. Te incluyo dos opciones: una para un solo archivo y otra automatizada por si quieres escanear varios archivos `.db` dentro de una carpeta.

---



### Opción 1: Consultar las tablas de un solo archivo `.db`

Este script se conecta al archivo específico, extrae los nombres de las tablas y cierra la conexión inmediatamente.



````python
import pandas as pd
import sqlite3

# 1. Ruta del archivo .db que deseas consultar
ruta_db = r"C:\Users\marco\Documentos\docencia\fundamentos_de_programacion\2_clases\unidad4\2_datos\cafeteria.db"

# 2. Abrir la conexión
conexion = sqlite3.connect(ruta_db)

# 3. Consulta SQL a la tabla maestra de SQLite
# Filtramos por type='table' para ignorar índices o vistas automáticas
query = "SELECT name AS nombre_tabla FROM sqlite_master WHERE type='table';"

# 4. Leer el resultado con Pandas
df_tablas = pd.read_sql_query(query, conexion)

# 5. Cerrar conexión
conexion.close()

# 6. Mostrar el resultado
print(f"--- Tablas encontradas en: {ruta_db.split('\\')[-1]} ---")
print(df_tablas)
````


# Opción 2: Escanear una carpeta entera y listar las tablas de CADA archivo `.db`

Si estás administrando varias clases o proyectos y quieres saber qué hay dentro de cada archivo `.db` de esa carpeta de forma automatizada, este enfoque con `pathlib` te ahorrará mucho tiempo:



````python
import pandas as pd
import sqlite3
from pathlib import Path

# 1. Definir la ruta de la carpeta contenedora
ruta_carpeta = Path(r"C:\Users\marco\Documentos\docencia\fundamentos_de_programacion\2_clases\unidad4\2_datos")

# 2. Buscar todos los archivos que terminen en .db en esa carpeta
archivos_db = list(ruta_carpeta.glob("*.db"))

print(f"Se encontraron {len(archivos_db)} archivos .db para analizar.\n")

# 3. Iterar sobre cada archivo encontrado
for ruta_archivo in archivos_db:
    try:
        # Conectar al archivo actual
        conn = sqlite3.connect(str(ruta_archivo))
        
        # Consulta para traer los nombres de las tablas
        query = "SELECT name FROM sqlite_master WHERE type='table';"
        df = pd.read_sql_query(query, conn)
        
        conn.close()
        
        # Mostrar resultados de forma ordenada
        print(f"==================================================")
        print(f"ARCHIVO: {ruta_archivo.name}")
        print(f"==================================================")
        if not df.empty:
            # Convertimos la columna a una lista para una lectura más limpia en consola
            lista_tablas = df['name'].tolist()
            for tabla in lista_tablas:
                print(f"  • {tabla}")
        else:
            print("  [La base de datos está vacía / no contiene tablas]")
        print("\n")
        
    except Exception as e:
        print(f"Error al leer el archivo {ruta_archivo.name}: {e}\n")
````
