````python
import qrcode
from IPython.display import display
from PIL import Image

def generar_qr(url, version=1, box_size=4, border=4, fill_color="black", back_color="white"):
    """
    Genera y muestra un código QR para una URL dada en un Jupyter Notebook.
    
    Parámetros:
    - url: La URL para la cual generar el código QR
    - version: Tamaño del QR (1-40, donde 1 es el más pequeño)
    - box_size: Número de píxeles por cada "caja" del QR
    - border: Grosor del borde blanco alrededor del QR (en cajas)
    - fill_color: Color del código QR
    - back_color: Color de fondo del código QR
    
    Retorna:
    - Muestra el código QR directamente en el notebook
    - Retorna el objeto QR generado
    """
    # Configurar el generador de QR
    qr = qrcode.QRCode(
        version=version,
        error_correction=qrcode.constants.ERROR_CORRECT_L,
        box_size=box_size,
        border=border,
    )
    
    # Añadir la URL al QR
    qr.add_data(url)
    qr.make(fit=True)
    
    # Crear la imagen del QR
    img = qr.make_image(fill_color=fill_color, back_color=back_color)
     
    return img

# Ejemplo de uso:
# generar_qr("https://www.ejemplo.com")

````

## Una perspectiva para estudiantes de Matemáticas Puras



En el capítulo anterior trabajamos la **entrada y salida de archivos en Python**.

En matemáticas puras estamos acostumbrados a:

* Definir con precisión.
* Establecer hipótesis.
* Construir demostraciones.
* Analizar estructuras.



El análisis de datos no es ajeno a esta lógica. De hecho, puede entenderse como una **forma aplicada del pensamiento matemático**, donde:

* Los datos son objetos.
* Los modelos son estructuras.
* Las conclusiones son resultados inferenciales.



Vivimos en la **era de los datos**. La cantidad de información generada crece exponencialmente. Pero los datos por sí solos no tienen valor matemático hasta que:

* Se organizan.

Ahí es donde interviene el análisis de datos.

# ¿Qué es el análisis de datos desde una mirada matemática?

De manera general:

> El análisis de datos es el proceso de extraer información estructurada y significativa a partir de datos sin procesar, utilizando métodos matemáticos, estadísticos y computacionales.



Observa que esta definición involucra tres dimensiones:

1. **Estructura** → organización formal.
2. **Transformación** → aplicación de operadores.
3. **Inferencia** → extracción de conclusiones.



Desde una perspectiva formal, el análisis de datos puede verse como:

$$
\text{Datos crudos} \xrightarrow{\text{Transformaciones}} \text{Estructuras interpretables}
$$

Es un proceso secuencial, similar a una demostración matemática: cada paso depende del anterior.

---



# Análisis de datos vs. Analítica de datos

Es importante distinguir conceptualmente ambos términos.

## Análisis de datos

* Estudia datos pasados.
* Describe y explica fenómenos.
* Responde:
  **¿Qué ocurrió?**
  **¿Por qué ocurrió?**



## Analítica de datos

* Utiliza modelos avanzados.
* Predice comportamientos futuros.
* Responde:
  **¿Qué podría ocurrir?**



Desde una perspectiva matemática:

* El análisis se apoya en estadística descriptiva e inferencial.
* La analítica involucra modelación, optimización y aprendizaje automático.

# Tipos de análisis de datos (Una lectura estructural)

Podemos entenderlos como niveles de profundidad en la interpretación.



## 1. Análisis descriptivo

Responde: **¿Qué sucedió?**



Ejemplo:

* Resumen estadístico.
* Media, varianza, distribución.
* Reportes de ventas.

Matemáticamente:
$$
\text{Resumen} = f(\text{Datos})
$$

---



## 2. Análisis diagnóstico

Responde: **¿Por qué sucedió?**

Aquí buscamos relaciones causales o correlacionales.

Matemáticamente:

* Modelos de regresión.
* Análisis de dependencia.
* Estudio de correlaciones.

---



## 3. Análisis predictivo

Responde: **¿Qué puede suceder?**

Se construyen modelos:

$$
Y = f(X) + \varepsilon
$$

Donde:

* $ f $ es un modelo.
* $ \varepsilon $ es el error.



Aquí entran:

* Series de tiempo.
* Modelos probabilísticos.
* Machine Learning.

---



## 4. Análisis prescriptivo

Responde: **¿Qué decisión debemos tomar?**

Implica optimización:

$$
\min_{x \in S} f(x)
$$

Sujeto a restricciones.

Es el punto donde la matemática se convierte en acción.

# El Proceso de Análisis de Datos



## Una lectura estructural para matemáticos en formación

En matemáticas estamos acostumbrados a procesos bien definidos:

$$
\text{Hipótesis} \rightarrow \text{Desarrollo lógico} \rightarrow \text{Conclusión}
$$



El análisis de datos sigue una lógica análoga:

$$
\text{Problema} \rightarrow \text{Datos} \rightarrow \text{Transformación} \rightarrow \text{Modelo} \rightarrow \text{Interpretación}
$$



No es un procedimiento improvisado; es un **proceso estructurado**, comparable al método matemático.

---



# 1. Requisitos: Recopilación y Planificación

*(Formulación del problema)*

Esta fase equivale a **definir el problema matemático**.

Aquí no se calculan todavía resultados; se formaliza el contexto.



Preguntas fundamentales:

* ¿Qué queremos estudiar?
* ¿Cuál es el dominio del problema?
* ¿Qué tipo de análisis necesitamos?
* ¿Qué datos representan el fenómeno?
* ¿Qué limitaciones existen?



Desde una perspectiva matemática:

* Estamos definiendo el **espacio de trabajo**.
* Identificamos variables.
* Determinamos restricciones.
* Establecemos objetivos (función objetivo implícita).

Es la etapa donde pasamos del lenguaje cotidiano a una **formulación estructurada**.

---



# 2. Recopilación de Datos

*(Construcción del conjunto de trabajo)*

Aquí construimos el conjunto:

$$
D = {x_1, x_2, \dots, x_n}
$$

donde cada $x_i$ puede verse como un vector en $\mathbb{R}^k$.



Fuentes posibles:

* Web
* Sistemas transaccionales
* Sensores
* Archivos estructurados



Aspectos fundamentales:

* Momento de medición (dependencia temporal).
* Frecuencia de muestreo.
* Calidad del dato.



En términos matemáticos, estamos definiendo:

* La muestra.
* La dimensión del espacio.
* La naturaleza del dominio.

---



# 3. Limpieza de Datos

*(Depuración estructural)*

En esta etapa eliminamos inconsistencias.

Si consideramos el conjunto original $D$, la limpieza produce:

$$
D_{\text{limpio}} \subseteq D
$$



Eliminamos:

* Valores atípicos erróneos.
* Registros corruptos.
* Datos irrelevantes.



Esto es análogo a:

* Verificar hipótesis en una demostración.
* Eliminar contradicciones.
* Asegurar coherencia interna.



Un modelo construido sobre datos inconsistentes es como una demostración con un supuesto falso.

---



# 4. Preparación de Datos

*(Transformaciones matemáticas)*

Aquí aplicamos operadores sobre los datos.

Si cada dato es un vector $x \in \mathbb{R}^k$, podemos aplicar transformaciones:

$$
T: \mathbb{R}^k \rightarrow \mathbb{R}^m
$$



Ejemplos:

* Normalización.
* Cambio de escala.
* Creación de variables derivadas.



El ejemplo del número telefónico es trivial, pero conceptualmente representa una **descomposición estructural**.

En modelación real:

* Transformamos variables no lineales.
* Construimos variables indicadoras.
* Realizamos reducción de dimensión.

---



# 5. Análisis de Datos

*(Aplicación de modelos)*

Aquí aplicamos herramientas matemáticas:

* Estadística descriptiva.
* Regresión.
* Series de tiempo.
* Métodos de optimización.
* Álgebra lineal aplicada.



Un modelo típico puede escribirse como:

$$
Y = f(X) + \varepsilon
$$

Donde:

* $ f $ es una función o transformación.
* $ \varepsilon$ es el término de error.



En este punto, el matemático puro aporta:

* Análisis crítico de supuestos.
* Comprensión de estabilidad numérica.
* Evaluación de convergencia.
* Interpretación estructural del modelo.

---



# 6. Interpretación de Resultados

*(Discusión matemática)*

Un resultado numérico no es suficiente.

Debemos preguntarnos:

* ¿Es consistente con las hipótesis?
* ¿Es estable frente a perturbaciones?
* ¿Es generalizable?

Interpretar no es solo leer números; es comprender el significado estructural del resultado.

---



# 7. Visualización

*(Representación geométrica)*

La visualización es una forma de geometrización del dato.

Un gráfico puede verse como:

$$
\phi: D \rightarrow \mathbb{R}^2
$$

Es una proyección que permite comprender patrones.



Gráficos comunes:

* Barras → Comparaciones discretas.
* Histogramas → Distribuciones.
* Líneas → Dependencia temporal.

Visualizar es interpretar geométricamente la estructura algebraica.

# Estructura de la Clase

---



## 1. Requisitos → Modelado matemático del problema

### En el flujo de datos:

Se define el problema y los objetivos.



### En álgebra lineal:

Se traduce el problema a un modelo:

Si tenemos:

* $ n$ observaciones
* $ p $ variables

Construimos una matriz:

$$
X \in \mathbb{R}^{n \times p}
$$

Cada:

* Fila = vector en ℝᵖ
* Columna = vector en ℝⁿ



- Discusión con estudiantes:

* ¿El conjunto de columnas genera un subespacio de ℝⁿ?
* ¿Qué significa que haya variables redundantes?
* ¿Cómo se interpreta la dependencia lineal en datos reales?

---



## 2. Recopilación de datos → Construcción de la matriz

Aquí se construye explícitamente:

$$
X =
\begin{bmatrix}
x_{11} & x_{12} & \dots & x_{1p} \\ 
x_{21} & x_{22} & \dots & x_{2p} \\ 
\vdots & \vdots & \ddots & \vdots \\ 
x_{n1} & x_{n2} & \dots & x_{np}
\end{bmatrix}
$$



- Pregunta clave:

> ¿Qué pasa si una columna es combinación lineal de otras?

→ El rango disminuye.
→ El sistema pierde información independiente.

Aquí conectamos con:

* Rango
* Base
* Dimensión del subespacio generado

---



## 3. Limpieza de datos → Operaciones matriciales

Eliminar datos corruptos puede verse como:

* Eliminar filas → restricción del espacio
* Eliminar columnas → reducción de dimensión
* Reemplazar valores → transformación del operador



Normalización:

$$
\tilde{X} = X - \mu
$$

Esto es:

* Traslación en ℝᵖ
* Cambio de base implícito

- Concepto clave:
Centrar datos equivale a proyectar respecto al vector constante.

---



## 4. Preparación de datos → Transformaciones lineales

Agregar variables derivadas es aplicar una transformación:

$$
T: \mathbb{R}^p \rightarrow \mathbb{R}^k
$$

Ejemplo:

$$
Z = XA
$$

Donde:

* $ A $ es una matriz de transformación.
* Estamos cambiando representación.



Discusión:

* ¿Cuándo es invertible?
* ¿Cuándo pierde información?
* ¿Qué implica que $det(A)=0$?

---



## 5. Análisis de datos → Proyección y mínimos cuadrados

Aquí conectamos con uno de los resultados más bellos del curso:

### Regresión lineal como proyección ortogonal

Dado:

$$
X\beta = y
$$

Si no tiene solución exacta:

$$
\hat{\beta} = (X^TX)^{-1}X^Ty
$$

Esto es:

$$
\text{Proyección de } y \text{ sobre } \text{Col}(X)
$$



- Aquí se puede demostrar:

* Ecuaciones normales
* Existencia si $ X^TX $ es invertible
* Relación con independencia lineal

Este momento suele ser revelador para estudiantes de matemáticas puras.

---



## 6. Reducción de dimensión → Autovalores y autovectores

Aquí entra:

### PCA (Análisis de Componentes Principales)

Matriz de covarianza:

$$
\Sigma = \frac{1}{n} X^T X
$$



Buscamos:

$$
\Sigma v = \lambda v
$$

Interpretación:

* Autovectores → direcciones principales
* Autovalores → varianza explicada
* Elegimos los mayores λ → reducción de dimensión



Aquí conectamos con:

* Teorema espectral
* Bases ortogonales
* Diagonalización

---



## 7. Visualización → Cambio de base geométrico

Cuando proyectamos datos a 2 dimensiones:

$$
Z = XV_k
$$

Estamos:

* Cambiando de base
* Representando el subespacio dominante
* Eliminando direcciones de baja varianza

Es literalmente álgebra lineal geométrica aplicada.

# Parte 1 — Datos como matriz y rango

## 1.1 Construcción del conjunto de datos



````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Creamos un conjunto de datos simple con dependencia lineal
X = np.array([
    [1, 2],
    [2, 4],
    [3, 6]
])

df = pd.DataFrame(X, columns=["x1", "x2"])
df
````


## 1.2 Rango y dependencia lineal



````python
rank = np.linalg.matrix_rank(X)
print("Rango de X:", rank)
````


### Análisis teórico

* Las columnas son linealmente dependientes.
* x2 = 2x1
* El espacio columna tiene dimensión 1.
* Col(X) es una recta en ℝ³.

# Parte 2 — Regresión como Proyección Ortogonal

Ahora construimos un problema:

$$
X\beta = y
$$

---



## 2.1 Definimos el vector objetivo



````python
y = np.array([1, 2, 2]).reshape(-1,1)
y
````

## 2.2 Intentamos resolver el sistema directamente


````python
try:
    beta = np.linalg.inv(X.T @ X) @ X.T @ y
    print("Beta:", beta)
except:
    print("X^T X no es invertible (dependencia lineal).")
````

Aquí se observa el problema algebraico:

* $ X^T X $ no es invertible.
* El sistema no tiene solución única.

---



## 2.3 Usamos pseudoinversa (proyección mínima)

````python
beta_hat = np.linalg.pinv(X) @ y
print("Solución mínima norma (pseudoinversa):")
print(beta_hat)
````


## 2.4 Verificación geométrica

Calculamos la proyección:


````python
y_hat = X @ beta_hat
residuo = y - y_hat

print("Producto X^T * residuo:")
print(X.T @ residuo)
````


- Resultado esperado:

$$
X^T (y - \hat{y}) = 0
$$

Interpretación:
El residuo es ortogonal al espacio columna.

Esto demuestra formalmente que la regresión es una proyección ortogonal.

---



# Parte 3 — PCA como Problema Espectral

Ahora generamos datos con correlación alta.

---



## 3.1 Generación de datos correlacionados

````python
np.random.seed(0)

x1 = np.random.randn(100)
x2 = 2*x1 + np.random.randn(100)*0.5

X = np.column_stack((x1, x2))
X
````


## 3.2 Centramos los datos


````python
X_centered = X - np.mean(X, axis=0)
````

Interpretación algebraica:
Restar la media es una traslación en ℝ².

---



## 3.3 Matriz de covarianza

$$
\Sigma = \frac{1}{n} X^T X
$$



````python
cov_matrix = (X_centered.T @ X_centered) / len(X_centered)
print("Matriz de covarianza:")
print(cov_matrix)
````


## 3.4 Autovalores y autovectores


````python
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

print("Autovalores:")
print(eigenvalues)

print("Autovectores:")
print(eigenvectors)
````


Interpretación:

* Autovectores → direcciones principales.
* Autovalores → varianza en esa dirección.

Aquí aplicamos directamente el Teorema Espectral.

---



# Parte 4 — Visualización Geométrica del Cambio de Base

---



## 4.1 Graficar datos originales


````python
plt.figure()
plt.scatter(X_centered[:,0], X_centered[:,1])
plt.title("Datos Centrados")
plt.xlabel("x1")
plt.ylabel("x2")
plt.axhline(0)
plt.axvline(0)
plt.show()
````


## 4.2 Dibujar autovectores


````python
plt.figure()
plt.scatter(X_centered[:,0], X_centered[:,1])

for i in range(len(eigenvalues)):
    vec = eigenvectors[:, i] * np.sqrt(eigenvalues[i]) * 3
    plt.arrow(0, 0, vec[0], vec[1], head_width=0.2)

plt.title("Autovectores (Direcciones Principales)")
plt.axhline(0)
plt.axvline(0)
plt.show()
````


Interpretación geométrica:

* El autovector asociado al mayor autovalor indica la dirección de mayor varianza.
* PCA es simplemente un cambio de base hacia esa dirección.

## 1.3

Explique por qué la dependencia lineal implica que ( X^T X ) puede no ser invertible.

Si las columnas de $X$ son dependientes, entonces:

$$
\exists v \neq 0 \text{ tal que } Xv = 0

Entonces:

X^T X v = X^T(0) = 0

Por tanto, $X^T X$ tiene núcleo no trivial ⇒ no es invertible.

Conclusión:

$$
\det(X^T X) = 0
$$

# PARTE II — Regresión como Proyección Ortogonal (30%)

Sea:

$$
y =
\begin{pmatrix}
1 \\ 
2 \\ 
1 \\ 
0 \\ 
3
\end{pmatrix}
$$

---



## 2.1 (Formal)

Explique por qué el problema de mínimos cuadrados consiste en proyectar $y$ sobre Col(X).



### Solución

Queremos minimizar:

$$
|y - X\beta|^2
$$

Esto equivale a encontrar el vector en Col(X) más cercano a $y$.

Por teoría de proyecciones en espacios con producto interno:

El vector más cercano es la proyección ortogonal.



Por tanto:

$$
\hat{y} = P_{\text{Col}(X)} y
$$

---



## 2.2 (Python)



````python
y = np.array([1,2,1,0,3]).reshape(-1,1)

beta_hat = np.linalg.pinv(X) @ y
y_hat = X @ beta_hat
residuo = y - y_hat

beta_hat
````


## 2.3

Demuestre que:

$$
X^T(y - X\hat{\beta}) = 0
$$



### Solución

Partimos de:

$$
\hat{\beta} = (X^TX)^{-1}X^Ty
$$

Multiplicando:

$$
X^T y - X^T X \hat{\beta} = 0
$$



Entonces:

$$
X^T(y - X\hat{\beta}) = 0
$$



Interpretación:

El residuo es ortogonal a cada columna de X.

---



## 2.4 Interpretación geométrica



### Solución

* $ y \in \mathbb{R}^5 $
* $ \hat{y} \in \text{Col}(X) \subset \mathbb{R}^5 $
* Residuo $ r \in \text{Col}(X)^\perp $



Descomposición ortogonal:

$$
\mathbb{R}^5 = \text{Col}(X) \oplus \text{Col}(X)^\perp
$$

Según su teoría semiótica, el aprendizaje significativo se potencia cuando el estudiante logra traducir y articular distintos sistemas de representación: gráfico, algebraico, numérico, verbal y, en nuestro tiempo, digital.



De igual manera, el **Enfoque Ontosemiótico de Godino (2002, 2017)** muestra que los significados matemáticos emergen de prácticas diversas, lo que justifica la incorporación de estos lenguajes como recursos que enriquecen las configuraciones de práctica matemática, en sintonía con las exigencias de la investigación y la práctica profesional.



En el ámbito internacional, organizaciones como el **NCTM (2000, 2014)** han subrayado que la tecnología es esencial para la enseñanza y el aprendizaje de la matemática, dado que facilita la experimentación, la validación de conjeturas y la comunicación de ideas. 

Asimismo, instancias como la **ACM** y el **IEEE** han señalado la importancia de desarrollar habilidades computacionales como requisito fundamental en la formación universitaria, tanto en el sector público como privado.



Por tanto, enseñar matemáticas integrando Python, LaTeX, Markdown e inglés en ambientes como Jupyter Notebooks contribuye a la formación de profesionales con **competencias científicas, comunicativas y tecnológicas**, capaces de comprender la matemática de manera interactiva y experimental, y de responder a las demandas del mundo laboral. 

Esta integración, lejos de ser inadecuada, constituye un **puente entre la enseñanza académica y las necesidades de la sociedad**, fortaleciendo la pertinencia de los programas universitarios.

#  Estado del arte



El interés por integrar la tecnología en la enseñanza de las matemáticas tiene raíces en las propuestas de **Papert (1980, 1996)**, quien con su enfoque constructionista concibió la computadora como un “objeto para pensar”, capaz de mediar la construcción activa de significados matemáticos mediante la programación. 

Esta perspectiva se enlaza con los planteamientos de **Piaget (1970)** y **Vygotsky (1978)**, quienes reconocieron la importancia de la experimentación y de las herramientas culturales en la formación del pensamiento. 

Bajo esta mirada, el uso de entornos digitales no constituye una amenaza para la enseñanza tradicional de las matemáticas, sino un recurso que amplía las posibilidades de aprendizaje.



A comienzos del siglo XXI, la noción de **pensamiento computacional** introducida por **Wing (2006)** renovó el debate al proponer que esta competencia debía enseñarse de manera transversal desde los niveles iniciales hasta la educación superior, situándola al mismo nivel que la lectura, la escritura y las matemáticas. 

Este enfoque se alinea con los planteamientos del **NCTM (2000, 2014)**, que destacó la tecnología como un componente esencial para la experimentación, la representación y la comunicación en matemáticas. En paralelo, aportes teóricos como la **teoría de registros de representación semiótica de Duval (1993, 2006)** y el **Enfoque Ontosemiótico de Godino (2007, 2017)** muestran que el aprendizaje matemático se enriquece al transitar entre múltiples sistemas de representación, incluyendo los digitales.



En el contexto actual, organizaciones como la **ACM** y el **IEEE** insisten en que las competencias computacionales son indispensables en la formación profesional, lo que refuerza la pertinencia de integrar lenguajes como **Python, LaTeX, Markdown e inglés** en ambientes como **Jupyter Notebooks** para la enseñanza universitaria de las matemáticas. 

Así, el estado del arte evidencia un consenso creciente: lejos de ser un obstáculo, el pensamiento computacional y el uso de lenguajes digitales constituyen una vía para fortalecer la comprensión matemática, formar profesionales más competentes y responder a las demandas del mundo laboral y científico contemporáneo.

# Estado del arte con citas



El uso de la tecnología en la enseñanza de las matemáticas encuentra uno de sus pilares en **Papert (1980)**, quien afirmaba que *“los niños no aprenden sobre la computadora, aprenden con la computadora”* (p. 6). 

En su visión, la programación se convierte en un medio para la construcción activa de significados matemáticos, idea que se enlaza con la epistemología genética de **Piaget (1970)**, quien sostuvo que *“el conocimiento se construye a través de la acción”* (p. 15), y con la perspectiva de **Vygotsky (1978)**, al señalar que *“toda función en el desarrollo cultural del niño aparece dos veces: primero en el plano social y después en el individual”* (p. 57). 

Estos enfoques coinciden en otorgar a las herramientas culturales un papel esencial en la mediación del aprendizaje.



Posteriormente, **Wing (2006)** acuñó el término pensamiento computacional, definiéndolo como *“una habilidad fundamental para todos, no solo para los científicos de la computación”* (p. 33). 

Esta autora subrayó que debía enseñarse de manera transversal, en paralelo con la lectura, la escritura y la matemática. 

En la misma línea, el **NCTM (2000)** destacó que *“la tecnología es esencial en la enseñanza y el aprendizaje de las matemáticas; influye en la matemática que se enseña y mejora el aprendizaje de los estudiantes”* (p. 24).



Desde la didáctica de las matemáticas, **Duval (2006)** mostró que *“la comprensión matemática requiere la coordinación de al menos dos registros de representación diferentes”* (p. 107), lo que legitima la incorporación de registros digitales como LaTeX, Markdown y Python. 

Por su parte, **Godino (2017)** planteó que *“el significado de un objeto matemático está constituido por el sistema de prácticas en el que interviene”* (p. 7), con lo cual la práctica digital se convierte en parte constitutiva del aprendizaje matemático.



Finalmente, organismos internacionales como la **ACM y el IEEE (2013)** enfatizan que *“las habilidades computacionales son esenciales para el ejercicio profesional en cualquier disciplina”* (p. 12). 

De esta manera, se consolida un consenso: la enseñanza universitaria de las matemáticas debe integrar herramientas contemporáneas como **Jupyter Notebooks, Python, LaTeX, Markdown e inglés**, no para competir con la enseñanza tradicional, sino para **fortalecer la comprensión matemática, potenciar la comunicación científica y responder a las demandas del mundo laboral**.
