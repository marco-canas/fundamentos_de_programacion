````python
import qrcode
from IPython.display import display
from PIL import Image

def generar_qr(url, version=1, box_size=10, border=4, fill_color="black", back_color="white"):
    """
    Genera y muestra un c√≥digo QR para una URL dada en un Jupyter Notebook.
    
    Par√°metros:
    - url: La URL para la cual generar el c√≥digo QR
    - version: Tama√±o del QR (1-40, donde 1 es el m√°s peque√±o)
    - box_size: N√∫mero de p√≠xeles por cada "caja" del QR
    - border: Grosor del borde blanco alrededor del QR (en cajas)
    - fill_color: Color del c√≥digo QR
    - back_color: Color de fondo del c√≥digo QR
    
    Retorna:
    - Muestra el c√≥digo QR directamente en el notebook
    - Retorna el objeto QR generado
    """
    # Configurar el generador de QR
    qr = qrcode.QRCode(
        version=version,
        error_correction=qrcode.constants.ERROR_CORRECT_L,
        box_size=box_size,
        border=border,
    )
    
    # A√±adir la URL al QR
    qr.add_data(url)
    qr.make(fit=True)
    
    # Crear la imagen del QR
    img = qr.make_image(fill_color=fill_color, back_color=back_color)
    
    # Mostrar la imagen en el notebook
    display(img)
    
    return img

# Ejemplo de uso:
# generar_qr("https://www.ejemplo.com")

````

## ¬øVariables en Python?  



Una variable es el nombre de una ubicaci√≥n de memoria reservada que contiene un valor. 

Por ejemplo: supongamos a = 10. Aqu√≠, "a" es el nombre de la variable, el signo igual (=) es un operador de asignaci√≥n y 10 es el valor o literal. 

Por lo tanto, al usar un operador de asignaci√≥n (=) en Python, podemos reservar memoria para un valor sin declararlo expl√≠citamente.

## Operadores aritm√©ticos.   

Los operadores aritm√©ticos son necesarios para realizar operaciones aritm√©ticas como suma, resta, multiplicaci√≥n, divisi√≥n, etc. La siguiente tabla muestra la lista de operadores aritm√©ticos en Python:


# Operadores aritm√©ticos



| Nombre del operador      | S√≠mbolo del operador | Descripci√≥n                                                      | Ejemplo       |
| ------------------------ | -------------------- | ---------------------------------------------------------------- | ------------- |
| Suma                     | `+`                  | Suma dos valores                                                 | `3 + 2 = 5`   |
| Resta                    | `-`                  | Resta el segundo valor al primero                                | `5 - 2 = 3`   |
| Multiplicaci√≥n           | `*`                  | Multiplica dos valores                                           | `4 * 3 = 12`  |
| Divisi√≥n                 | `/`                  | Divide el primer valor entre el segundo (retorna un n√∫mero real) | `7 / 2 = 3.5` |
| Divisi√≥n entera          | `//`                 | Divide y retorna solo la parte entera del cociente               | `7 // 2 = 3`  |
| M√≥dulo (residuo)         | `%`                  | Retorna el residuo de la divisi√≥n                                | `7 % 2 = 1`   |
| Exponenciaci√≥n           | `**`                 | Eleva un n√∫mero a la potencia de otro                            | `2 ** 3 = 8`  |
| Operador unario positivo | `+`                  | Indica que el n√∫mero es positivo (no cambia el valor)            | `+5 = 5`      |
| Operador unario negativo | `-`                  | Cambia el signo del n√∫mero                                       | `-5 = -5`     |

---



# Nota did√°ctica 

* La **divisi√≥n `/` siempre devuelve flotante** en Python.
* El operador `//` es clave para **modelar conteos, ciclos y discretizaciones**.
* `**` permite conectar f√°cilmente con **potencias, crecimiento exponencial y modelos matem√°ticos**.



# Precedencia de operadores aritm√©ticos (Python)

En Python, **la precedencia** determina el orden en que se eval√∫an los operadores cuando no hay par√©ntesis.



# Tabla de precedencia (de mayor a menor)



| Nivel | Operador(es) | Descripci√≥n |
|------:|-------------|-------------|
| 1 | `()` | Par√©ntesis (agrupaci√≥n) |
| 2 | `**` | Exponenciaci√≥n |
| 3 | `+x`, `-x` | Operadores unarios (signo) |
| 4 | `*`, `/`, `//`, `%` | Multiplicaci√≥n, divisi√≥n, divisi√≥n entera y m√≥dulo |
| 5 | `+`, `-` | Suma y resta |




- **Notas clave para los estudiantes**

* La **exponenciaci√≥n se eval√∫a de derecha a izquierda**:
  `2 ** 3 ** 2 = 2 ** (3 ** 2) = 512`
* Los **par√©ntesis siempre mandan**.
* `*`, `/`, `//` y `%` tienen **la misma precedencia** y se eval√∫an de izquierda a derecha.

# Operadores relacionales.   

Los operadores relacionales se utilizan para comprobar la relaci√≥n entre operandos y comparar valores. Seg√∫n la condici√≥n, estos operadores devuelven "Verdadero"(`True`) o "Falso"(`False`). Revise los operadores relacionales en Python que se listan a continuaci√≥n:


# Operadores relacionales (comparaci√≥n) en Python



| Nombre del operador | S√≠mbolo del operador | Descripci√≥n                                                 | Ejemplo         |
| ------------------- | -------------------- | ----------------------------------------------------------- | --------------- |
| Igual que           | `==`                 | Verifica si dos valores son iguales                         | `5 == 5 ‚Üí True` |
| Distinto de         | `!=`                 | Verifica si dos valores son diferentes                      | `5 != 3 ‚Üí True` |
| Mayor que           | `>`                  | Verifica si el primer valor es mayor que el segundo         | `7 > 4 ‚Üí True`  |
| Menor que           | `<`                  | Verifica si el primer valor es menor que el segundo         | `3 < 6 ‚Üí True`  |
| Mayor o igual que   | `>=`                 | Verifica si el primer valor es mayor o igual que el segundo | `5 >= 5 ‚Üí True` |
| Menor o igual que   | `<=`                 | Verifica si el primer valor es menor o igual que el segundo | `4 <= 6 ‚Üí True` |

---



## Nota did√°ctica importante

* El operador `=` **no es relacional**: se usa para **asignaci√≥n**, no para comparaci√≥n.
* El resultado de **todo operador relacional es un booleano**: `True` o `False`.
* Estos operadores son la base para:

  * estructuras condicionales (`if`, `elif`, `else`)
  * bucles (`while`)
  * filtros de datos en ciencia de datos
