```python
import qrcode
from IPython.display import display
from PIL import Image

def generar_qr(url, version=1, box_size=4, border=4, fill_color="black", back_color="white"):
    """
    Genera y muestra un código QR para una URL dada en un Jupyter Notebook.
    
    Parámetros:
    - url: La URL para la cual generar el código QR
    - version: Tamaño del QR (1-40, donde 1 es el más pequeño)
    - box_size: Número de píxeles por cada "caja" del QR
    - border: Grosor del borde blanco alrededor del QR (en cajas)
    - fill_color: Color del código QR
    - back_color: Color de fondo del código QR
    
    Retorna:
    - Muestra el código QR directamente en el notebook
    - Retorna el objeto QR generado
    """
    # Configurar el generador de QR
    qr = qrcode.QRCode(
        version=version,
        error_correction=qrcode.constants.ERROR_CORRECT_L,
        box_size=box_size,
        border=border,
    )
    
    # Añadir la URL al QR
    qr.add_data(url)
    qr.make(fit=True)
    
    # Crear la imagen del QR
    img = qr.make_image(fill_color=fill_color, back_color=back_color)
     
    return img

# Ejemplo de uso:
# generar_qr("https://www.ejemplo.com")

```

```python
url = "https://colab.research.google.com/github/marco-canas/fundamentos_de_programacion/blob/main/clases/0_formato_clases_fundamentos_programacion/4_formato_clase_funda_programacion.ipynb"
generar_qr(url) 
```

# 3. ¿Qué es la visualización de datos?

La **visualización de datos** es la representación gráfica de información mediante diagramas, gráficos o figuras.



Su objetivo es **comunicar información de manera clara y eficiente**.



Matemáticamente, muchas estructuras de datos pueden representarse visualmente:

* una función $f(x)$ puede representarse en el plano cartesiano
* un vector $v \in \mathbb{R}^2$ puede representarse como un segmento orientado
* una distribución de datos puede representarse mediante un histograma

En ciencia de datos, la visualización permite descubrir patrones que no son evidentes únicamente observando tablas de números.



Por ejemplo, una serie temporal $x(t)$ puede mostrar tendencias, periodicidad o crecimiento exponencial.

---



# 4. ¿Qué es Matplotlib?

**Matplotlib** es una biblioteca de Python diseñada para crear gráficos y visualizaciones.

Permite construir gráficos similares a los utilizados en matemáticas, física, estadística y ciencia de datos.



Entre sus características principales:

* permite graficar funciones matemáticas
* funciona perfectamente con **NumPy**
* permite personalizar completamente los gráficos
* permite exportar resultados para artículos científicos

Matplotlib puede generar gráficos como:

* funciones matemáticas
* series temporales
* histogramas
* gráficos estadísticos
* diagramas científicos

# 6. Gráfico de línea simple

El gráfico más básico es el **gráfico de líneas**.

Supongamos que queremos graficar la función

$f(x)=x^2$


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5,5,100)

y = x**2

plt.plot(x,y)

plt.show()
```

Este código:

1. genera 100 puntos en el intervalo $[-5,5]$
2. evalúa la función
3. dibuja la gráfica

---



# 7. API orientada a objetos en Matplotlib

Matplotlib también permite trabajar mediante una **interfaz orientada a objetos**.



```python
fig, ax = plt.subplots()

ax.plot(x,y)

plt.show()
```


Aquí:

* `fig` representa la figura
* `ax` representa el sistema de ejes

Esta forma es más adecuada para gráficos complejos.

---



# 8. Uso de subplots

Los **subplots** permiten mostrar varios gráficos en una misma figura.

Ejemplo:

graficar

$f(x)=x^2$  
$f(x)=x^3$



```python
fig, ax = plt.subplots(1,2)

x = np.linspace(-3,3,100)

ax[0].plot(x,x**2)
ax[1].plot(x,x**3)

plt.show()
```

Esto es muy útil para comparar funciones.


# 9.7 Grosor de línea


# 10. Tipos básicos de gráficos




# 10.1 Gráfico de barras

Representa datos categóricos.



```python
categorias = ["A","B","C","D"]

valores = [10,15,7,12]

plt.bar(categorias,valores)

plt.show()

```


# 10.2 Histogramas

Muestran la distribución de datos.


```python
data = np.random.randn(1000)

plt.hist(data,bins=30)

plt.show()
```


# 10.3 Scatter plot

Representa pares ordenados.



```python
x = np.random.rand(100)

y = np.random.rand(100)

plt.scatter(x,y)

plt.show()
```


# 10.4 Pie chart

Representa proporciones.



```python
sizes = [30,40,20,10]

labels = ["A","B","C","D"]

plt.pie(sizes,labels=labels)

plt.show()

```
