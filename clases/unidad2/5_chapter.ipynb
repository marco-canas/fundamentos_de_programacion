````python
import qrcode
from IPython.display import display
from PIL import Image

def generar_qr(url, version=1, box_size=4, border=4, fill_color="black", back_color="white"):
    """
    Genera y muestra un código QR para una URL dada en un Jupyter Notebook.
    
    Parámetros:
    - url: La URL para la cual generar el código QR
    - version: Tamaño del QR (1-40, donde 1 es el más pequeño)
    - box_size: Número de píxeles por cada "caja" del QR
    - border: Grosor del borde blanco alrededor del QR (en cajas)
    - fill_color: Color del código QR
    - back_color: Color de fondo del código QR
    
    Retorna:
    - Muestra el código QR directamente en el notebook
    - Retorna el objeto QR generado
    """
    # Configurar el generador de QR
    qr = qrcode.QRCode(
        version=version,
        error_correction=qrcode.constants.ERROR_CORRECT_L,
        box_size=box_size,
        border=border,
    )
    
    # Añadir la URL al QR
    qr.add_data(url)
    qr.make(fit=True)
    
    # Crear la imagen del QR
    img = qr.make_image(fill_color=fill_color, back_color=back_color)
     
    return img

# Ejemplo de uso:
# generar_qr("https://www.ejemplo.com")

````

## 6.2 Ciclo `for` (`for` loop)

Se usa cuando el número de iteraciones es conocido.



Ejemplo matemático:


````python
list(range(0, 20, 2)) # La función range() tiene trs parámetros sin nombre,
# pero el primero es parámetro es el primer término de la progresión aritmética, 
# el segundo menos la diferencia me da el último termino de la progresión aritmética, 
# el tercer parámetro es la diferencia de la progresión aritmética, 
# que en python generalmente se le llama paso. 
````

## 6.3 `while` loop

Se usa cuando la iteración depende de una condición y no se conoce el número de iteraciones que vas a realizar.



````python
x = 1
num_iteraciones = 0
while x < 100:
    x *= 2
    num_iteraciones +=1
    
x, num_iteraciones   # Esta línea se ejecuta solo cuando el ciclo finaliza o para de iterar. 
````

Relación:

* crecimiento exponencial
* iteraciones hasta convergencia

---


````python
p = 1 # La función exponencial 2^n
t = 0

while p < 1000:
    t +=1 
    p = 2**t
    print(p)
    
p, t
````

## 6.5 `else` con loops



````python
# algoritmo para buscar un número en una lista con break y else
n = 5 # número a buscar
lista_donde_voy_a_buscar = list(range(0, 10, 5))
for i in lista_donde_voy_a_buscar:
    if i == n:
        print("Se encontró el", n, ", se detiene el ciclo con break")
        print("y la ubicacion del", n, "es en la posición ",\
            (lista_donde_voy_a_buscar.index(i)+1))
        break
else:
    print("El ciclo terminó normalmente (sin break) y no se encontró el número", n)
````


## 6.6 Control de loops

### `break`

Interrumpe el ciclo.



````python
for i in range(0, 10, 2):
    print(i)
    if i == 6:
        break
````


### `continue`

Salta a la siguiente iteración.


````python
for i in range(0,9+1):
    if i % 2 == 0:
        print(f"{i} es par → se omite con continue")
        continue
    
    print(f"{i} es impar → sí se ejecuta esta parte del ciclo")

````

### `pass`

Marcador sintáctico.



La sentencia `pass` se usa como “marcador de posición”: no hace nada, pero permite que la estructura del programa sea válida mientras decides qué lógica agregar después.


````python
for i in range(10):
    if i % 2 == 0:
        # Aquí luego agregaremos lógica para los números pares
        pass
    else:
        print(f"{i} es impar y se procesa normalmente")
````


**Qué muestra este ejemplo:**

* `pass` mantiene la estructura del `if` sin ejecutar ninguna acción.
* El programa sigue funcionando sin errores de sintaxis.
* Es útil cuando estás diseñando el algoritmo y aún no implementas una parte del código.

## 7.2 Aproximación numérica

Ejemplo:

* método iterativo
* sucesión



````python
x = 1  # inicializamos la variable x
while x < 10:
    x = x + 2
x
````

````python
# algoritmo para aproximar el número pi con ciclo mientras y break
pi_aproximado = 0
n = 1
while True:
    pi_aproximado += 4 * ((-1)**(n+1)) / (2*n - 1)
    n += 1
    if abs(pi_aproximado - 3.14159) < 0.00001:
        break
pi_aproximado    
````

Tu algoritmo está bien encaminado (usa la serie de Serie de Leibniz), pero tiene un problema conceptual fuerte:



- Estás comparando contra un valor fijo de $\pi$ (`3.14159`), lo cual **no es adecuado** si queremos un criterio interno de convergencia.

Lo correcto, como planteas, es usar:

$$
| \pi_n - \pi_{n-1} | < \text{tolerancia}
$$

---



# Versión mejorada (criterio entre aproximaciones)

```python


````python
# Aproximación de pi con criterio de tolerancia entre iteraciones

tolerancia = float(input("Ingrese la tolerancia: "))

pi_aproximado = 0
pi_anterior = 0
n = 1

while True:
    pi_anterior = pi_aproximado
    
    pi_aproximado += 4 * ((-1)**(n+1)) / (2*n - 1)
    
    # criterio de parada
    if abs(pi_aproximado - pi_anterior) < tolerancia:
        break
    
    n += 1

print("Aproximación de pi:", pi_aproximado)
print("Número de iteraciones:", n)
````

# Mejora conceptual (muy importante para tus estudiantes)

Este algoritmo ahora:

* No usa el valor real de $\pi$
* Usa un **criterio de convergencia numérica**
* Evalúa estabilidad de la sucesión
* Es consistente con métodos de análisis numérico

---



# Observación matemática clave

La serie de Leibniz:

$$
\pi = 4 \sum_{n=1}^{\infty} \frac{(-1)^{n+1}}{2n-1}
$$

- **Converge MUY lento**

Por ejemplo:

* Para $10^{-3}$ → miles de iteraciones
* Para $10^{-6}$ → millones

---



# Versión optimizada (evitando potencias)

Puedes mejorar eficiencia evitando `(-1)**(n+1)`:



````python
tolerancia = float(input("Ingrese la tolerancia: "))

pi_aproximado = 0
pi_anterior = 0
n = 1
signo = 1  # alterna entre +1 y -1

while True:
    pi_anterior = pi_aproximado
    
    pi_aproximado += 4 * signo / (2*n - 1)
    
    if abs(pi_aproximado - pi_anterior) < tolerancia:
        break
    
    signo *= -1
    n += 1

print("Aproximación de pi:", pi_aproximado)
print("Iteraciones:", n)
````


# Nivel alto (para Matemáticas Puras)

veamos que:

* Están trabajando con una **sucesión de sumas parciales**
* El criterio implementa:

$$
\lim_{n \to \infty} |\pi_n - \pi_{n-1}| = 0
$$

* Esto conecta con:

  * convergencia de series
  * criterio de Cauchy
  * análisis numérico

---



# Versión con gráfica de convergencia

Este script:

* Calcula las aproximaciones sucesivas
* Guarda los valores
* Grafica cómo converge hacia $\pi$

---

## Código con `matplotlib`

```python


````python
import matplotlib.pyplot as plt
import math

# Parámetros
tolerancia = float(input("Ingrese la tolerancia: "))
max_iter = 100000  # límite de seguridad

# Inicialización
pi_aproximado = 0
pi_anterior = 0
n = 1
signo = 1

# Listas para graficar
aproximaciones = []
iteraciones = []

while n <= max_iter:
    pi_anterior = pi_aproximado
    
    pi_aproximado += 4 * signo / (2*n - 1)
    
    aproximaciones.append(pi_aproximado)
    iteraciones.append(n)
    
    if abs(pi_aproximado - pi_anterior) < tolerancia:
        break
    
    signo *= -1
    n += 1

# Gráfica
plt.figure()
plt.plot(iteraciones, aproximaciones, label="Aproximación de π")
plt.axhline(y=math.pi, linestyle='--', label="π real", color='red')

plt.xlabel("Número de iteraciones")
plt.ylabel("Valor aproximado")
plt.title("Convergencia de la serie de Leibniz para π")
plt.legend()
plt.grid()
plt.savefig(r"C:\Users\marco\Downloads\convergencia_pi.png")  # Guardar la gráfica como imagen
plt.show()

# Resultados
print("Aproximación:", pi_aproximado)
print("Iteraciones:", n)
````


## Versión con `plotly` (interactiva)

Ideal para tus clases tipo laboratorio o notebooks



````python
import plotly.graph_objects as go
import math

tolerancia = float(input("Ingrese la tolerancia: "))
max_iter = 100000

pi_aproximado = 0
pi_anterior = 0
n = 1
signo = 1

aproximaciones = []
iteraciones = []

while n <= max_iter:
    pi_anterior = pi_aproximado
    
    pi_aproximado += 4 * signo / (2*n - 1)
    
    aproximaciones.append(pi_aproximado)
    iteraciones.append(n)
    
    if abs(pi_aproximado - pi_anterior) < tolerancia:
        break
    
    signo *= -1
    n += 1

# Gráfica interactiva
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=iteraciones,
    y=aproximaciones,
    mode='lines',
    name='Aproximación de π'
))

fig.add_trace(go.Scatter(
    x=iteraciones,
    y=[math.pi]*len(iteraciones),
    mode='lines',
    name='π real',
    line=dict(dash='dash')
))

fig.update_layout(
    title="Convergencia de la serie de Leibniz",
    xaxis_title="Iteraciones",
    yaxis_title="Valor",
)

fig.show()
fig.write_image(r"C:\Users\marco\Downloads\convergencia_pi.png")

print("Aproximación:", pi_aproximado)
print("Iteraciones:", n)
````
